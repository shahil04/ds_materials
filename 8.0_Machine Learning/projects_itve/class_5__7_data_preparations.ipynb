```python
name =df.columns  # store all columnms name in a list

# df["Manufacturer"].dtypes
numberical_columnss =[]

for i in range(len(name)):
    if df[name[i]].dtypes !='O':
        numberical_columnss.append(name[i])

print(numberical_columnss)
```

```python
for i in numberical_columnss:
    plt.figure(figsize=(5,3))
    sns.histplot(np.log(df[i]), kde=True, label ='skewness:%.2f'% (np.log(df[i]).skew()))
    plt.legend()
    plt.show()
```
