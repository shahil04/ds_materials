```python
# create a dataframe
import pandas as pd

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df["target"] = iris.target
df.head()
```

```python
df["target"].unique()
```

```python
# create feature and target in seprate variable
x = df.drop("target", axis=1)

y = df["target"]

# train and  test split data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# model load KNN
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

# train the model
model.fit(x_train, y_train)
```

```python
# prediction
y_pred = model.predict(x_test)
```

```python
# Accuracy check confusion_matrix, classification_report
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# simple accuracy percentage
print(accuracy_score(y_test, y_pred))
```

```python
# show the classfication data matrix
print(confusion_matrix(y_test, y_pred))
```

```python
# details report about accuracy
print(classification_report(y_test, y_pred))
```

## hyper parameter Tuning


### 1. manually finding best parameter using loops

```python
2# check the k value from 1 to 50
for i in range(1, 50):
    model2 = KNeighborsClassifier(n_neighbors=i,leaf_size=20)
    model2.fit(x_train, y_train)
    y_pred2 = model2.predict(x_test)
    print(f"K={i} Accuracy: {accuracy_score(y_test, y_pred2)}")
```

```python
# check the k value from 1 to 50
for i in range(1, 50):
    for j in range(10, 20):
      model2 = KNeighborsClassifier(n_neighbors=i,leaf_size=j)
      model2.fit(x_train, y_train)
      y_pred2 = model2.predict(x_test)
      print(f"K={i} Leaf size= {j} Accuracy: {accuracy_score(y_test, y_pred2)}")
```

### 2. using predfine methods in sklearn finding best parameter using GridSearchCV, RandomizedSearchCV

```python
# Hyper parameter tuning
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
# Define the parameter grid for KNN
model3 = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9],  # Different values for K
    'weights': ['uniform', 'distance'] # Different weighting schemes
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=model3, param_grid=param_grid, cv=5, scoring='accuracy', verbose=3)
# Fit the grid search to the data
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
# final
model5 = grid_search.best_estimator_
```
