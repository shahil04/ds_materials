### 1.Handling Outliers
An outlier is an observation that lies an abnormal distance from other values in a random sample from a population.

Handling outliers involves identifying extreme data points and deciding whether to remove, transform (like log or square root), impute (replace with mean/median/model), or use robust methods (like median/IQR) that are less sensitive, depending on their cause and relevance, to prevent distortion in analysis while preserving data integrity. The best approach depends on the dataset and analysis goals, often requiring investigation into why the outlier exists (error vs. genuine extreme).

### 2.Handling Skewness

Skewness is a key statistical measure that shows how data is spread out in a dataset. It tells us if the data points are skewed to the left (negative skew) or to the right (positive skew) in relation to the mean.

Skewness is a statistical measure that describes the asymmetry of a probability distribution, indicating whether the data is stretched more toward the left or right of the mean

```python
from sklearn.preprocessing import PowerTransformer
# Initialize transformer for Box-Cox
pt = PowerTransformer(method='box-cox')
# Box-Cox requires strictly positive values
df['total_sqft_boxcox'] = pt.fit_transform(df[['total_sqft']])

# Check skewness
df['total_sqft_boxcox'].skew()
# # Get lambda value
# lambda_total_sqft = pt.lambdas_[0]
```

```python
from scipy.stats import boxcox

df['total_sqft_boxcox'], lambda_total_sqft = boxcox(df['total_sqft']) 

df['total_sqft_boxcox'].skew()
```

```python
from scipy.stats import boxcox
# Apply Box-Cox transformation to the 'total_sqft' column
df['total_sqft_boxcox'], lambda_total_sqft = boxcox(df['total_sqft'])

# Display the skewness of the original and transformed total_sqft
print(f"Original total_sqft Skewness: {df['total_sqft'].skew()}")
print(f"Box-Cox Transformed total_sqft Skewness: {df['total_sqft_boxcox'].skew()}")
# Visualize the distribution of the transformed total_sqft
plt.Figure(figsize=(4,4))
plt.subplot(1,2,1)
sns.histplot(df['total_sqft'], kde=True)

plt.subplot(1,2,2)
sns.histplot(df['total_sqft_boxcox'], kde=True)

plt.title('Distribution of Box-Cox Transformed total_sqft')
plt.xlabel('Box-Cox Transformed Price')
plt.ylabel('Frequency')
plt.show()
```

### Feature Scaling -Normalization and Standardization

#### Scaling
- change the range of value in fixed range
- Normalization (min-max scaler) -->range (0-1)/(-1 to 1)
- Standarization (z-score) --> mean=0, std=1

```python
# x-xmin/xmax-xmin  --> manual
li = [1,2,3,4,9,10]
max(li)
min(li)

for x in li:
  new_x = (x-min(li))/(max(li)-min(li))
  print(new_x)
```

```python
# Why use robust scaling
# Handles outliers: It is robust to the presence of outliers, which can distort other scaling methods that rely on the mean and standard deviation.
# Less sensitive to skewed data: It is suitable for datasets with non-normal distributions, which are common in real-world data.
# Improves model performance:
# X_scaled=(X-X_median)/IQR)
```
