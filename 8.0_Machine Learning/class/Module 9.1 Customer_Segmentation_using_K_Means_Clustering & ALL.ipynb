## Choosing the number of clusters

###  WCSS in K-Means Clustering

##  Definition

**WCSS (Within-Cluster Sum of Squares):**
It is the **sum of the squared distances** between each data point and the **centroid** of the cluster to which it belongs.

Mathematically:
[
WCSS = \sum_{k=1}^{K} \sum_{i \in C_k} ||x_i - \mu_k||^2
]

Where:

* ( K ) = Number of clusters
* ( C_k ) = Cluster ( k )
* ( x_i ) = Data point in cluster ( C_k )
* ( \mu_k ) = Centroid of cluster ( C_k )

---

##  Why WCSS is Important

* Measures **compactness** of clusters (lower WCSS = better clustering).
* Helps in **choosing the optimal number of clusters** using the **Elbow Method**.
* Indicates **how close** data points are to their cluster center.

---

##  Real-World Example

| Scenario                  | Description                               | How WCSS Helps                                                                |
| ------------------------- | ----------------------------------------- | ----------------------------------------------------------------------------- |
|  Customer Segmentation | Grouping customers by purchasing behavior | Lower WCSS means customers in each group have **similar buying patterns**     |
|  Cab Demand Prediction  | Clustering city areas by ride demand      | Lower WCSS ensures each cluster represents **geographically close** locations |
|  Patient Clustering     | Grouping patients based on symptoms       | Lower WCSS ensures each group has **similar symptom profiles**                |

---

##  Elbow Method (Finding Optimal K)

To find the **optimal number of clusters**, we plot **WCSS vs. K (number of clusters)**:

* Initially, WCSS drops quickly as K increases.
* After a certain point, the decrease slows down and forms an **"elbow"**.
* The **elbow point** is considered the best ( K ).

---

 **Summary:**

* WCSS is a metric to measure **intra-cluster similarity**.
* It decreases as the number of clusters increases.
* Use it with the **Elbow Method** to find the best number of clusters for your K-Means model.


```python
# finding wcss value for different number of clusters
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
# plot an elbow graph
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of Clusters = 5

###  Silhouette Score in K-Means (and Clustering)

The **Silhouette Score** is a powerful metric used to evaluate how **well data points are clustered**. It tells you **how similar** an object is to its **own cluster** compared to **other clusters**.

---

##  Definition

The **Silhouette Score (S)** for a single data point is calculated as:

[
S = \frac{b - a}{\max(a, b)}
]

Where:

* ( a ) = Average distance of the point to all other points **in the same cluster** (intra-cluster distance)
* ( b ) = Average distance of the point to all points **in the nearest different cluster** (inter-cluster distance)

---

##  Score Range & Interpretation

| Silhouette Score Range | Meaning                                                       |
| ---------------------- | ------------------------------------------------------------- |
| **+1.0**               | Perfectly clustered  far away from other clusters            |
| **0.5  1.0**          | Well-clustered  data points are appropriately grouped        |
| **0.0  0.5**          | Overlapping clusters  may not be well-separated              |
| **< 0**                | Wrong clustering  point may be assigned to the wrong cluster |

---

##  Why Its Important

* Measures **cluster quality** based on both **cohesion** (how close points in a cluster are) and **separation** (how far clusters are apart).
* Helps validate **optimal number of clusters**.
* Works well with **K-Means, Hierarchical, DBSCAN**, etc.

---

##  Real-World Examples

| Scenario                  | How Silhouette Score Helps                                        | Interpretation                          |
| ------------------------- | ----------------------------------------------------------------- | --------------------------------------- |
|  Customer Segmentation | Evaluate if customers in each group have distinct buying patterns | **>0.6** means clusters are meaningful  |
|  Cab Demand Clustering | Check if geographic zones are well-separated                      | **>0.5** means each zone is distinct    |
|  Patient Grouping       | Confirm patients with similar symptoms are clustered together     | **Close to 1**  clusters are very good |

---

##  Silhouette vs WCSS  Quick Comparison

| Metric               | Meaning                                               | Best Value            | Use Case                                |
| -------------------- | ----------------------------------------------------- | --------------------- | --------------------------------------- |
| **WCSS**             | Measures cluster compactness (distance to centroid)   | Lower is better       | Used for **Elbow Method**               |
| **Silhouette Score** | Measures how well-separated and cohesive clusters are | Closer to 1 is better | Used for **Cluster Quality Validation** |

---

 **Summary:**

* Silhouette Score is a **quality measure** for clustering.
* It combines **cohesion + separation** into one metric.
* Use it alongside **WCSS** to validate the **optimal number of clusters**.

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch

from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
# Create a dendrogram to visualize the hierarchical clustering
plt.figure(figsize=(10, 7))

dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
# Apply Agglomerative Clustering
# Based on the dendrogram, let's choose 5 clusters (you can adjust this based on the dendrogram)
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
print("Cluster labels for each data point:")
print(y_hc)
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

# Apply DBSCAN
# You may need to adjust the eps and min_samples parameters based on your data
dbscan = DBSCAN(eps=5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

print("Cluster labels for each data point:")
print(y_dbscan)
```

- **Different colored points:** Each color represents a different cluster identified by DBSCAN. These are groups of customers with similar annual income and spending scores based on the eps and min_samples parameters used.

- **Purple points (label -1):** These points represent "noise" points. DBSCAN was not able to assign these customers to any specific cluster based on the defined parameters. They are considered outliers or points in sparse areas of the data.
Density-based: Unlike K-Means and Hierarchical Clustering which form spherical or hierarchical clusters, DBSCAN finds clusters based on density. This is why the clusters might have irregular shapes and there are points not assigned to any cluster.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Visualize the DBSCAN clusters
plt.figure(figsize=(8, 8))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_dbscan, palette='viridis', legend='full')
plt.title('DBSCAN Clusters')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# Add the DBSCAN cluster labels to the original DataFrame
customer_data_with_clusters_dbscan = customer_data.copy()
customer_data_with_clusters_dbscan['DBSCAN_Cluster'] = y_dbscan

# Analyze the characteristics of each cluster
cluster_characteristics_dbscan = customer_data_with_clusters_dbscan.groupby('DBSCAN_Cluster')[['Annual Income (k$)', 'Spending Score (1-100)']].agg(['mean', 'median', 'count'])

print("Characteristics of each DBSCAN Cluster:")
display(cluster_characteristics_dbscan)
```
