# CrossValidations in DT

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Define the model
model2 = DecisionTreeClassifier(criterion="entropy", max_depth=6)

# Perform cross-validation
cv_scores = cross_val_score(model2, X_train, y_train, cv=5, scoring='accuracy')

# Print the cross-validation results
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", np.mean(cv_scores))
```

#  hyperper parameter

```python
#  hyperper parameter
from sklearn.model_selection import GridSearchCV

# Define the model
model3 = DecisionTreeClassifier()

# Define the parameter grid to search
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 4, 5, 6, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=model3, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1,verbose=True)

# Fit the model
grid_search.fit(X_train, y_train)
```

```python

# Best parameters found
print("Best hyperparameters:", grid_search.best_params_)

# Best model based on grid search
best_model = grid_search.best_estimator_
```

```python

# Predict on the test set
predict = best_model.predict(X_test)

# Evaluate accuracy on the test set
test_accuracy = accuracy_score(y_test, predict)
print("Test set accuracy with best model:", test_accuracy)
```

R-squared (\( R^2 \)) is a statistical measure that represents the proportion of the variance in the dependent variable (the outcome or response variable) that is explained by the independent variables (predictors) in a regression model. It is used to evaluate how well the regression model fits the data.

### Key Points about \( R^2 \):
1. **Range of Values**: \( R^2 \) values range from 0 to 1:
   - **0** means the model explains none of the variance in the dependent variable.
   - **1** means the model explains all of the variance in the dependent variable.
   - A value between 0 and 1 indicates the proportion of the variance that is explained by the model.

2. **Interpretation**:
   - If \( R^2 = 0.8 \), it means 80% of the variance in the dependent variable is explained by the model, while the remaining 20% is unexplained (due to random noise or other factors not included in the model).
   - Higher \( R^2 \) values generally indicate a better fit, but this is not always the case. A high \( R^2 \) can also occur if the model is overfitting, i.e., it captures noise in the data as if it were a meaningful pattern.

3. **Limitations**:
   - \( R^2 \) does not imply causation; it only quantifies the fit of the model.
   - It can increase if more predictors are added to the model, even if the added predictors are not genuinely useful (this is why adjusted \( R^2 \) is often used for multiple regression).
   - It doesn't capture the possibility of a poor model fit when other diagnostic measures (like residual analysis) suggest otherwise.

### Formula for \( R^2 \):
\[
R^2 = 1 - \frac{\text{Sum of Squared Residuals (SSR)}}{\text{Total Sum of Squares (SST)}}
\]
Where:
- SSR (Sum of Squared Residuals) is the sum of squared differences between the observed and predicted values.
- SST (Total Sum of Squares) is the sum of squared differences between the observed values and their mean.

In simple terms, \( R^2 \) tells you how much better your model is at explaining the variation in the data compared to a baseline model (which simply predicts the mean of the dependent variable for all observations).
