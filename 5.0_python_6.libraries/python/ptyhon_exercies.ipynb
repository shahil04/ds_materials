### Types of Python Operators

Python offers several types of operators, each serving a unique purpose in programming. Below is a summary of the different types of Python operators:

1. **Arithmetic Operators**:
   - Used for basic mathematical operations.
   - **Operators**:
     - `+` (Addition): Adds two operands.
     - `-` (Subtraction): Subtracts the second operand from the first.
     - `*` (Multiplication): Multiplies two operands.
     - `/` (Division): Divides the first operand by the second.
     - `//` (Floor Division): Divides and returns the integer part of the quotient.
     - `%` (Modulo): Returns the remainder of the division.
     - `**` (Exponentiation): Raises the first operand to the power of the second.

2. **Assignment Operators**:
   - Used to assign values to variables.
   - **Operators**:
     - `=` (Simple Assignment): Assigns the value on the right to the variable on the left.
     - `+=` (Addition Assignment): Adds the right operand to the left operand and assigns the result to the left operand.
     - `-=` (Subtraction Assignment): Subtracts the right operand from the left operand and assigns the result to the left operand.
     - `*=` (Multiplication Assignment): Multiplies the right operand with the left operand and assigns the result to the left operand.
     - `/=` (Division Assignment): Divides the left operand by the right operand and assigns the result to the left operand.
     - `%=` (Modulo Assignment): Takes the modulus using two operands and assigns the result to the left operand.
     - `**=` (Exponentiation Assignment): Performs an exponential calculation on operands and assigns the result to the left operand.

3. **Comparison Operators**:
   - Used to compare two values or variables.
   - **Operators**:
     - `==` (Equal to): Returns True if the operands are equal.
     - `!=` (Not Equal to): Returns True if the operands are not equal.
     - `>` (Greater than): Returns True if the left operand is greater than the right.
     - `<` (Less than): Returns True if the left operand is less than the right.
     - `>=` (Greater than or Equal to): Returns True if the left operand is greater than or equal to the right.
     - `<=` (Less than or Equal to): Returns True if the left operand is less than or equal to the right.

4. **Logical Operators**:
   - Used to combine conditional statements.
   - **Operators**:
     - `and`: Returns True if both operands are true.
     - `or`: Returns True if at least one of the operands is true.
     - `not`: Returns True if the operand is false.

5. **Bitwise Operators**:
   - Operate on binary digits of integers.
   - **Operators**:
     - `&` (AND): Sets each bit to 1 if both bits are 1.
     - `|` (OR): Sets each bit to 1 if one of two bits is 1.
     - `~` (NOT): Inverts all the bits.
     - `^` (XOR): Sets each bit to 1 if only one of the two bits is 1.
     - `>>` (Right Shift): Shifts bits of the left operand to the right by the number of bits specified by the right operand.
     - `<<` (Left Shift): Shifts bits of the left operand to the left by the number of bits specified by the right operand.

6. **Special Operators**:
   - Python provides special operators for certain conditions.
   - **Identity Operators**:
     - `is`: Returns True if both operands refer to the same object.
     - `is not`: Returns True if both operands do not refer to the same object.
   - **Membership Operators**:
     - `in`: Returns True if a value is found in the specified sequence.
     - `not in`: Returns True if a value is not found in the specified sequence.

### Python `if...else` Statement - Short Notes

**1. Python `if` Statement:**
- **Purpose**: The `if` statement is used to execute a block of code only when a specified condition is met.
- **Syntax**:
  ```python
  if condition:
      # body of if statement
  ```
- **Example**:
  ```python
  number = int(input('Enter a number: '))
  if number > 0:
      print(f'{number} is a positive number.')
  ```

**2. Indentation in Python:**
- Python uses indentation (usually 4 spaces) to define blocks of code, such as within an `if` statement.
- Improper indentation leads to errors.

**3. Python `if...else` Statement:**
- **Purpose**: The `else` clause allows you to execute an alternative block of code if the condition in the `if` statement evaluates to False.
- **Syntax**:
  ```python
  if condition:
      # body of if statement
  else:
      # body of else statement
  ```
- **Example**:
  ```python
  number = int(input('Enter a number: '))
  if number > 0:
      print('Positive number')
  else:
      print('Not a positive number')
  ```

**4. Python `ifâ€¦elifâ€¦else` Statement:**
- **Purpose**: The `ifâ€¦elifâ€¦else` statement is used to evaluate multiple conditions and execute corresponding blocks of code.
- **Syntax**:
  ```python
  if condition1:
      # code block 1
  elif condition2:
      # code block 2
  else: 
      # code block 3
  ```
- **Example**:
  ```python
  number = -5
  if number > 0:
      print('Positive number')
  elif number < 0:
      print('Negative number')
  else:
      print('Zero')
  ```

**5. Python Nested `if` Statements:**
- **Purpose**: You can place an `if` statement inside another `if` statement to check multiple conditions in a nested manner.
- **Example**:
  ```python
  number = 5
  if number >= 0:
      if number == 0:
          print('Number is 0')
      else:
          print('Number is positive')
  else:
      print('Number is negative')
  ```

**6. Python Ternary Operator:**
- **Purpose**: A shorthand way to write simple `if...else` statements.
- **Syntax**:
  ```python
  result = value_if_true if condition else value_if_false
  ```

### Short Notes on Python `for` Loop

#### 1. **Introduction to Python `for` Loop**
A `for` loop in Python is used to iterate over sequences such as lists, strings, tuples, dictionaries, and sets. The loop iterates over each element of the sequence and executes the code inside the loop body for each element.

**Syntax:**
```python
for variable in sequence:
    # body of the loop
```

#### 2. **Iterating Over a List**
When using a `for` loop with a list, the loop will iterate over each item in the list in sequence.

**Example:**
```python
languages = ['Swift', 'Python', 'Go']

for lang in languages:
    print(lang)
```

**Output:**
```
Swift
Python
Go
```
- Here, the loop iterates over the list `languages` and prints each element.

#### 3. **Indentation in a `for` Loop**
Indentation defines the block of code that belongs to the loop. The body of the loop must be indented consistently. If there is no indentation, Python will throw an error.

**Example:**
```python
languages = ['Swift', 'Python', 'Go']

for lang in languages:
    print(lang)
    print('-----')

print('Last statement')
```
- All code inside the loop body is indented, and the `print('Last statement')` is outside the loop because it is not indented.

#### 4. **Iterating Over a String**
A string in Python can be iterated through character by character using a `for` loop.

**Example:**
```python
language = 'Python'

for letter in language:
    print(letter)
```

**Output:**
```
P
y
t
h
o
n
```
- Here, the loop iterates over each character of the string.

#### 5. **Using `range()` in a `for` Loop**
The `range()` function generates a sequence of numbers, which can be iterated over using a `for` loop.

**Example:**
```python
for i in range(4):
    print(i)
```

**Output:**
```
0
1
2
3
```
- `range(4)` generates the numbers 0 to 3. The loop iterates over these numbers and prints them.

#### 6. **`for` Loop with `else` Clause**
An `else` clause can be used with a `for` loop. The `else` block is executed after the loop completes its iteration over the sequence, provided that no `break` statement was encountered.

**Example:**
```python
for i in range(3):
    print(i)
else:
    print("Loop finished")
```

**Output:**
```
0
1
2
Loop finished
```

#### 7. **Using a `for` Loop Without Accessing Items**
You can use a `for` loop without directly accessing the items in the sequence. This is useful when you just need to perform an action a set number of times.

**Example:**
```python
for _ in range(3):
    print("Hello!")
```

**Output:**
```
Hello!
Hello!
Hello!
```
- The loop runs three times, but we don't use the variable (indicated by `_`).

#### 8. **Nested `for` Loops**
A `for` loop can be nested within another `for` loop. This allows you to iterate over multiple sequences simultaneously.

**Example:**
```python
for i in range(3):
    for j in range(2):
        print(f"i = {i}, j = {j}")
```

**Output:**
```
i = 0, j = 0
i = 0, j = 1
i = 1, j = 0
i = 1, j = 1
i = 2, j = 0
i = 2, j = 1
```
- The outer loop runs three times, and for each iteration of the outer loop, the inner loop runs twice.

#### 9. **Exiting a Loop Early: `break` Statement**
You can use the `break` statement to exit a loop before it finishes iterating over all the elements.

**Example:**
```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

**Output:**
```
0
1
2
```
- The loop terminates when `i` equals 3 because of the `break` statement.

#### 10. **Skipping Iterations: `continue` Statement**
The `continue` statement is used to skip the current iteration and move to the next one.

**Example:**
```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

**Output:**
```
0
1
3
4
```
- When `i` equals 2, the `continue` statement is encountered, skipping the rest of the code for that iteration.

- Python for loop with else clause
    - A for loop can have an optional else clause. This else clause executes after the iteration completes.
    - Note: The else block will not execute if the for loop is stopped by a break statement.
```
digits = [0, 1, 5]

for i in digits:
    print(i)
else:
    print("No items left.")

```

- Using for loop without accessing items
 - # using _ for placeholder variable
```

```
These are the key concepts of Python's `for` loop. It is a powerful tool for iterating over sequences in Python, making it essential for working with collections of data.

### Short Notes on Python `while` Loop

#### 1. **Introduction to Python `while` Loop**
The `while` loop in Python is used to repeatedly execute a block of code as long as a specified condition remains true.

**Syntax:**
```python
while condition:
    # body of the while loop
```
- The loop evaluates the `condition`. If it evaluates to `True`, the code inside the loop is executed. 
- After each iteration, the condition is re-evaluated. 
- The loop continues until the condition becomes `False`.

#### 2. **Example: Basic `while` Loop**
```python
number = 1

while number <= 3:
    print(number)
    number = number + 1
```

**Output:**
```
1
2
3
```
- The loop prints numbers 1 to 3 and stops when `number` becomes greater than 3.

#### 3. **Termination of `while` Loop**
To prevent an infinite loop, ensure the condition eventually evaluates to `False` by updating variables inside the loop.

**Example:**
```python
number = int(input('Enter a number: '))

while number != 0:
    print(f'You entered {number}.')
    number = int(input('Enter a number: '))

print('The end.')
```

**Output:**
```
Enter a number: 3
You entered 3.
Enter a number: 0
The end.
```
- The loop runs until the user inputs `0`, then it terminates.

#### 4. **Infinite `while` Loop**
If the loop's condition is always `True`, it forms an infinite loop.

**Example:**
```python
age = 32

while age > 18:
    print('You can vote')
```

**Output:**
```
You can vote
You can vote
... (infinite)
```

Alternatively, you can create an infinite loop explicitly:

```python
while True:
    print('You can vote')
```

#### 5. **Exiting a `while` Loop with `break`**
The `break` statement allows you to exit a loop prematurely before the condition becomes `False`.

**Example:**
```python
while True:
    number = int(input("Enter a number: "))
    if number == 0:
        break
    print(f'You entered {number}')
```
- The loop terminates when the user enters `0`.

#### 6. **`while` Loop with `else` Clause**
An `else` clause can be added to a `while` loop. It executes when the loop finishes normally (i.e., not terminated by a `break` statement).

**Example:**
```python
number = 0

while number < 3:
    print(number)
    number += 1
else:
    print("Loop finished")
```

**Output:**
```
0
1
2
Loop finished
```
- The `else` block executes after the loop completes all its iterations.

#### 7. **`for` Loop vs `while` Loop**
- **`for` loop**: Used when you know the number of iterations in advance or are iterating over a sequence (like lists, strings).
- **`while` loop**: Used when the number of iterations isn't known beforehand and depends on a condition that must be checked repeatedly.

These are the key concepts of the Python `while` loop, including its syntax, examples, and different variations such as loops with `break`, `else`, and infinite loops.
