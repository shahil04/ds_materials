```python
import os

    _id_counter = 1

    def __init__(self, name, age, marks, roll_number=None):
```

```python
# student_management_tkinter.py

import tkinter as tk
from tkinter import ttk, messagebox
import csv
import os

class Student:
    _id_counter = 1

    def __init__(self, name, age, marks, roll_number=None):
        if roll_number:
            self.roll_number = roll_number
        else:
            self.roll_number = f"S{Student._id_counter:03d}"

        self.name = name
        self.age = age
        self.marks = marks

    def to_dict(self):
        return {
            "RollNumber": self.roll_number,

            "Age": self.age,
            "Marks": self.marks
        }

class StudentManagementSystem:
    def __init__(self, master):
        self.master = master
        self.master.title("Student Management System")
        self.filename = "students.csv"
        self.students = []
        self.load_students()

        # GUI Elements
        self.create_widgets()
        self.populate_table()

    # ---------------- CSV Methods ----------------
    def load_students(self):
        if os.path.exists(self.filename):
            with open(self.filename, "r", newline="") as f:
                reader = csv.DictReader(f)
                for row in reader:
                    roll_num = row["RollNumber"]
                    num_part = int(roll_num[1:])
                    if num_part >= Student._id_counter:
                        Student._id_counter = num_part + 1
                    student = Student(row["Name"], int(row["Age"]), int(row["Marks"]), roll_number=roll_num)
                    self.students.append(student)

    def save_students(self):
        with open(self.filename, "w", newline="") as f:
            fieldnames = ["RollNumber", "Name", "Age", "Marks"]
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            for student in self.students:
                writer.writerow(student.to_dict())

    # ---------------- GUI Methods ----------------
    def create_widgets(self):
        # Labels & Entry Fields
        tk.Label(self.master, text="Name:").grid(row=0, column=0, padx=5, pady=5)
        self.name_entry = tk.Entry(self.master)
        self.name_entry.grid(row=0, column=1, padx=5, pady=5)

        tk.Label(self.master, text="Age:").grid(row=1, column=0, padx=5, pady=5)
        self.age_entry = tk.Entry(self.master)
        self.age_entry.grid(row=1, column=1, padx=5, pady=5)

        tk.Label(self.master, text="Marks:").grid(row=2, column=0, padx=5, pady=5)
        self.marks_entry = tk.Entry(self.master)
        self.marks_entry.grid(row=2, column=1, padx=5, pady=5)

        # Buttons
        tk.Button(self.master, text="Add Student", command=self.add_student).grid(row=3, column=0, columnspan=2, pady=5)
        tk.Button(self.master, text="Update Marks", command=self.update_marks).grid(row=4, column=0, columnspan=2, pady=5)
        tk.Button(self.master, text="Remove Student", command=self.remove_student).grid(row=5, column=0, columnspan=2, pady=5)
        tk.Button(self.master, text="Search Student", command=self.search_student).grid(row=6, column=0, columnspan=2, pady=5)

        # Treeview for displaying students
        self.tree = ttk.Treeview(self.master, columns=("Roll", "Name", "Age", "Marks"), show="headings")
        self.tree.heading("Roll", text="Roll No")
        self.tree.heading("Name", text="Name")
        self.tree.heading("Age", text="Age")
        self.tree.heading("Marks", text="Marks")
        self.tree.grid(row=0, column=2, rowspan=7, padx=10, pady=5)

    def populate_table(self):
        # Clear table
        for row in self.tree.get_children():
            self.tree.delete(row)
        # Add students
        for student in self.students:
            self.tree.insert("", "end", values=(student.roll_number, student.name, student.age, student.marks))

    # ---------------- Student Methods ----------------
    def add_student(self):
        name = self.name_entry.get()
        age = self.age_entry.get()
        marks = self.marks_entry.get()
        if not name or not age.isdigit() or not marks.isdigit():
            messagebox.showerror("Error", "Invalid input!")
            return
        student = Student(name, int(age), int(marks))
        self.students.append(student)
        self.save_students()
        self.populate_table()
        messagebox.showinfo("Success", f"Student {name} added!")

    def update_marks(self):
        selected = self.tree.focus()
        if not selected:
            messagebox.showwarning("Warning", "Select a student from the table!")
            return
        new_marks = self.marks_entry.get()
        if not new_marks.isdigit():
            messagebox.showerror("Error", "Invalid marks!")
            return
        roll_number = self.tree.item(selected)["values"][0]
        for student in self.students:
            if student.roll_number == roll_number:
                student.marks = int(new_marks)
                self.save_students()
                self.populate_table()
                messagebox.showinfo("Success", f"{student.name}'s marks updated!")
                return

    def remove_student(self):
        selected = self.tree.focus()
        if not selected:
            messagebox.showwarning("Warning", "Select a student from the table!")
            return
        roll_number = self.tree.item(selected)["values"][0]
        for student in self.students:
            if student.roll_number == roll_number:
                self.students.remove(student)
                self.save_students()
                self.populate_table()
                messagebox.showinfo("Success", f"{student.name} removed!")
                return

    def search_student(self):
        name = self.name_entry.get().lower()
        found = False
        for student in self.students:
            if student.name.lower() == name:
                messagebox.showinfo("Found", f"Roll: {student.roll_number}\nName: {student.name}\nAge: {student.age}\nMarks: {student.marks}")
                found = True
                break
        if not found:
            messagebox.showwarning("Not Found", "Student not found!")

# ---------------- Main Program ----------------
```

```python
            return

        if choice == '1':
            print("Result:", add(num1, num2))
        elif choice == '2':
            print("Result:", subtract(num1, num2))
        elif choice == '3':
            print("Result:", multiply(num1, num2))
        elif choice == '4':
            print("Result:", divide(num1, num2))
    else:
        print("Invalid choice. Please select from 1 to 4.")

if __name__ == "__main__":
    main()
```
