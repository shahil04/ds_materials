## Set
A Python set is an unordered collection of unique and mutable items. This means:

- Unordered: Elements within a set do not maintain a specific order. The order in which elements are added is not necessarily the order in which they are stored or retrieved.
- Unique: Sets automatically handle and remove duplicate values; each element in a set must be distinct.
- Mutable: You can add or remove elements from a set after it has been created.

#### Create a Set

```python
# 1.Using curly braces
my_set = {1, 2, 3, 4}

#2.Using the set() constructor with an iterable
data = [5, 6, 7, 8]
set2 = set(data)
```

```python
type(set2)
```

```python
# Accessing Elements in Sets
# Sets do not support indexing
# Iterating through a set using for loop
# set2[3]
for i in set2:
  print(i)
```

```python
s1 ={1,2,3,4,5}
s2 ={4,5,6,7,8}
```

## maths funcions in sets

```python
# 1. union
s1.union(s2)
# intersections
s1.intersection(s2)
# difference
s1.difference(s2)

# symmatric diffrence
s1.symmetric_difference(s2)
```

```python
s2.difference(s1)
```

## Methods in set

```python
# new data add
s1.add(34)

# deep copy
s3 = s1.copy()

# update
# add multiple value from list/tuple/set
s1.update(s2)

# pop --> remove random value
s1.pop()
```

```python
# remove --> remove specific value from set
s1.remove(5)
```

```python
# discard --> remove specific value from set  if value does not exist in set it not show error
s1.discard(8)
s1
```

```python
# clear
s1.clear()
s1
```

```python
# for permanenent delete the variable
del s1
```

- isdisjoint(other_set): Returns True if the set has no elements in common with other_set, False otherwise.

- issubset(other_set): Returns True if all elements of the calling set are present in other_set, False otherwise.

- issuperset(other_set): Returns True if all elements of other_set are present in the calling set, False otherwise.

```python

s7.issubset(s2)

s8.isdisjoint(s2)
```
