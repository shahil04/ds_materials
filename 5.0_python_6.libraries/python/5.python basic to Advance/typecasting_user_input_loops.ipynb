 # break
 the break statement is a control flow tool used to prematurely terminate a loop. When encountered within a for or while loop, the break statement immediately halts the loop's execution,

```python
for i in range(1, 11):
  print(i)
  if i == 5:
    break
```

```python
#  using break loop
for i in range(1, 4):
  print(i)
  u = input("Enter the user name ")
  p = int(input("Enter you password "))

  if user_id==u and  passwd== p:
    print("Login successfull")
    break
  else:
    print("Faild to login")

# last line
print("end loop")
```

# continue
The continue statement in Python is a control flow statement used within loops (for and while loops) to skip the rest of the current iteration and proceed to the next iteration. When continue is encountered, the loop immediately jumps to the beginning of the next cycle, effectively skipping any remaining code in the current iteration. This is useful when you want to bypass specific conditions without terminating the loop entirely.

```python
for i in range(1, 11):
  if i == 5:
    continue

  print(i)
```

```python
for i in range(1, 11):
  print("before",i)
  if i == 5:
    continue

  print(i)
  # print("hii",i)
```

## Nested for loops
A loop can also contain another loop inside it. These loops are called nested loops.

In a nested loop, the inner loop is executed once for each iteration of the outer loop.

```python
# outer loop
attributes = ['Electric', 'Fast']

cars = ['Tesla', 'Porsche', 'Mercedes']

for a in attributes:

  for c in cars:
    print(a, c)


    # this statement is outside the inner loop
  print("-----")
```
