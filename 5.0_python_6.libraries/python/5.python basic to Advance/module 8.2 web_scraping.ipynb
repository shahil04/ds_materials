```python
import gradio as gr

# -----------------------------
# In-memory "database"
# -----------------------------
students_db = {}

# -----------------------------
# CREATE
# -----------------------------
def add_student(student_id, name, age, course, email):
    if student_id in students_db:
        return "❌ Student ID already exists!"
    
    students_db[student_id] = {
        "Name": name,
        "Age": age,
        "Course": course,
        "Email": email
    }
    return "✅ Student added successfully!"

# -----------------------------
# READ
# -----------------------------
def view_students():
    if not students_db:
        return "📭 No students found."
    
    result = ""
    for sid, details in students_db.items():
        result += (
            f"ID: {sid}\n"
            f"Name: {details['Name']}\n"
            f"Age: {details['Age']}\n"
            f"Course: {details['Course']}\n"
            f"Email: {details['Email']}\n"
            "----------------------\n"
        )
    return result

# -----------------------------
# UPDATE
# -----------------------------
def update_student(student_id, name, age, course, email):
    if student_id not in students_db:
        return "❌ Student not found!"
    
    students_db[student_id] = {
        "Name": name,
        "Age": age,
        "Course": course,
        "Email": email
    }
    return "✅ Student updated successfully!"

# -----------------------------
# DELETE
# -----------------------------
def delete_student(student_id):
    if student_id not in students_db:
        return "❌ Student not found!"
    
    del students_db[student_id]
    return "🗑️ Student deleted successfully!"

# -----------------------------
# GRADIO UI
# -----------------------------
with gr.Blocks(title="Student Management System") as app:
    gr.Markdown("# 🎓 Student Management System (CRUD)")
    gr.Markdown("Manage students using **Create, Read, Update, Delete** operations")

    with gr.Tab("➕ Add Student"):
        sid = gr.Textbox(label="Student ID")
        name = gr.Textbox(label="Name")
        age = gr.Number(label="Age")
        course = gr.Textbox(label="Course")
        email = gr.Textbox(label="Email")
        add_btn = gr.Button("Add Student")
        add_output = gr.Textbox(label="Status")
        add_btn.click(add_student, [sid, name, age, course, email], add_output)

    with gr.Tab("📄 View Students"):
        view_btn = gr.Button("View All Students")
        view_output = gr.Textbox(lines=15, label="Student Records")
        view_btn.click(view_students, outputs=view_output)

    with gr.Tab("✏️ Update Student"):
        usid = gr.Textbox(label="Student ID")
        uname = gr.Textbox(label="Name")
        uage = gr.Number(label="Age")
        ucourse = gr.Textbox(label="Course")
        uemail = gr.Textbox(label="Email")
        update_btn = gr.Button("Update Student")
        update_output = gr.Textbox(label="Status")
        update_btn.click(update_student, [usid, uname, uage, ucourse, uemail], update_output)

    with gr.Tab("🗑️ Delete Student"):
        dsid = gr.Textbox(label="Student ID")
        delete_btn = gr.Button("Delete Student")
        delete_output = gr.Textbox(label="Status")
        delete_btn.click(delete_student, dsid, delete_output)

app.launch()
```
