```python
soup.find("div", class_="popularTitleTextBlock")
```

```python
soup.find("div", class_="popularTitleTextBlock").find("a")
```

```python
url = soup.find("div", class_="popularTitleTextBlock").find("a")["href"] # soup.a["href"]
```

```python
base_url = "http://yellowpages.in"
base_url+url
# store all website url in website_link variable
```

```python
all_link = soup.find_all("div", class_="popularTitleTextBlock")
```

```python
website_links=[]
base_url = "http://yellowpages.in"
for i in all_link:
    url = i.a["href"]
    website_links.append(base_url+url)

    print(base_url+url)
    print("=======================")
```

##### 3.extract the phone number

```python
soup.find_all("a", class_="businessContact")
```

```python
# extract the phone number
# businessContact
ph=[]
for i in soup.find_all("a", class_="businessContact"):
  print(i.text)
  ph.append(i.text)
```
