# ðŸ§  Understanding `if __name__ == "__main__":` in Python

---

## ðŸ”° 1. What is `__name__` in Python?

In every Python file, there is a special built-in variable called `__name__`.

* When you **run a Python file directly**, Python sets `__name__` to `"__main__"` in that file.
* When you **import a Python file into another**, Python sets `__name__` to the **filename** (module name), not `"__main__"`.

---

## âœ… 2. Syntax

```python
if __name__ == "__main__":
    # this code runs only when the file is run directly
```

---

## ðŸ” 3. Why is it Used?

It helps to:

* **Run test code** only when the file is executed directly
* **Prevent certain parts** of the code from running when the file is imported as a module
* **Organize** your code into reusable parts

---

## ðŸ”§ 4. Example 1: Direct Execution

**File: `my_script.py`**

```python
print("Top-level code in my_script.py")

def greet():
    print("Hello from greet function!")

if __name__ == "__main__":
    print("Running directly!")
    greet()
```

### â–¶ï¸ Output when you run:

```bash
python my_script.py
```

```
Top-level code in my_script.py
Running directly!
Hello from greet function!
```

---

## ðŸ”„ 5. Example 2: Imported as Module

**File: `main.py`**

```python
import my_script
```

### â–¶ï¸ Output:

```
Top-level code in my_script.py
```

âœ… Notice: `Running directly!` and `Hello from greet function!` **did not run**.

Because `my_script` was **imported**, so `__name__ == "my_script"` â€” not `"__main__"`.

| --------------------------------- | ---------------------------------------- |

| Writing code to be reused later   | âœ… Yes                                    |

| Concept                     | Explanation                                         |

## ðŸ“Œ What is Testing?

**Testing** is the process of executing code to find bugs or ensure functionality is correct.
In Python, you can automate testing using built-in and third-party frameworks.

---

## âœ… Types of Testing in Python

| Type                    | Description                                             |
| ----------------------- | ------------------------------------------------------- |
| **Unit Testing**        | Tests individual units or functions in isolation.       |
| **Integration Testing** | Tests how different units work together.                |
| **Functional Testing**  | Validates output for a given input as per requirements. |
| **Regression Testing**  | Ensures new code doesnâ€™t break existing features.       |
| **Acceptance Testing**  | Validates the system against user needs.                |

---

## ðŸ”§ Common Testing Tools in Python

| Tool/Library  | Purpose                                             |
| ------------- | --------------------------------------------------- |
| `unittest`    | Built-in unit testing framework.                    |
| `pytest`      | Popular third-party framework with rich features.   |
| `doctest`     | Tests embedded in docstrings.                       |

## 1ï¸âƒ£ `unittest` - Python Built-in Testing Framework

### âœ… Structure

```python
import unittest

class TestMath(unittest.TestCase):

    def test_add(self):
        self.assertEqual(2 + 3, 5)

    def test_subtract(self):
        self.assertEqual(5 - 3, 2)

if __name__ == '__main__':
    unittest.main()
```

### âœ… Common `unittest` Assertions

| Assertion Method       | Description                  |
| ---------------------- | ---------------------------- |
| `assertEqual(a, b)`    | Passes if `a == b`           |
| `assertNotEqual(a, b)` | Passes if `a != b`           |
| `assertTrue(x)`        | Passes if `bool(x)` is True  |
| `assertFalse(x)`       | Passes if `bool(x)` is False |
| `assertIs(a, b)`       | Passes if `a is b`           |
| `assertIsNone(x)`      | Passes if `x is None`        |
| `assertRaises(Error)`  | Checks exception raised      |

## 2ï¸âƒ£ `pytest` - Popular Python Testing Tool

### âœ… Basic Example

## 4ï¸âƒ£ Mocking in Python

Used to simulate parts of code (like APIs, DB calls) during testing.

### Example with `unittest.mock`

```python
from unittest.mock import Mock

mock_api = Mock()
mock_api.get_user.return_value = {'id': 1, 'name': 'Shahil'}

assert mock_api.get_user() == {'id': 1, 'name': 'Shahil'}
```

---

## 5ï¸âƒ£ Fixtures in `pytest`

Fixtures provide setup and teardown for tests.

```python

@pytest.fixture
def sample_data():
    return {"id": 1, "name": "Shahil"}

def test_user(sample_data):
    assert sample_data["id"] == 1
```

## ðŸ§® Parameterized Testing in `pytest`
