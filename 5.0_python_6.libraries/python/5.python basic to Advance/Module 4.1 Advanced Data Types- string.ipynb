Here’s a **comprehensive list of common Python string methods** with **definition** and **example** so you can easily teach or revise them.

---

## **🔹 Python String Methods — Definitions & Examples**

### 1. **`upper()`**

**Definition:** Returns a new string with all characters converted to uppercase.

```python
text = "hello world"
print(text.upper())  
# Output: HELLO WORLD
```

---

### 2. **`lower()`**

**Definition:** Returns a new string with all characters converted to lowercase.

```python
text = "HELLO WORLD"
print(text.lower())  
# Output: hello world
```

---

### 3. **`title()`**

**Definition:** Returns a new string where the first letter of each word is capitalized.

```python
text = "python programming"
print(text.title())  
# Output: Python Programming
```

---

### 4. **`capitalize()`**

**Definition:** Capitalizes the first character of the string and makes the rest lowercase.

```python
text = "hello world"
print(text.capitalize())  
# Output: Hello world
```

---

### 5. **`strip()`**

**Definition:** Removes leading and trailing spaces (or specified characters).

```python
text = "   hello world   "
print(text.strip())  
# Output: hello world
```

---

### 6. **`lstrip()`**

**Definition:** Removes spaces (or specified characters) from the left side.

```python
text = "   hello"
print(text.lstrip())  
# Output: hello
```

---

### 7. **`rstrip()`**

**Definition:** Removes spaces (or specified characters) from the right side.

```python
text = "hello   "
print(text.rstrip())  
# Output: hello
```

---

### 8. **`replace()`**

**Definition:** Replaces all occurrences of a substring with another string.

```python
text = "I love Java"
print(text.replace("Java", "Python"))  
# Output: I love Python
```

---

### 9. **`split()`**

**Definition:** Splits the string into a list based on a separator.

```python
text = "apple,banana,cherry"
print(text.split(","))  
# Output: ['apple', 'banana', 'cherry']
```

---

### 10. **`join()`**

**Definition:** Joins elements of an iterable into a single string with a separator.

```python
words = ["I", "love", "Python"]
print(" ".join(words))  
# Output: I love Python
```

---

### 11. **`find()`**

**Definition:** Returns the index of the first occurrence of a substring (or `-1` if not found).

```python
text = "I love Python"
print(text.find("Python"))  
# Output: 7
```

---

### 12. **`index()`**

**Definition:** Similar to `find()`, but raises an error if the substring is not found.

```python
text = "I love Python"
print(text.index("Python"))  
# Output: 7
```

---

### 13. **`count()`**

**Definition:** Returns the number of times a substring appears in the string.

```python
text = "banana"
print(text.count("a"))  
# Output: 3
```

---

### 14. **`startswith()`**

**Definition:** Checks if the string starts with the given substring.

```python
text = "Python is great"
print(text.startswith("Python"))  
# Output: True
```

---

### 15. **`endswith()`**

**Definition:** Checks if the string ends with the given substring.

```python
text = "Python is great"
print(text.endswith("great"))  
# Output: True
```

---

### 16. **`isalpha()`**

**Definition:** Returns `True` if all characters are alphabets (A–Z).

```python
text = "Hello"
print(text.isalpha())  
# Output: True
```

---

### 17. **`isdigit()`**

**Definition:** Returns `True` if all characters are digits.

```python
text = "12345"
print(text.isdigit())  
# Output: True
```

---

### 18. **`isalnum()`**

**Definition:** Returns `True` if all characters are alphanumeric (letters and numbers).

```python
text = "Python3"
print(text.isalnum())  
# Output: True
```

---

### 19. **`isspace()`**

**Definition:** Returns `True` if the string contains only whitespace characters.

```python
text = "   "
print(text.isspace())  
# Output: True
```

---

### 20. **`swapcase()`**

**Definition:** Swaps the case of each letter (uppercase → lowercase, lowercase → uppercase).

```python
text = "PyThOn"
print(text.swapcase())  
# Output: pYtHoN
```

In Python, the **`print()` function has several important arguments** that control how output appears.

---

## 🔹 `print()` Syntax

```python
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

---

## 🔑 Important Arguments Explained

### 1️⃣ **`objects`**

The values you want to print.

```python
print("Hello", 10, 3.5)
```

---

### 2️⃣ **`sep` (separator)**

Defines what goes **between** multiple values.

```python
print("A", "B", "C", sep=", ")
```

**Output:**

```
A, B, C
```

---

### 3️⃣ **`end`**

Defines what is printed **at the end** (default is a newline).

```python
print("Hello", end=" ")
print("World")
```

**Output:**

```
Hello World
```

---

### 4️⃣ **`file`**

Specifies where to print the output (default is the screen).

```python
with open("output.txt", "w") as f:
    print("Hello file", file=f)
```

---

### 5️⃣ **`flush`**

Forces the output buffer to flush immediately.

```python
print("Loading...", flush=True)
```

Useful for real-time logs or progress updates.

---

## 📌 Common Examples

### Print without newline

```python
print("Hello", end="")
print("World")
```

### Custom separator

```python
print(1, 2, 3, 4, sep=" | ")
```

---

## ✅ Most Used Arguments

| Argument | Purpose                  |
| -------- | ------------------------ |
| `sep`    | Separator between values |
| `end`    | Line ending              |
| `file`   | Output destination       |
| `flush`  | Immediate output         |
