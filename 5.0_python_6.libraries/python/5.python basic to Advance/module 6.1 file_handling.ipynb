# File Handling

File handling in Python refers to the process of interacting with files on a computer's file system, allowing for operations such as creating, opening, reading, writing, appending, renaming, and deleting files.

##### Key Concepts and Operations:

1. Opening a File:

The open() function is used to open a file, returning a file object (also known as a file handle).
```
Syntax: file_object = open(filename, mode)

filename: The name or path of the file to open.
```

mode: Specifies the purpose of opening the file (e.g., 'r' for read, 'w' for write, 'a' for append, 'x' for exclusive creation, 'b' for binary, 't' for text, '+' for read and write).

**File Modes:**
- 'r' (Read): Opens for reading (default).
- 'w' (Write): Opens for writing, creates the file if it doesn't exist, and truncates (empties) it if it does.
- 'a' (Append): Opens for appending, creates the file if it doesn't exist, and adds content to the end if it does.
- 'x' (Exclusive Creation): Creates a new file, but raises an error if the file already exists.
- 't' (Text): Opens in text mode (default for text files).
- 'b' (Binary): Opens in binary mode (for non-text files like images).

- Adding '+' to any mode (e.g., 'r+', 'w+') allows both reading and writing.
Reading from a File:

- read(size): Reads size bytes (or characters in text mode) from the file.
- readline(): Reads a single line from the file.
- readlines(): Reads all lines from the file and returns them as a list of strings.

**Writing to a File:**
- write(string): Writes a string to the file.
- writelines(list_of_strings): Writes a list of strings to the file.

**Closing a File:**

- file_object.close(): Closes the opened file, releasing system resources.
- Using with statement: The preferred method for opening files as it automatically handles file closing, even if errors occur.

```

        with open("example.txt", "r") as file:
            content = file.read()
        # File is automatically closed here
Example:
Python

# Writing to a file
with open("my_file.txt", "w") as f:
    f.write("Hello, Python!\n")
    f.write("This is a new line.")

# Reading from a file
with open("my_file.txt", "r") as f:
    content = f.read()
    print(content)

# Appending to a file
with open("my_file.txt", "a") as f:
    f.write("\nAppending more text.")

```

## pickle module pickling and unpickling

The pickle module in Python provides a way to serialize and deserialize Python object structures. This process is often referred to as "pickling" and "unpickling," respectively.

**Pickling (Serialization):**

â€¢ This is the process of converting a Python object (e.g., a list, dictionary, custom class instance) into a byte stream.
â€¢ This byte stream can then be stored in a file, transmitted over a network, or otherwise persisted.
â€¢ The pickle.dump() function is used to write a pickled object to a file-like object.
â€¢ The pickle.dumps() function returns the pickled object as a bytes object.

**Unpickling (Deserialization):**

â€¢ This is the reverse process of reconstructing the original Python object from its byte stream representation.
â€¢ The pickle.load() function reads a pickled object from a file-like object.
â€¢ The pickle.loads() function reconstructs an object from a bytes object.


**Key Features and Considerations:**

â€¢ Object Preservation: pickle aims to preserve the exact state of the object, including its data and structure, allowing for faithful reconstruction.
â€¢ Python-Specific: pickle is a Python-native serialization format, meaning pickled objects generally cannot be unpickled in other programming languages.

â€¢ Security Warning: Unpickling data from untrusted or unauthenticated sources is highly discouraged due to potential security risks, as malicious pickle data can execute arbitrary code during deserialization.
â€¢ Use Cases: Commonly used for saving trained machine learning models, storing application state across sessions, or transferring complex Python objects between processes within a Python environment.

```
Example Usage:
import pickle

# An object to pickle
data = {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Pickling (serializing) to a file
with open('data.pkl', 'wb') as f:
    pickle.dump(data, f)

# Unpickling (deserializing) from a file
with open('data.pkl', 'rb') as f:
    loaded_data = pickle.load(f)

print(loaded_data)

```
