```python
# account_open funtion
#  2. deposit
# 3. withdraw
# 4. show balance
amount =0
def account_open(name, ph, account_number, balance=0):
    global amount 
    amount = balance
    return name ,account_number, amount

def deposit(account_num, amt):
    global amount 
    res = amount
    amount= res+amt
    print(f"ac_no {account_num} has deposit {amt} , now Total amount is {amount}")
    return  amount

def withdraw(account_num, amt):
    global amount 
    res = amount
    amount = res-amt
    print(f"ac_no {account_num} has withdraw {amt} , now Total amount is {amount}")
    return  amount

def balance_show(account_num):
    return f" total Amount  is {amount}"
```
