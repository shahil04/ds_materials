# Operators
Operators are special symbols that instruct the Python interpreter to perform mathematical, logical, or relational operations on operands (variables and values). 

* Arithmetic: Perform math operations (+, -, *, /, //, %, **). 
<!-- * Comparison: Compare values (==, !=, &gt;, &lt;, &gt;=, &lt;=).  -->
* Logical: Evaluate combined boolean expressions (and, or, not). 
* Assignment: Assign values to variables (=, +=, -=, *=, etc.). 
* Membership: Check if a value exists in a sequence (in, not in).

### * * Logical: Evaluate combined boolean expressions (and, or, not). 

```python
# 
balance = 1500
withdrawal_amount = 600
atm_card = True

balance >= withdrawal_amount and atm_card # Check if balance is sufficient and card is present
#and operator returns True only if both conditions are True
```

```python
is_weekend = False
is_holiday = True

print(is_weekend or is_holiday) # or operator returns True if at least one condition is True
```

```python
not is_weekend   #not operator reverses the boolean value, so it returns True if is_weekend is False
```

| Operator | Meaning                             | Example           |
| -------- | ----------------------------------- | ----------------- |
| `and`    | Both conditions must be true        | `x > 0 and y > 0` |
| `or`     | At least one condition must be true | `x > 0 or y > 0`  |
| `not`    | Reverses True/False                 | `not x > 0`       |

Assignment: Assign values to variables

| Operator | Example   | Equivalent To |
| -------- | --------- | ------------- |
| `=`      | `x = 5`   | Assign 5 to x |
| `+=`     | `x += 3`  | `x = x + 3`   |
| `-=`     | `x -= 3`  | `x = x - 3`   |
| `*=`     | `x *= 3`  | `x = x * 3`   |
| `/=`     | `x /= 3`  | `x = x / 3`   |
| `//=`    | `x //= 3` | `x = x // 3`  |
| `%=`     | `x %= 3`  | `x = x % 3`   |
| `**=`    | `x **= 3` | `x = x ** 3`  |

```python
# Membership: Check if a value exists in a sequence (in, not in). 
# 
fruits = ["apple", "banana", "mango"]

print("banana" in fruits)
```

in → Checks if a value exists.

not in → Checks if a value does not exist.

# identity: Check if two variables refer to the same object (is, is not).


| Operator | Meaning                                                     |
| -------- | ----------------------------------------------------------- |
| `is`     | Returns `True` if both variables refer to the same object   |
| `is not` | Returns `True` if both variables refer to different objects |
