# ğŸ“˜ Python `sys` Module â€“ Detailed Notes with Use Cases & Examples

---

## ğŸ”° What is the `sys` Module?

The **`sys` module** in Python provides access to variables, functions, and tools used to interact **directly with the Python interpreter and system-level operations**.

ğŸ“Œ It is a **built-in module** â€“ no need to install it separately.

```python
import sys
```

---

## ğŸ§¾ Key Uses of `sys` Module

* Access **command-line arguments**
* **Exit** a Python program
* Work with **Python runtime environment**
* Get the **Python version**
* Modify the **module search path**
* Redirect **standard input/output**
* Handle **exceptions info**

---

## âœ… Commonly Used `sys` Functions & Variables

---

### 1. `sys.argv` â€“ Command-Line Arguments

ğŸ“Œ Returns a list of command-line arguments passed to the script.

```python
# script.py
import sys
print("Script name:", sys.argv[0])
print("Arguments:", sys.argv[1:])
```

**Use Case:** CLI-based programs or tools like parsers.

---

### 2. `sys.exit([status])` â€“ Exit from Program

ğŸ“Œ Exits the program. You can pass an optional exit code (`0` = success, non-zero = error).

```python
import sys
if some_condition:
    sys.exit("Something went wrong!")
```

**Use Case:** Terminate script early if error or invalid input is found.

---

### 3. `sys.version` â€“ Python Version String

ğŸ“Œ Returns full Python version (as string).

```python
import sys
print(sys.version)
```

**Use Case:** Compatibility check (for example, between Python 3.6 vs 3.10).

---

### 4. `sys.version_info` â€“ Version as Tuple

ğŸ“Œ Returns version as tuple-like object.

```python
if sys.version_info < (3, 6):
    print("Please upgrade your Python version!")
```

---

### 5. `sys.path` â€“ Module Search Paths

ğŸ“Œ A list of directories that Python searches to import modules.

```python
import sys
print(sys.path)
```

You can **append custom directories** to `sys.path`:

```python
sys.path.append('/my/custom/module/path')
```

**Use Case:** Importing modules from outside the project folder.

---

### 6. `sys.platform` â€“ Get OS Type

ğŸ“Œ Tells you the current operating system.

```python
print(sys.platform)  # 'win32', 'linux', 'darwin' (for macOS)
```

**Use Case:** Write platform-specific code.

---

### 7. `sys.stdin`, `sys.stdout`, `sys.stderr`

ğŸ“Œ Standard input/output/error stream objects.

```python
sys.stdout.write("Hello\n")
sys.stderr.write("This is an error!\n")
```

**Use Case:** Redirect output in CLI tools, custom logging, or automation.

---

### 8. `sys.getsizeof(object)` â€“ Size in Bytes

ğŸ“Œ Returns the memory size of an object in bytes.

```python
a = [1, 2, 3]
print(sys.getsizeof(a))  # Size of list in memory
```

**Use Case:** Optimize memory in data processing.

---

### 9. `sys.modules` â€“ Loaded Modules Dictionary

ğŸ“Œ Shows all currently loaded modules.

```python
print(sys.modules.keys())
```

**Use Case:** Debugging module dependencies.

---

### 10. `sys.exc_info()` â€“ Exception Details

ğŸ“Œ Returns info about the last exception (type, value, traceback).

```python
import sys
try:
    1 / 0
except:
    exc_type, exc_value, exc_traceback = sys.exc_info()
    print("Exception type:", exc_type)
    print("Exception message:", exc_value)
```

**Use Case:** Advanced exception handling/logging.

# ğŸ“˜ Python Time, Date & Datetime Modules â€“ Full Notes

---

## ğŸ”° Overview

| Module     | Purpose                                                  |
| ---------- | -------------------------------------------------------- |
| `time`     | Deals with **Unix timestamps**, delays, sleep            |
| `datetime` | Works with **dates and times**, formatting, calculations |
| `date`     | Submodule of `datetime` (only date part)                 |

---

## âœ… 1. `time` Module

ğŸ“Œ The `time` module provides access to **time-related functions** based on the system clock and **Unix epoch** (January 1, 1970).

```python
import time
```

### ğŸ”¹ Common Functions

| Function           | Description                             |
| ------------------ | --------------------------------------- |
| `time.time()`      | Returns current time in seconds (float) |
| `time.sleep(n)`    | Delays execution for `n` seconds        |
| `time.ctime()`     | Converts time to readable string        |
| `time.gmtime()`    | UTC time tuple                          |
| `time.localtime()` | Local time tuple                        |
| `time.strftime()`  | Format time as string                   |
| `time.strptime()`  | Parse string to time struct             |

### ğŸ”¸ Examples

```python
import time

print(time.time())            # Unix timestamp
print(time.ctime())           # 'Tue Jul  2 18:00:00 2025'
time.sleep(2)                 # Pause 2 seconds
print(time.strftime("%Y-%m-%d %H:%M:%S"))
```

---

## âœ… 2. `datetime` Module

ğŸ“Œ `datetime` is the most powerful and commonly used module for working with **dates and times**.

```python
import datetime
```

### ğŸ”¹ Classes in `datetime` module

| Class       | Purpose                               |
| ----------- | ------------------------------------- |
| `date`      | Represents a date (Y-M-D)             |
| `time`      | Represents a time (H\:M\:S)           |
| `datetime`  | Combination of date and time          |
| `timedelta` | Difference between two dates or times |

---

### ğŸ”¸ `datetime.date`

```python
from datetime import date

today = date.today()
print(today)                # 2025-07-02
print(today.year)           # 2025
print(today.month)          # 7
print(today.day)            # 2
```

---

### ğŸ”¸ `datetime.time`

```python
from datetime import time

t = time(14, 30, 45)
print(t)             # 14:30:45
print(t.hour)        # 14
```

---

### ğŸ”¸ `datetime.datetime`

```python
from datetime import datetime

now = datetime.now()
print(now)                       # 2025-07-02 18:05:00
print(now.strftime("%A"))        # Day of the week (e.g., Tuesday)
```

#### ğŸ“Œ Custom DateTime

```python
dt = datetime(2025, 12, 25, 10, 30)
print(dt.strftime("%Y-%m-%d %I:%M %p"))  # 2025-12-25 10:30 AM
```

---

### ğŸ”¸ `datetime.timedelta`

Used to **add or subtract time**.

```python
from datetime import timedelta, datetime

today = datetime.now()
tomorrow = today + timedelta(days=1)
print("Tomorrow:", tomorrow)

diff = timedelta(days=10)
print("10 days ago:", today - diff)
```

---

### ğŸ”¹ `datetime.strptime()` â€“ String to datetime

```python
date_string = "2025-07-02"
dt_obj = datetime.strptime(date_string, "%Y-%m-%d")
print(dt_obj)  # 2025-07-02 00:00:00
```

---

### ğŸ”¹ `datetime.strftime()` â€“ datetime to String

```python
now = datetime.now()
print(now.strftime("%d-%b-%Y %I:%M %p"))  # 02-Jul-2025 06:10 PM
```

A **virtual environment** is an **isolated workspace** that allows you to manage project-specific Python dependencies **without affecting your global Python installation**.

ğŸ“Œ It creates a self-contained directory with:

* Its own `python` executable
* Its own installed packages via `pip`

---

## ğŸ¯ Why Use Virtual Environments?

| Reason                         | Benefit                                            |
| ------------------------------ | -------------------------------------------------- |
| Avoid version conflicts        | Different projects need different package versions |
| Keeps global environment clean | Avoid clutter in system Python                     |
| Reproducibility                | Easy to recreate using `requirements.txt`          |
| Collaboration                  | Ensures all team members use same dependencies     |

# ğŸ² Python `random` Module â€“ Complete Notes

---

## ğŸ”° What is the `random` Module?

The **`random` module** in Python provides tools to **generate random numbers**, **shuffle items**, **pick random elements**, and simulate randomness for games, simulations, testing, etc.

ğŸ“Œ It uses **pseudo-random numbers** generated using a seed.

```python
import random
```

---

## ğŸ”¹ 1. Basic Random Number Generation

| Function                 | Description                           |
| ------------------------ | ------------------------------------- |
| `random.random()`        | Returns float between `0.0` to `<1.0` |
| `random.uniform(a, b)`   | Float in range `[a, b]`               |
| `random.randint(a, b)`   | Integer in range `[a, b]`             |
| `random.randrange(a, b)` | Random int from range `[a, b)`        |
| `random.seed(n)`         | Sets the seed for reproducibility     |

### âœ… Examples

```python
import random

print(random.random())         # e.g. 0.7353
print(random.uniform(10, 20))  # e.g. 16.25
print(random.randint(1, 6))    # e.g. 4
print(random.randrange(100, 200, 10))  # e.g. 130
```

---

## ğŸ”¹ 2. Random Choice from List / Sequence

| Function                   | Description                                   |
| -------------------------- | --------------------------------------------- |
| `random.choice(seq)`       | Randomly select one element                   |
| `random.choices(seq, k=n)` | Randomly pick n elements **with replacement** |
| `random.sample(seq, k=n)`  | Pick n elements **without replacement**       |
| `random.shuffle(seq)`      | Shuffle the original list (in-place)          |

### âœ… Examples

```python
items = ['apple', 'banana', 'mango', 'cherry']

print(random.choice(items))             # One item
print(random.choices(items, k=2))       # e.g. ['banana', 'banana']
print(random.sample(items, k=2))        # e.g. ['cherry', 'apple']
random.shuffle(items)
print(items)                            # List shuffled
```

## ğŸ”¹ 3. Seeding Randomness

Use `random.seed(n)` to make random behavior **predictable** (useful for testing).

```python
random.seed(42)
print(random.random())  # Always gives same output for same seed
```
