## Recursive Functions:
 These are functions that call themselves to solve a problem. They are often used when a problem can be broken down into smaller, similar sub-problems. A base case is required to prevent infinite recursion.

A recursive function in Python is a function that calls itself during its execution. This technique is used to solve problems that can be broken down into smaller, self-similar subproblems.
Key components of a recursive function:

Base Case:
This is a condition that terminates the recursion. Without a base case, the function would call itself indefinitely, leading to a RecursionError (stack overflow). The base case provides a direct solution to the simplest instance of the problem.

Recursive Case:
This part of the function involves calling the function itself with a modified input, moving closer to the base case. The problem is broken down into a smaller version of the same problem.

### Variable Scope - Global vs Local

variables are categorized as either local or global based on their scope, which dictates where they can be accessed and modified within a program.
- Local Variables:
Definition:
Local variables are defined inside a function or a block of code.

- Scope:
Their scope is limited to the function or block in which they are defined. They are only accessible within that specific function or block.

- Lifetime:
They are created when the function is called and are destroyed when the function finishes execution.

- Encapsulation:
They promote encapsulation by keeping data isolated within functions, reducing the risk of unintended side effects in other parts of the program

##### Global Variables:
- Definition:
Global variables are defined outside of any function or class, typically at the top level of a script.
- Scope:
Their scope spans the entire program. They can be accessed and, with the global keyword, modified from anywhere in the code, including inside functions.
- Lifetime:
They are created when the program starts and remain in memory until the program terminates.
- Modification within Functions:
To modify a global variable from within a function, you must explicitly declare it as global inside that function using the global keyword. If you assign a new value to a variable with the same name as a global variable inside a function without using global, you are creating a new local variable with that name, not modifying the global one.

````python
global_var = "I am global"

def modify_global():
    global global_var  # Declare intent to modify the global variable
    global_var = "I have been modified globally"

def access_global():
    print(global_var)

access_global()
modify_global()
access_global()
````

#### Nonlocal Variables (in Nested Functions):
Nonlocal variables are used in the context of nested functions (where one function is defined inside another).

The nonlocal keyword allows an inner function to modify a variable that belongs to its immediate enclosing (outer) function's scope, but is not a global variable.
