### The map() function
in Python is a built-in function used to apply a specified function to each item in an iterable (such as a list, tuple, or string) and return an iterator (a map object) with the results.

````python
# map()
# map(function, iterable, ...) syntax
data= ["male","female","male","female","Female","female"]
````

````python
# create custom funtion for mapping
def gender(v):
  if v == "male":
    return 1
  else:
    return 0
````

````python
# change gender function into lambda function
g = lambda v : 1 if v == "male" else 0
````

````python
g("f")
````

````python
# traditional way of map value
for i in data:
  print(gender(i))
````

````python
data
````

````python
# use map funtion and return as tuple
tuple(map(gender,data))
````

````python
d = list(map(gender,data))
````

````python
print(d)
````

````python
# using lambda function
list(map(lambda v : 1 if v == "male" else 0 ,data ))
````

### The filter() function
in Python is a built-in function used to construct an iterator from elements of an iterable for which a function returns true. It offers a concise way to filter out elements from sequences like lists, tuples, or sets based on a specified condition.

````python
# filter
# filter(function, iterable)
nums =[5,4,65,767,232,32]

def even(num):
  if num%2==0:
    return num

# lambda num : num if num%2==0 else None
````

````python
list(filter(even ,nums))
````

````python
list(filter(lambda num : num if num%2!=0 else None ,nums))
````

### The reduce() function
in Python is a functional programming tool that applies a given function cumulatively to the items of an iterable, reducing the iterable to a single value. It is found within the functools module and needs to be imported explicitly.

````python
# Syntax:
from functools import reduce
# reduce(function, iterable[, initializer])
````

````python
# reduce function
numbers = [1, 2, 3, 4, 5]

def total(n,t):
  t =t+n
  return t

reduce(total,numbers)

# lambda n,t : t+n
````

````python
reduce(lambda n,t : n+t ,numbers)
````

###â€¢ Generator Functions:
These are special types of functions that return an iterator (generator object) using the yield keyword. They are memory-efficient as they generate values on the fly rather than storing them all in memory at once.

    def count_up_to(n):
        i = 1
        while i <= n:
            yield i
            i += 1


#### Key Differences and Benefits:
- Implementation: Iterators are typically implemented using classes with __iter__ and __next__ methods, while generators are implemented using functions with the yield keyword.
- Memory Efficiency: Generators are more memory-efficient as they generate values on demand (lazy evaluation) and do not store the entire sequence in memory. This is particularly beneficial for large or infinite sequences.
- Code Simplicity: Generators often lead to more concise and readable code compared to writing custom iterator classes.
- State Management: Generators automatically handle saving and restoring the function's state between yield calls.

````python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
print(next(gen)) # Output: 1
````

````python
print(next(gen)) # Output: 2
````

### Dunder methods,
also known as magic methods or special methods, are a core feature of Python's object-oriented programming. These methods are distinguished by their naming convention, which involves double leading and trailing underscores

(e.g., __init__, __str__, __add__).

## Key characteristics and uses of dunder methods:
* Automatic invocation: Dunder methods are not typically called directly by the programmer. Instead, the Python interpreter automatically invokes them in response to specific actions or operations. For instance, when you use the + operator to add two objects, Python internally calls the __add__ dunder method of the first object.
* Operator overloading: A primary use of dunder methods is to enable operator overloading. This allows you to define how standard Python operators (like +, -, *, ==, []) behave when applied to instances of your custom classes.
Customizing built-in functions: Dunder methods also allow you to customize the behavior of built-in functions like len(), str(), repr(), int(), and float() when applied to your objects.
* Object lifecycle management: Methods like __init__ (constructor) and __del__ (destructor) are dunder methods that control the creation and destruction of objects.
* Protocol implementation: Dunder methods define various protocols that objects can adhere to, such as the sequence protocol (e.g., __len__, __getitem__), the numeric protocol (e.g., __add__, __sub__), and the context manager protocol (e.g., __enter__, __exit__).

Examples of common dunder methods:
- __init__(self, ...): Initializes a new instance of a class.
- __str__(self): Returns a user-friendly string representation of an object.
- __repr__(self): Returns an unambiguous, developer-focused string representation of an object, often used for debugging.
- __add__(self, other): Defines the behavior of the + operator.
- __len__(self): Defines the behavior of the len() function.




#  Decorators

````python
def outer_function():
    enclosing_var = "Hello" # Variable in the enclosing scope

    def inner_function():
        nonlocal enclosing_var # Declare intent to modify the enclosing variable
        enclosing_var = "World"
        print(f"Inside inner function: {enclosing_var}")

    inner_function()
    print(f"Inside outer function (after inner function call): {enclosing_var}")

outer_function()
````

````python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Execution time: {end - start:.4f} seconds")
        return result
    return wrapper


````

````python
@timer
def slow_function():
    time.sleep(2)
    print("Finished slow task")
````

## âœ… What is a Decorator?

A **decorator** in Python is a **function that takes another function as input, adds some functionality, and returns it** â€” without modifying the original function's source code.

ðŸ“Œ **Use Case:** Decorators are used to **extend** the behavior of a function **without changing its code** â€” e.g., logging, timing, authentication.

---

## ðŸ”¹ Decorators = Higher-Order Functions + Closures

To understand decorators, you need to know two concepts:

### 1. **Higher-Order Function**

A function that **takes another function as argument** or **returns a function**.

```python
def greet(name):
    return f"Hello {name}"

def call_func(func):
    return func("Alice")

print(call_func(greet))  # Hello Alice
```

### 2. **Closure**

A function defined inside another function that **remembers** the outer function's variables.

```python
def outer():
    msg = "Hello"

    def inner():
        print(msg)

    return inner

fn = outer()
fn()  # Hello
```

---

## âœ… Basic Decorator Syntax

```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

### Output:

```
Before the function runs
Hello!
After the function runs
```

> `@my_decorator` is just shorthand for:

```python
say_hello = my_decorator(say_hello)
```

## âœ… Decorators with `functools.wraps`

Use `functools.wraps` to preserve the original function's metadata (like `__name__`, `__doc__`).

```python
from functools import wraps

def decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Before")
        return func(*args, **kwargs)
    return wrapper
```

### ðŸ”¹ **Decorator with Parameters (Decorator Factory)**
