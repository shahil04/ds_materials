# ğŸš€ Part 2: Student Management System Project (with OOP + Streamlit)

---

## ğŸ”¹ Step 1: OOP Student Classes

```python
class Student:
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def to_dict(self):
        return {"Name": self.name, "Age": self.age, "Marks": self.marks}
```

---

## ğŸ”¹ Step 2: Streamlit App Skeleton

```python
import streamlit as st
import pandas as pd

# Initialize storage
if "students" not in st.session_state:
    st.session_state.students = []

st.title("ğŸ“ Student Management System")
menu = st.sidebar.radio("Menu", ["Add Student", "View Students", "Update Marks", "Analytics"])
```

---

## ğŸ”¹ Step 3: Add Student Page

```python
if menu == "Add Student":
    st.subheader("Add New Student")
    name = st.text_input("Enter Name")
    age = st.number_input("Enter Age", 5, 100)
    marks = st.slider("Enter Marks", 0, 100, 50)

    if st.button("Add Student"):
        st.session_state.students.append({"Name": name, "Age": age, "Marks": marks})
        st.success(f"Student {name} added successfully!")
```

---

## ğŸ”¹ Step 4: View Students Page

```python
elif menu == "View Students":
    st.subheader("All Students")
    if st.session_state.students:
        df = pd.DataFrame(st.session_state.students)
        st.dataframe(df)
    else:
        st.warning("No students added yet!")
```

---

## ğŸ”¹ Step 5: Update Marks

```python
elif menu == "Update Marks":
    st.subheader("Update Student Marks")
    if st.session_state.students:
        names = [s["Name"] for s in st.session_state.students]
        student_name = st.selectbox("Select Student", names)
        new_marks = st.slider("New Marks", 0, 100, 50)

        if st.button("Update"):
            for s in st.session_state.students:
                if s["Name"] == student_name:
                    s["Marks"] = new_marks
            st.success(f"{student_name}'s marks updated to {new_marks}")
    else:
        st.warning("No students to update!")
```

---

## ğŸ”¹ Step 6: Analytics

```python
elif menu == "Analytics":
    st.subheader("Analytics Dashboard")
    if st.session_state.students:
        df = pd.DataFrame(st.session_state.students)
        st.bar_chart(df.set_index("Name")["Marks"])
        st.write("Average Marks:", df["Marks"].mean())
    else:
        st.warning("No data available!")
```
