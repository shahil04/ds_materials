Connecting to a database in Python primarily involves using specific libraries or modules tailored to the database system in question. The general steps and common methods are outlined below:

1. Using Database-Specific Connectors (DB-API 2.0 Compliant):
Most relational databases have dedicated Python connectors that adhere to the Python DB-API 2.0 specification. This ensures a consistent interface for database interactions across different systems.

- Installation: Install the relevant connector using pip.
Code

 ```   pip install mysql-connector-python # For MySQL
    pip install psycopg2-binary       # For PostgreSQL
    pip install cx_Oracle             # For Oracle
    pip install pymssql               # For SQL Server
    ```
Connection and Operations.
```
Python

    import mysql.connector # Example for MySQL

    try:
        connection = mysql.connector.connect(
            host="localhost",
            user="your_user",
            password="your_password",
            database="your_database"
        )
        cursor = connection.cursor()
        cursor.execute("SELECT * FROM your_table")
        results = cursor.fetchall()
        # Process results
        connection.commit() # For DML operations
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        if 'connection' in locals() and connection.is_connected():
            cursor.close()
            connection.close()
```

2. Using ORMs (Object-Relational Mappers):
ORMs like SQLAlchemy provide an abstraction layer over raw SQL, allowing interaction with databases using Python objects and methods. This can simplify development and improve code readability. installation.

``` Code

    pip install SQLAlchemy
Connection and Operations (SQLAlchemy example).
Python

    from sqlalchemy import create_engine, text

    # Define connection string (e.g., for PostgreSQL)
    DATABASE_URL = "postgresql://user:password@host:port/database"
    engine = create_engine(DATABASE_URL)

    with engine.connect() as connection:
        result = connection.execute(text("SELECT * FROM your_table"))
        for row in result:
            print(row)
        connection.commit()

```
3. Built-in SQLite3 Module:
Python includes a built-in module for interacting with SQLite databases, which are file-based and do not require a separate server installation. Connection and Operations.
Python
```
    import sqlite3

    conn = sqlite3.connect('your_database.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT)")
    cursor.execute("INSERT INTO users (name) VALUES (?)", ("Alice",))
    conn.commit()
    conn.close()
```
* Key Considerations:

- Credentials:
Securely manage database credentials (e.g., environment variables, configuration files).

- Error Handling:
Implement robust error handling using try-except-finally blocks to manage connection issues and ensure resources are closed.

- Resource Management:
Always close database connections and cursors after use to release resources.

- Transactions:
Use commit() for persistent changes and rollback() to revert changes if an error occurs during a transaction.

## ‚úÖ Step 3: `app.py` ‚Äì Flask API Code

```python
from flask import Flask, request, jsonify
import mysql.connector

app = Flask(__name__)

# Database connection
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="yourpassword",  # <-- Change this
    database="inventorydb"
)
cursor = db.cursor(dictionary=True)

# -------------------------------
# ROUTES
# -------------------------------

# GET all items
@app.route('/items', methods=['GET'])
def get_items():
    cursor.execute("SELECT * FROM inventory")
    items = cursor.fetchall()
    return jsonify(items)

# GET item by id
@app.route('/items/<int:item_id>', methods=['GET'])
def get_item(item_id):
    cursor.execute("SELECT * FROM inventory WHERE id = %s", (item_id,))
    item = cursor.fetchone()
    if item:
        return jsonify(item)
    return jsonify({"message": "Item not found"}), 404

# ADD new item
@app.route('/items', methods=['POST'])
def add_item():
    data = request.get_json()
    name = data['name']
    quantity = data['quantity']
    price = data['price']
    cursor.execute("INSERT INTO inventory (name, quantity, price) VALUES (%s, %s, %s)", (name, quantity, price))
    db.commit()
    return jsonify({"message": "Item added successfully!"}), 201

# UPDATE item
@app.route('/items/<int:item_id>', methods=['PUT'])
def update_item(item_id):
    data = request.get_json()
    name = data['name']
    quantity = data['quantity']
    price = data['price']
    cursor.execute("UPDATE inventory SET name=%s, quantity=%s, price=%s WHERE id=%s", (name, quantity, price, item_id))
    db.commit()
    return jsonify({"message": "Item updated successfully!"})

# DELETE item
@app.route('/items/<int:item_id>', methods=['DELETE'])
def delete_item(item_id):
    cursor.execute("DELETE FROM inventory WHERE id = %s", (item_id,))
    db.commit()
    return jsonify({"message": "Item deleted successfully!"})

# -------------------------------
# Run server
# -------------------------------
if __name__ == '__main__':
    app.run(debug=True)
```

from fastapi import FastAPI, HTTPException
from models import User
from database import get_connection

app = FastAPI()

@app.get("/")
def home():
    return {"message": "FastAPI + MySQL is working!"}

@app.post("/users")
def create_user(user: User):
    db = get_connection()
    cursor = db.cursor()
    cursor.execute("INSERT INTO users (name, email) VALUES (%s, %s)", (user.name, user.email))
    db.commit()
    db.close()
    return {"message": "User created successfully"}

@app.get("/users")
def read_users():
    db = get_connection()
    cursor = db.cursor(dictionary=True)
    cursor.execute("SELECT * FROM users")
    users = cursor.fetchall()
    db.close()
    return users

@app.get("/users/{user_id}")
def read_user(user_id: int):
    db = get_connection()
    cursor = db.cursor(dictionary=True)
    cursor.execute("SELECT * FROM users WHERE id = %s", (user_id,))
    user = cursor.fetchone()
    db.close()
    if user:
        return user
    raise HTTPException(status_code=404, detail="User not found")

@app.delete("/users/{user_id}")
def delete_user(user_id: int):
    db = get_connection()
    cursor = db.cursor()
    cursor.execute("DELETE FROM users WHERE id = %s", (user_id,))
    db.commit()
