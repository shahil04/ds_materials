## If-else

```python
# syntax
# if conditions:
#   statements
# else:
#   statements
age =12
if age>=18:
  print("You are adult")
else:
  print("You are not adult")
```

```python
# nested if-else
user = input("Enter your userid ")

if user=="shahil":
  print("userid is correct")

  passwod = input("Enter your password ")

  if passwod=="123":
    print("password is correct")
  else:
    print("password is incorrect")

else:
  print("userid is incorrect")
```

A company needs to determine employee bonus eligibility:

If the employee has worked more than = 5 years and has a performance rating above = 4, they get a bonus.


Otherwise, no bonus.


```python
year =float(input("Enter your year of experience "))
rating =float(input("Enter your performance rating "))
# simple if-else with logical operators
if year>=5 and rating>=4:
  print("You are eligible for bonus")
else:
  print("You are not eligible for bonus")
```

```python
year =float(input("Enter your year of experience "))
rating =float(input("Enter your performance rating "))
# nested if
if year>=5:
  if rating>=4:
    print("You are eligible for bonus")
else:
  print("You are not eligible for bonus")
```

# using if-elif-else

```python

# When you have multiple conditions to check.
# Syntax:
# if condition1:
#     statement(s)
# elif condition2:
#     statement(s)
# elif condition3:
#     statement(s)
# else:
#     statement(s)

marks=65
if marks>=90:
  print("A")
elif marks>=75:
  print("B")
elif marks>=60:
  print("C")
else:
  print("Fail")
```

```python
# retail discount system
total_buy = input("enter the total amount of customer: ")

if total_buy.isnumeric():
  total_buy = float(total_buy)
else:
  print("Invalid input")
  total_buy = 0

discount = 0

if total_buy>=5000:
  discount = 0.25

elif total_buy>2000:
  discount = .15

elif total_buy>=500:
  discount = .10
else:
  discount=0

after_discount = total_buy -total_buy*discount
print("Total bill amount ",total_buy, "discount is ", discount," after discount bill Amount", after_discount)
```
