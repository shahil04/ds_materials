# Multiple `if` vs Nested `if`

Consider a college scholarship.

Requirements:

* Marks ≥ 80
* Attendance ≥ 75

---

## Using Multiple `if`

````python
marks = 85
attendance = 70

if marks >= 80:
    print("Marks condition satisfied")

if attendance >= 75:
    print("Attendance condition satisfied")
````

**Output:**

```
Marks condition satisfied

what happen here 

Each condition is checked separately.


````python
## Using Nested `if`
marks = 85
attendance = 79

if marks >= 80:
    print("Marks condition satisfied")

    if attendance >= 75:
        print("Attendance condition satisfied")
        
````

````python
## Using Nested `if`

marks = 85
attendance = 79

if marks >= 80:
    print("Marks condition satisfied")

    if attendance >= 75:
        print("Scholarship Approved")
````

**Output:**

```
Marks condition satisfied


The attendance check happens only after the marks condition is satisfied.


---

## Visual Difference

### Multiple `if`

```text
Check Marks
   |
Check Attendance
```

Both checks happen independently.

### Nested `if`

```text
Check Marks
   |
   +-- True --> Check Attendance
                  |
                  +-- True --> Approve
```

The second check depends on the first.

### 4.using if-elif-else

````python
# When you have multiple conditions to check.
# Syntax:
# if condition1:
#     statement(s)
# elif condition2:
#     statement(s)
# elif condition3:
#     statement(s)
# else:
#     statement(s)

marks=79
if marks>=90:
  print("A")
elif marks>=75:
  print("B")

elif marks>=60:
  print("C")
else:
  print("Fail")
````

# When to Use Which?

### Use Multiple `if` When Conditions Are Independent



# Summary Table

| Type                  | Purpose                            | Example                    |
| --------------------- | ---------------------------------- | -------------------------- |
| `if`                  | Check one condition                | Age ≥ 18                   |
| `if-else`             | Two outcomes                       | Pass/Fail                  |
| `if-elif-else`        | Multiple alternatives              | Grade A/B/C                |
| Multiple `if`         | Independent checks                 | Academic and Sports awards |
| Nested `if`           | One condition depends on another   | Admit card → ID card       |
| Nested `if-else`      | Dependent checks with alternatives | Loan approval              |
| Nested `if-elif-else` | Multi-level decisions              | University admission       |

### A Helpful Rule

* **Multiple `if`** = "Check everything."
* **Nested `if`** = "Check the next condition only if the previous condition is true."
* **`if-elif-else`** = "Choose exactly one option from many possibilities."
