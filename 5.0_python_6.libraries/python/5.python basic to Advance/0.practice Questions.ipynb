```python
# Find the largest and smallest number in a list NOT USING 
# ANY INBUIT FUNCTION
data = [4,5,40,12,0]
def min_max(numbers):
    small =numbers[0]
    large =numbers[0]

    for i in numbers:
        if i <small:
            small = i
        if i>large:
            large =i
    return small , large

min_max(data)


# Count vowels in a string "HELLO PYTHON"

# Sum all elements in a list WITOUT USING SUM FUCTIONS [4,5,4,12,1]

# Swap two variables without using a third variable A=5 ,B=6

# Remove duplicates from a list [4,5,4,12,1]

# Reverse a string without using built-in functions "HELLO PYTHON"
```
