## List
1. Introduction to Lists
A list in Python is a collection of items that can store different types of data (integers, strings, floats, etc.) and is ordered and mutable (can be changed).

- ✅ Lists are defined using square brackets [ ].
-
✅ Items are separated by commas.
* Example:
```
my_list = [1, 2, 3, "apple", 4.5]
print(my_list)  # [1, 2, 3, 'apple', 4.5]
```

````python
# sequetical datatype (list ,tuple, set , dictionary) 
# create ,update, read,delete(CURD opertions)

# Way to create a List
# 1. empty list
list1 = [] # square brackets use for list --> empty list
print(list1)

# 2.  # initialization with value
list2 =[34,6,52,3]
print(list2)

# 3. using list() function
list3 = list() 
print(list3)

 # 4.using list functions with other datatypes
list4 = list({3,4,5,6})
print(list4)
````
