## List
1. Introduction to Lists
A list in Python is a collection of items that can store different types of data (integers, strings, floats, etc.) and is ordered and mutable (can be changed).

- ✅ Lists are defined using square brackets [ ].
-
✅ Items are separated by commas.
* Example:
```
my_list = [1, 2, 3, "apple", 4.5]
print(my_list)  # [1, 2, 3, 'apple', 4.5]
```
