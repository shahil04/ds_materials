* Example:

````python
import gradio as gr

# Global cart list
cart = []


def add_item(product_name):
    if product_name.strip():
        cart.append(product_name)
    return "\n".join(cart)


def show_cart():
    if not cart:
        return "Cart is empty."
    return "\n".join(cart)


def update_cart(old_name, new_name):
    if old_name in cart:
        index = cart.index(old_name)
        cart[index] = new_name
        return f"{old_name} updated to {new_name}\n\nCart:\n" + "\n".join(cart)
    else:
        return f"{old_name} not found in cart."


def delete_item(product_name):
    if product_name in cart:
        cart.remove(product_name)
        return f"{product_name} removed.\n\nCart:\n" + ("\n".join(cart) if cart else "Cart is empty.")
    else:
        return f"{product_name} not found in cart."


with gr.Blocks(title="Shopping Cart") as demo:
    gr.Markdown("# 🛒 Shopping Cart")

    # Add Item
    with gr.Row():
        add_input = gr.Textbox(label="Product Name")
        add_btn = gr.Button("Add Item")

    # Update Item
    with gr.Row():
        old_input = gr.Textbox(label="Old Product Name")
        new_input = gr.Textbox(label="New Product Name")
        update_btn = gr.Button("Update Item")

    # Delete Item
    with gr.Row():
        delete_input = gr.Textbox(label="Product to Delete")
        delete_btn = gr.Button("Delete Item")

    show_btn = gr.Button("Show Cart")

    output = gr.Textbox(label="Cart", lines=10)

    # Button Actions
    add_btn.click(add_item, inputs=add_input, outputs=output)
    update_btn.click(update_cart, inputs=[old_input, new_input], outputs=output)
    delete_btn.click(delete_item, inputs=delete_input, outputs=output)
    show_btn.click(show_cart, outputs=output)

demo.launch()
````

````python
import gradio as gr
import google.generativeai as genai

# Replace with your Gemini API Key
# API_KEY = "YOUR_GEMINI_API_KEY"
API_KEY = ""
genai.configure(api_key=API_KEY)

model = genai.GenerativeModel("gemini-3.5-flash")

SYSTEM_PROMPT = """
You are a DevOps Tutor chatbot.

Your job is to answer questions related to:
- Linux
- Docker
- Kubernetes
- Git & GitHub
- Jenkins
- CI/CD
- Terraform
- Ansible
- AWS, Azure, GCP
- Monitoring (Prometheus, Grafana)
- DevOps Interview Questions

Explain concepts in simple language with examples.

If the question is not related to DevOps, politely reply:
'I am a DevOps assistant. Please ask DevOps-related questions.'
"""

def chatbot(message, history):
    prompt = SYSTEM_PROMPT + "\n\nStudent: " + message

    response = model.generate_content(prompt)

    return response.text


demo = gr.ChatInterface(
    fn=chatbot,
    title="🤖 DevOps Student Chatbot",
    description="Ask anything about Docker, Kubernetes, Git, Linux, Jenkins, AWS, CI/CD, Terraform, and DevOps."
)

demo.launch()
````

#### List comprehension

List comprehension in Python offers a concise, elegant, and efficient way to create a new list based on the values of an existing iterable (such as a list, tuple, or string). It is a single line of code that replaces traditional loops for many common tasks like filtering and data transformation, often resulting in faster execution. 

- Syntax
- new_list = [expression for item in iterable if condition]

````python
# WAP to create a square of number 1 to 10 and store in list 
square =[]
for i in range(1,11):
    square.append(i**2)
print(square)
````

````python
var  = [i **2 for i in range(1,11)]
var
````

````python
# Normal loop --> even number store in list
evens = []
for x in range(11):
    if x % 2 == 0:
        evens.append(x)
evens

# list comprehension
[i for  i in range (1,11) if i%2==0]
````

````python
result = []
for n in range(11):
    if n % 2 == 0:
        result.append(f'Even {n}')
    else:
        result.append(f'Odd {n}')
result

# [ Expression if condition else expression for item in itrable]
result = [f"Even {n}" if n%2==0 else f'odd {n}' for n in range(11)]
result
````
