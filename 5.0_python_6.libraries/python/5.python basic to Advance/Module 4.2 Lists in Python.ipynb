#### List comprehension

List comprehension in Python offers a concise, elegant, and efficient way to create a new list based on the values of an existing iterable (such as a list, tuple, or string). It is a single line of code that replaces traditional loops for many common tasks like filtering and data transformation, often resulting in faster execution. 

- Syntax
- new_list = [expression for item in iterable if condition]

````python
# WAP to create a square of number 1 to 10 and store in list 
square =[]
for i in range(1,11):
    square.append(i**2)
print(square)
````

````python
var  = [i **2 for i in range(1,11)]
var
````

````python
# Normal loop --> even number store in list
evens = []
for x in range(11):
    if x % 2 == 0:
        evens.append(x)
evens

# list comprehension
[i for  i in range (1,11) if i%2==0]
````

````python
result = []
for n in range(11):
    if n % 2 == 0:
        result.append(f'Even {n}')
    else:
        result.append(f'Odd {n}')
result

# [ Expression if condition else expression for item in itrable]
result = [f"Even {n}" if n%2==0 else f'odd {n}' for n in range(11)]
result
````
