## ðŸ”° What is Comprehension?

**Comprehension** in Python is a concise way to construct sequences (like lists, dictionaries, sets, or generators) using a single line of code. It replaces multiple lines of loop and `append()` calls.

---

## âœ… 1. **List Comprehension**

### ðŸ“Œ Syntax:

```python
[expression for item in iterable if condition]
```

### ðŸ“Ž Examples:

#### âž¤ Basic list comprehension

```python
squares = [x**2 for x in range(6)]
print(squares)  # [0, 1, 4, 9, 16, 25]
```

#### âž¤ With condition (Even numbers)

```python
evens = [x for x in range(10) if x % 2 == 0]
print(evens)  # [0, 2, 4, 6, 8]
```

#### âž¤ Nested loop (2D to 1D - flattening)

```python
matrix = [[1, 2], [3, 4]]
flat = [num for row in matrix for num in row]
print(flat)  # [1, 2, 3, 4]
```

#### âž¤ String manipulation

```python
words = ["hello", "world"]
upper = [w.upper() for w in words]
print(upper)  # ['HELLO', 'WORLD']
```

---

## âœ… 2. **Dictionary Comprehension**

### ðŸ“Œ Syntax:

```python
{key_expression: value_expression for item in iterable if condition}
```

### ðŸ“Ž Examples:

#### âž¤ Create dictionary from list

```python
nums = [1, 2, 3]
squares = {x: x**2 for x in nums}
print(squares)  # {1: 1, 2: 4, 3: 9}
```

#### âž¤ Filter dictionary (keep only even values)

```python
original = {'a': 1, 'b': 2, 'c': 3}
filtered = {k: v for k, v in original.items() if v % 2 == 0}
print(filtered)  # {'b': 2}
```

#### âž¤ Swap keys and values

```python
d = {'a': 1, 'b': 2}
swapped = {v: k for k, v in d.items()}
print(swapped)  # {1: 'a', 2: 'b'}
```

---

## âœ… 3. **Set Comprehension**

### ðŸ“Œ Syntax:

```python
{expression for item in iterable if condition}
```

### ðŸ“Ž Examples:

#### âž¤ Create a set of squares

```python
squares = {x**2 for x in range(5)}
print(squares)  # {0, 1, 4, 9, 16}
```

#### âž¤ Remove duplicates and filter

```python
nums = [1, 2, 2, 3, 4, 4]
evens = {x for x in nums if x % 2 == 0}
print(evens)  # {2, 4}
```

---

## âœ… 4. **Generator Comprehension**

### ðŸ“Œ Syntax:

```python
(expression for item in iterable if condition)
```

> âš  Returns a **generator object** instead of a list. It uses **less memory** and is **lazy-evaluated**.

### ðŸ“Ž Examples:

#### âž¤ Basic usage

```python
gen = (x**2 for x in range(5))
print(next(gen))  # 0
print(next(gen))  # 1
```

#### âž¤ Convert to list

```python
gen = (x**2 for x in range(5))
print(list(gen))  # [0, 1, 4, 9, 16]
```

# ðŸ“˜ Python: **Iterators vs Generators**

---

## âœ… **1. What is an Iterator?**

An **iterator** is any object that implements two methods:

* `__iter__()` â†’ returns the iterator object itself.
* `__next__()` â†’ returns the next value from the sequence.

ðŸ“Œ **Used for looping through a sequence of data one element at a time.**

### ðŸ”¹ Example: Creating a custom iterator

```python
class CountUpTo:
    def __init__(self, max):
        self.max = max
        self.num = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.num <= self.max:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration

counter = CountUpTo(3)
for num in counter:
    print(num)
```

**Output:**

```
1
2
3
```

---

## âœ… **2. What is a Generator?**

A **generator** is a simpler way to create iterators using:

* The `yield` keyword instead of `return`.
* Automatically implements `__iter__()` and `__next__()`.

ðŸ“Œ **Generators are memory-efficient and easier to write than custom iterators.**

---

### ðŸ”¹ Example: Generator function

```python
def count_up_to(max):
    num = 1
    while num <= max:
        yield num
        num += 1

for n in count_up_to(3):
    print(n)
```

**Output:**

```
1
2
3
```

---

### âœ… **3. Generator Expression (Like List Comprehension)**

```python
gen = (x*x for x in range(5))
print(next(gen))  # 0
print(next(gen))  # 1
```

## ðŸ”„ **Iterator vs Generator: Key Differences**

| Feature      | Iterator                                              | Generator                               |
| ------------ | ----------------------------------------------------- | --------------------------------------- |
| Definition   | Class-based object with `__iter__()` and `__next__()` | Function or expression using `yield`    |
| Syntax       | Requires a class, methods                             | Uses simple function with `yield`       |
| Memory usage | Loads full data in memory (unless optimized)          | More memory-efficient (lazy evaluation) |
| Complexity   | More complex, boilerplate code                        | Easy to write and read                  |
| Performance  | Slower if not optimized                               | Faster for large data streams           |
| Reusability  | Reusable if written correctly                         | One-time use (unless re-called)         |
| Example      | Custom iterator class                                 | Generator function or expression        |

---

## ðŸ§  When to Use What?

* Use **iterators** when:

  * You need full control over iteration logic.
  * You're building reusable/custom iteration behavior.

* Use **generators** when:

  * You want **lazy evaluation**.
  * You're processing **large datasets** or streams.
  * You want concise code for producing a sequence.
