# üß© 1. Procedural Approach

Everything is **just variables**.


````python
# Procedural
student1_name = "Rahul"
student1_age = 15
student1_marks = 85

student2_name = "Priya"
student2_age = 14
student2_marks = 90

def display_student(name, age, marks):
    print(f"Name: {name}, Age: {age}, Marks: {marks}")

display_student(student1_name, student1_age, student1_marks)
display_student(student2_name, student2_age, student2_marks)
````


üëâ Problems:

* Too many separate variables.
* Hard to manage 100 students.
* No natural grouping of **data + behavior**.

---

# üß© 2. Functional + Dictionary Approach

We improve by grouping **related data** in a dictionary.

````python
# Functional + Dictionary
student1 = {"name": "Rahul", "age": 15, "marks": 85}
student2 = {"name": "Priya", "age": 14, "marks": 90}

def display_student(student):
    print(f"Name: {student['name']}, Age: {student['age']}, Marks: {student['marks']}")

display_student(student1)
display_student(student2)
````



üëâ Better:

* Easier to handle multiple students.
* Data grouped in one place.

üëâ Still Problems:

* What if someone writes `student1["marks"] = "eighty five"` (string instead of number)? ‚Üí No validation.
* Repetition: Every student dictionary needs the same keys manually.
* Functions and data are still **separate**.
* No reusability (e.g., a **Teacher** object will need a separate dictionary structure).

## üîπ Part 1: Why OOP? (Problem Statement)

* In real life, everything is an **object** (Student, Teacher, Car, Bank Account).
* Objects have **attributes (data)** and **behaviors (actions)**.
* In coding, if we use **procedural approach**, data and functions are separate ‚Üí leads to complexity.
* OOP solves this by grouping **data + functions** together in a **class**.


# üß© 3. Object-Oriented Approach

We bundle **data + behavior** into one structure called a **class**.


````python
class Student:
    def __init__(self, name, age, marks):   # constructor
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Marks: {self.marks}")

s1 = Student("Rahul", 15, 85)
s2 = Student("Priya", 14, 90)

s1.display()
s2.display()
````


üëâ Advantages over dictionary approach:

1. **Encapsulation** ‚Üí We can control how data is accessed (e.g., prevent invalid marks).
2. **Reusability** ‚Üí Create other classes like `Teacher`, `Course`, etc.
3. **Inheritance** ‚Üí `GraduateStudent` can inherit from `Student`.
4. **Polymorphism** ‚Üí Same method `display()` works differently in different classes.
5. **Scalability** ‚Üí Easy to maintain when project grows large.

````python
class MyClass:
    class_variable = 0

    def __init__(self, instance_variable):
        self.instance_variable = instance_variable

    # create a class method
    @classmethod
    def update_class_variable(cls, value):
        cls.class_variable = value

    def display_variables(self):
        print(f"Class Variable: {MyClass.class_variable}, Instance Variable: {self.instance_variable}")

# Calling the class method
MyClass.update_class_variable(10)

# Creating instances and displaying variables
instance1 = MyClass(5)
instance1.display_variables()  # Output: Class Variable: 10, Instance Variable: 5
instance2 = MyClass(15)
````

# 1. Inheritance

Inheritance in Python is a fundamental concept of object-oriented programming (OOP) that allows a new class (called a child class or subclass) to inherit attributes and methods from an existing class (called a parent class or superclass). This mechanism promotes code reusability, modularity, and the creation of hierarchical relationships between classes.

Key Aspects of Inheritance in Python:

- Code Reusability: Child classes can reuse the code defined in their parent classes, avoiding redundant code and making the codebase more efficient and maintainable.

- Class Hierarchy: Inheritance establishes a "is-a" relationship, meaning a child class "is a" type of its parent class. For example, a Dog class could inherit from an Animal class, as a dog "is an" animal.
Syntax:
Python

    class ParentClass:
        # attributes and methods of the parent class

    class ChildClass(ParentClass):
        # attributes and methods of the child class
        # can also override or extend parent's methods
Method Overriding:
A child class can provide its own implementation for a method that is already defined in its parent class. This allows for specialization of behavior in the child class while still maintaining the common interface.
super() Function:
The super() function is used within a child class to call methods or access attributes of its parent class, particularly useful for calling the parent's __init__ method to ensure proper initialization.
Types of Inheritance:
Single Inheritance: A class inherits from only one parent class.
Multiple Inheritance: A class inherits from multiple parent classes. Python resolves potential conflicts in method resolution order (MRO).
Multilevel Inheritance: A class inherits from another subclass, creating a chain of inheritance.
Hierarchical Inheritance: Multiple child classes inherit from a single parent class.

```python
class BankAccount:
    def __init__(self, name, account_number, balance):
        self.__name = name
        self.__account_number = account_number
        self.__balance = balance
        self.__refund_limit_percent = 0.5  # Only 50% of balance can be refunded

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"‚úÖ ‚Çπ{amount} deposited.")
        else:
            print("‚ùå Invalid deposit amount.")

    def withdraw(self, amount):
        if amount > 0 and self.__balance >= amount:
            self.__balance -= amount
            print(f"‚úÖ ‚Çπ{amount} withdrawn.")
        else:
            print("‚ùå Insufficient balance or invalid amount.")

    def customer_details(self):
        print("üìÑ Account Details:")
        print(f"Name: {self.__name}")
        print(f"Account Number: {self.__account_number}")
        print(f"Balance: ‚Çπ{self.__balance}")

    # Public method: Customer asks "Can I get this much refund?"
    def check_refund_eligibility(self, refund_amount):
        if self.__is_eligible_for_refund(refund_amount):
            print(f"‚úÖ You are eligible for ‚Çπ{refund_amount} refund.")
        else:
            print(f"‚ùå Refund of ‚Çπ{refund_amount} is not allowed.")

    # üîê Private method: Internal refund policy check
    def __is_eligible_for_refund(self, refund_amount):
        max_refund = self.__balance * self.__refund_limit_percent
        return 0 < refund_amount <= max_refund
```

## üéØ What is Abstraction?

> **Abstraction** is the process of hiding the internal implementation and **only showing essential features**.

### üìå Real-Life Example:

> When you use an **ATM**, you just press buttons (UI).

**You use the interface, not the internal complexity.**

---

## ‚úÖ In Python, we use:

* `abc` module ‚Üí `ABC` (Abstract Base Class)
* `@abstractmethod` ‚Üí force subclasses to implement the method

---

## üß† Why Use Abstraction?

* Enforce **common structure**
* Hide internal logic
* Encourage **interface-based design**

from abc import ABC, abstractmethod

# üîê Abstract base class (Abstraction)
class Account(ABC):
    def __init__(self, name, account_number, initial_balance):
        self._name = name                    # Protected
        self._account_number = account_number  # Protected
        self.__balance = initial_balance     # Private

    # üîê Abstract method
    @abstractmethod
    def calculate_interest(self):
        pass

    # ‚úÖ Encapsulated deposit method
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"‚úÖ ‚Çπ{amount} deposited.")
        else:
            print("‚ùå Invalid deposit amount.")

    # ‚úÖ Encapsulated withdraw method
    def withdraw(self, amount):
        if amount > 0 and self.__balance >= amount:
            self.__balance -= amount
            print(f"‚úÖ ‚Çπ{amount} withdrawn.")
        else:
            print("‚ùå Insufficient balance or invalid amount.")

    # ‚úÖ Polymorphic method (to be overridden)
    def display_account_type(self):
        print("Generic Bank Account")

    # üß† @property (getter)
    @property
    def balance(self):
        return self.__balance

    # üß† @balance.setter (with validation)
    @balance.setter
    def balance(self, value):
        if value < 0:
            print("‚ùå Balance cannot be negative.")
        else:
            self.__balance = value

    # ‚úÖ Public method to view account info
    def show_account_info(self):
        print(f"\nüìÑ Account Holder: {self._name}")
        print(f"üî¢ Account Number: {self._account_number}")
        print(f"üí∞ Balance: ‚Çπ{self.balance}")

            return True
        return False

            CREATE TABLE IF NOT EXISTS accounts (
                account_number TEXT PRIMARY KEY,
                name TEXT,
                type TEXT,
                balance REAL
            )
        ''')
        conn.execute('''
            CREATE TABLE IF NOT EXISTS transactions (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                account_number TEXT,
                type TEXT,
                amount REAL,
                time TEXT
            )
        ''')

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/create', methods=['GET', 'POST'])
def create_account():
    if request.method == 'POST':
        name = request.form['name']
        acc_num = request.form['acc_num']
        acc_type = request.form['acc_type']
        balance = float(request.form['balance'])

        with sqlite3.connect('database.db') as conn:
            conn.execute('INSERT INTO accounts VALUES (?, ?, ?, ?)', (acc_num, name, acc_type, balance))
        return redirect(url_for('dashboard', acc_num=acc_num))
    return render_template('create_account.html')

@app.route('/dashboard/<acc_num>')
def dashboard(acc_num):
    with sqlite3.connect('database.db') as conn:
        cur = conn.cursor()
        cur.execute('SELECT name, type, balance FROM accounts WHERE account_number=?', (acc_num,))
        data = cur.fetchone()
        if not data:
            return "Account not found"
        name, acc_type, balance = data
    return render_template('dashboard.html', name=name, acc_num=acc_num, acc_type=acc_type, balance=balance)

@app.route('/action/<acc_num>', methods=['POST'])
def action(acc_num):
    action_type = request.form['action']
    amount = float(request.form['amount'])

    with sqlite3.connect('database.db') as conn:
        cur = conn.cursor()
        cur.execute('SELECT type, balance FROM accounts WHERE account_number=?', (acc_num,))
        acc_type, current_balance = cur.fetchone()

        if action_type == 'deposit':
            new_balance = current_balance + amount
            trans_type = 'Deposit'
        elif action_type == 'withdraw' and current_balance >= amount:
            new_balance = current_balance - amount
            trans_type = 'Withdraw'
        else:
            return "Invalid transaction"

        # Update balance
        conn.execute('UPDATE accounts SET balance=? WHERE account_number=?', (new_balance, acc_num))
        # Log transaction
        conn.execute('INSERT INTO transactions (account_number, type, amount, time) VALUES (?, ?, ?, datetime("now"))',
                     (acc_num, trans_type, amount))

    return redirect(url_for('dashboard', acc_num=acc_num))

@app.route('/transactions/<acc_num>')
def view_transactions(acc_num):
    with sqlite3.connect('database.db') as conn:
        cur = conn.cursor()
        cur.execute('SELECT type, amount, time FROM transactions WHERE account_number=? ORDER BY time DESC', (acc_num,))
        logs = cur.fetchall()
    return render_template('transactions.html', acc_num=acc_num, logs=logs)

if __name__ == '__main__':
    init_db()
    app.run(debug=True)
