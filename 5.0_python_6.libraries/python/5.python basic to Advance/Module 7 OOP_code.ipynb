# 🧩 1. Procedural Approach

Everything is **just variables**.


````python
# Procedural 
student1_name = "Rahul"
student1_age = 15
student1_marks = 85

student2_name = "Priya"
student2_age = 14
student2_marks = 90

def display_student(name, age, marks):
    print(f"Name: {name}, Age: {age}, Marks: {marks}")

display_student(student1_name, student1_age, student1_marks)
display_student(student2_name, student2_age, student2_marks)
````


👉 Problems:

* Too many separate variables.
* Hard to manage 100 students.
* No natural grouping of **data + behavior**.

---

# 🧩 2. Functional + Dictionary Approach

We improve by grouping **related data** in a dictionary.

````python
# Functional + Dictionary
student1 = {"name": "Rahul", "age": 15, "marks": 85}
student2 = {"name": "Priya", "age": 14, "marks": 90}

def display_student(student):
    print(f"Name: {student['name']}, Age: {student['age']}, Marks: {student['marks']}")

display_student(student1)
display_student(student2)
````



👉 Better:

* Easier to handle multiple students.
* Data grouped in one place.

👉 Still Problems:

* What if someone writes `student1["marks"] = "eighty five"` (string instead of number)? → No validation.
* Repetition: Every student dictionary needs the same keys manually.
* Functions and data are still **separate**.
* No reusability (e.g., a **Teacher** object will need a separate dictionary structure).

---
## 🔹 Part 1: Why OOP? (Problem Statement)

* In real life, everything is an **object** (Student, Teacher, Car, Bank Account).
* Objects have **attributes (data)** and **behaviors (actions)**.
* In coding, if we use **procedural approach**, data and functions are separate → leads to complexity.
* OOP solves this by grouping **data + functions** together in a **class**.


# 🧩 3. Object-Oriented Approach

We bundle **data + behavior** into one structure called a **class**.


````python
class Student:
    def __init__(self, name, age, marks):   # constructor
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Marks: {self.marks}")

s1 = Student("Rahul", 15, 85)
s2 = Student("Priya", 14, 90)

s1.display()
s2.display()
````


👉 Advantages over dictionary approach:

1. **Encapsulation** → We can control how data is accessed (e.g., prevent invalid marks).
2. **Reusability** → Create other classes like `Teacher`, `Course`, etc.
3. **Inheritance** → `GraduateStudent` can inherit from `Student`.
4. **Polymorphism** → Same method `display()` works differently in different classes.
5. **Scalability** → Easy to maintain when project grows large.

## instance mathod and class mathod

````python
class MyClass:
    class_variable = 0

    def __init__(self, instance_variable):
        self.instance_variable = instance_variable

    # create a class method
    @classmethod   #@decorator
    def update_class_variable(cls, value):
        cls.class_variable = value

    def display_variables(self):
        print(f"Class Variable: {MyClass.class_variable}, Instance Variable: {self.instance_variable}")

# Calling the class method
MyClass.update_class_variable(10)

# Creating instances and displaying variables
instance1 = MyClass(5)
instance1.display_variables()  # Output: Class Variable: 10, Instance Variable: 5
instance2 = MyClass(15)
````

In Python, self is a conventional name for the first parameter of instance methods within a class. It is not a keyword, but a widely adopted convention that refers to the instance of the class on which the method is being called.

**Purpose of self:**

1. Accessing Instance Attributes:

self allows methods to access and modify the attributes (variables) specific to that particular instance of the class. Without self, a method would not know which instance's attributes it should operate on.

2. Calling Instance Methods:
Similarly, self enables methods within a class to call other instance methods belonging to the same object.

3. Distinguishing Instance and Local Variables:
It helps differentiate between local variables within a method and instance variables (attributes) of the object.

Inheritance in Python is a fundamental concept of object-oriented programming (OOP) that allows a new class (called a child class or subclass) to inherit attributes and methods from an existing class (called a parent class or superclass). This mechanism promotes code reusability, modularity, and the creation of hierarchical relationships between classes.


Types of Inheritance:
- Single Inheritance: A class inherits from only one parent class.
- Multiple Inheritance: A class inherits from multiple parent classes. Python resolves potential conflicts in method resolution order (MRO).
- Multilevel Inheritance: A class inherits from another subclass, creating a chain of inheritance.
- Hierarchical Inheritance: Multiple child classes inherit from a single parent class.

````python
# human class as a parents class
class Human:
  def __init__(self,name,age,gender):
    self.name = name
    self.a = age
    self.gender = gender

  def walk(self):
    return (self.name,"is walking")

  def talk(self):
    print(self.name,"is talking")

  def humandetails(self):
    print(f" Human Details:\n name:{self.name} \n age:{self.a} \n gender: {self.gender}")
````

````python
# student class with inheritance
class Student(Human):
  def __init__(self,name,age,gender,std_id, std):
    # Human.__init__(self,name,age,gender)
    super().__init__(name,age,gender)

    self.std_id = std_id
    self.std = std

  def studentdetails(self):
    print(f" Student Details:\n name:{self.name} \n id:{self.std_id} \n Class: {self.std}")

  def walk(self): # method overriding
    return (self.name,"is walking in the class")
````


Key Aspects of Inheritance in Python:

- Code Reusability: Child classes can reuse the code defined in their parent classes, avoiding redundant code and making the codebase more efficient and maintainable.

- Class Hierarchy: Inheritance establishes a "is-a" relationship, meaning a child class "is a" type of its parent class. For example, a Dog class could inherit from an Animal class, as a dog "is an" animal.
Syntax:
Python

    class ParentClass:
        # attributes and methods of the parent class

    class ChildClass(ParentClass):
        # attributes and methods of the child class
        # can also override or extend parent's methods

- Method Overriding:

A child class can provide its own implementation for a method that is already defined in its parent class. This allows for specialization of behavior in the child class while still maintaining the common interface.

- super() Function:

The super() function is used within a child class to call methods or access attributes of its parent class, particularly useful for calling the parent's __init__ method to ensure proper initialization.

## 🎯 What is `@property`?

The `@property` decorator in Python is used to:

✅ Access a **method like an attribute**
✅ **Control access** to private/protected variables
✅ Allow **read-only or read-write** behavior
✅ Provide **validation** logic during access or update


### 📌 With `@property`:

You can do:


## ✅ Real-World Example with `@property`

```python
class BankAccount:
    def __init__(self, name, account_number, balance):
        self._name = name
        self._account_number = account_number
        self.__balance = balance

    # 🌐 Public Method
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"✅ ₹{amount} deposited.")
        else:
            print("❌ Invalid deposit amount.")

    # 🌐 Public Method
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"✅ ₹{amount} withdrawn.")
        else:
            print("❌ Insufficient balance or invalid amount.")

    # ✅ Read-only property (getter)
    @property
    def balance(self):
        return self.__balance

    # ✅ Optional: setter if you want to allow external update
    @balance.setter
    def balance(self, value):
        if value < 0:
            print("❌ Balance can't be negative.")
        else:
            self.__balance = value
            print(f"✅ Balance set to ₹{value}")

    # ✅ Read-only properties for name and account
    @property
    def name(self):
        return self._name

    @property
    def account_number(self):
        return self._account_number
```

## 4. Abstractions
Abstraction in Object-Oriented Programming (OOP) is a design principle that hides complex implementation details, exposing only essential features to the user. It manages software complexity, reduces, and improves maintainability by allowing developers to focus on what an object does rather than how it does it. 

- Advantages of Abstraction:
1. Security: Hides internal data and implementation, protecting the system.
2. Maintainability: Internal changes can be made without affecting users, as long as the exposed interface remains the same.
3. Reusability & Modularity: Promotes organized, cleaner code. 

````python
from abc import ABC, abstractmethod

# ✅ Abstract Base Class
class Payment(ABC):

    @abstractmethod
    def authenticate(self):
        pass

    @abstractmethod
    def pay(self, amount):
        pass

    def type(self):
        pass
````

````python
class CreditCardPayment(Payment):
    def authenticate(self):
        print("🔐 Authenticating with OTP for Credit Card.")
````

- Abstraction vs. Encapsulation:

While sometimes confused, they differ:
1. Abstraction: Hides the complexity of the design.
2. Encapsulation: Hides the data (hiding state and requiring all interaction to occur through an object's methods). 

# 🎯 Abstraction?

Let's understand **Abstraction in Python** with a **real-world explanation** and **code implementation** using `abc` (Abstract Base Class) — in a way that's useful for both teaching and real project structure.

---

## 🎯 What is Abstraction?

> **Abstraction** is the process of hiding the internal implementation and **only showing essential features**.

---

### 📌 Real-Life Example:

> When you use an **ATM**, you just press buttons (UI).
> You don’t know how it connects to the bank server, processes balance, verifies OTP, etc.

That’s abstraction:
**You use the interface, not the internal complexity.**

---

## ✅ In Python, we use:

* `abc` module → `ABC` (Abstract Base Class)
* `@abstractmethod` → force subclasses to implement the method

---

## 🧠 Why Use Abstraction?

* Enforce **common structure**
* Hide internal logic
* Encourage **interface-based design**

---

## ✅ Real-World Example: Bank Payment System

We’ll create:

* `Payment` (abstract class/interface)
* `CreditCardPayment`, `UPIPayment`, etc. (subclasses)

---

### 🔐 Abstract Class Code

```python
from abc import ABC, abstractmethod

# ✅ Abstract Base Class
class Payment(ABC):
    @abstractmethod
    def authenticate(self):
        pass

    @abstractmethod
    def pay(self, amount):
        pass
```

---

### 🔧 Subclass Implementations

```python
class CreditCardPayment(Payment):
    def authenticate(self):
        print("🔐 Authenticating with OTP for Credit Card.")

    def pay(self, amount):
        print(f"💳 Paid ₹{amount} using Credit Card.")

class UPIPayment(Payment):
    def authenticate(self):
        print("🔐 Authenticating using UPI PIN.")

    def pay(self, amount):
        print(f"📲 Paid ₹{amount} using UPI.")
```

---

### 🧪 Usage Example

```python
def complete_payment(payment_method, amount):
    payment_method.authenticate()
    payment_method.pay(amount)

cc = CreditCardPayment()
upi = UPIPayment()

complete_payment(cc, 1000)
complete_payment(upi, 500)
```

## ❌ Error If Not Implemented:

If a subclass of `Payment` doesn’t implement all `@abstractmethod`s, Python will raise:

```plaintext
TypeError: Can't instantiate abstract class ...
```
