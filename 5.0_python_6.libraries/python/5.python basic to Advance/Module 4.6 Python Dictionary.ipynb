## Dictionary data type
A dictionary is a mutable, unordered collection of key-value pairs in Python where:

```python
# syntax
# {key1: value1, key2: value2}
```

```python
# create a dictionary
d1 ={}
print(type(d1))
```

```python
d2 = dict()
print(type(d2))
```

```python
# syntax of dict {key : value/values}
std = {"name": "raj", "id":1,'subject':'data science'}
print(type(std))
std
```

```python
# Q create a dictionary country name as key or country capital as value --> 4 country
countries = {"india":"delhi",
            "pakistan":"islamabad",
             "bangladesh":"dhaka",
             "nepal":"kathmandu"
             }
```

#### methods of Dictionary
- get()     return value from given key 
- keys()	Returns a view object of all the keys.
- values()	Returns a view object of all the values.
- items()	Returns a view object of all key-value pairs as tuples.
- update()	Merges the dictionary with another dictionary or iterable of key-value pairs.
- clear()	Removes all items from the dictionary.
- copy()    
- pop()
- popitems()


```python
# accesss the value -->print the captial of nepal
countries["nepal"] # index/key
# read value
```

```python
#use  get methods for access values from key 
countries.get("nepal")
```

```python
# only want keys -- use keys methods
print(countries.keys())
```

```python
# only values
print(countries.values())
```

```python
# items -return key value pair as tuple-wise
print(countries.items())
```
