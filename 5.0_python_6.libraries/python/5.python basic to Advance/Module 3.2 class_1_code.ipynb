Used to combine multiple conditional statements.
- and (Returns True if both statements are true)
- or (Returns True if at least one statement is true)
- not (Reverses the logical state of its operand)

```python
a =45
b =5
c = 2
# a greater
print(a>b)
print(a>c)
```

```python
# conditional statements
# if else
number =33

if number % 2 ==0:
  print("even")
else:
  print("odd")
```

```python
# tarniary operators
# value_if_true if condition else value_if_false
result = "Even" if number % 2 == 0 else "Odd"
print(result)
```

```python
# Even numbers always end with a 0 in their least significant bit (LSB), which represents the 1's place (2â°).
# Odd numbers always end with a 1
# bitwise Operators
if number & 1 == 0:
    print("Even")
else:
    print( "Odd")
```
