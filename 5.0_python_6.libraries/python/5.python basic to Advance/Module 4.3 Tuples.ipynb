A tuple in Python is an ordered collection of elements that is immutable. This means that once a tuple is created, its elements cannot be changed, added, or removed.

Key characteristics of Python tuples:
- Ordered:
- Immutable:
- Heterogeneous:

### Creating and Accessing Tuples


```python
# Defining tuples using parentheses ()
# 1.An empty tuple
t = ()

# 2.A tuple with mixed data types
my_tuple = ("apple", 10, 3.14, True)

# 3.
names =["a","b","c"]
t1 = tuple(names)

# 4.
t2 = (2,)

# 5.
t3 = tuple()
```

```python
type(t)
```

#### Accessing elements using indexing & slicing


```python
# indexing
my_tuple[2]

# slicing
my_tuple[1:3]
```

### Tuple Operations


```python
# Concatenation (+) and repetition (*)
t1+t2
t1*3
# Membership testing (in, not in)
"a" in t1

# Iterating through tuples
for i in t1:
  print(i)
```

```python
for i in t1:
  if(i=="b"):
    print(i*2)
    continue
  print(i)
```
