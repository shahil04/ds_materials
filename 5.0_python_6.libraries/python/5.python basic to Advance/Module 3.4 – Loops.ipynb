## While Loop

```python
# Syntax
# while condition:
#     # code block
```

```python
for i in range(1,11):
  print(i)
```

```python
j =1 # start
while j<=10: # condition
  print(j)
  j +=1  # if number use increment
```

```python
# print the number from 10 to 1(reverse counting) using while loop
num =10
while num>=1:
  print(num)
  num -=1
```

```python
# print the number from 10 to 1(reverse counting) using for loop
for i in range(10,0,-1):
  print(i)
```

```python
# run loop in 3times --> while
# print number from 1 to 3
db_user_name ="shahil"
db_password = 123

attampts =3
while attampts>0:
  print("you have only",attampts, "attampts")
  user = input("enter your user name: ")

  if user == db_user_name:
    password = int(input("Enter your password:(only numbers) "))
    if password == db_password:
      print("login success")
      break
    else:
      print("login failed incorrect password")
      attampts -=1

  else:
    print("login failed incorrect username")

  attampts -=1
```

```python
# print the star format
# *
# * *
# * * *
# * * * *
# reverse pattern
n = 4
i = 1
while i <= n:
    print(" " * (n - i) + "*" * i)
    i += 1
```
