## Loop

In programming, a loop is a control flow statement that executes a block of code repeatedly until a specified condition is met

* types of Loop
1. for
2. while

- 1. for Loop:

for loops are used for iterating(get one by one ) over a sequence (like a list, tuple, string, or range) or other iterable objects.  (fixed number of iterations)

2. while loops:
while loops execute a block of code as long as a specified condition remains True. It's crucial to ensure that the condition eventually becomes False to avoid an infinite loop.

```python
# Write a program to check if a number is an Armstrong number. 
# Note: Armstrong number = sum of cubes of its digits equals the
#  number Example: Input â†’ 153, Output â†’ True
# 1**3+5**3+3**3 = 153
# 51
# 5**2+1**2 =26
num = 1532
num2 =0
p= len((str(num)))
p
for i in str(num):
    digit = int(i)
    num2 = num2 + digit**p

print('new number',num2)
if num2 == num:
    print(num, "is armstrong number")
else:
    print(num, "is not armstrong number")
```
