The **continue** statement in Python is a control flow tool used inside loops (for and while) to skip the rest of the code in the current iteration and immediately move to the next iteration of the loop. 

```python
for i in range(10):
    print(i)
```

```python
# continue 
for  i in range(10):
    if i==5:
        continue
    print(i)
```

```python
# Continue
# The continue statement in Python is a control flow statement used within loops (both for and while loops) 
# to skip the current iteration and proceed to the next iteration of the loop
for i in range(10):
  if i ==3:
    continue
  print(i)
  print("hii"*i)
```

### 12. Interview Question

Q: What is the difference between for loop and while loop?

| for loop                                | while loop                          |
| --------------------------------------- | ----------------------------------- |
| Used when number of iterations is known | Used when iterations are unknown    |
| Example: 1 to 10                        | Example: ask until correct password |
| More compact                            | More flexible                       |

Sample Answer

“A for loop is used when we know how many times the loop should run, while a while loop is used when the loop should continue until a condition becomes false.”

### Remember

* while repeats code while a condition is True.

* Always update the loop variable.

* break stops the loop.

* continue skips the current iteration.

* Use while when the number of repetitions is not known in advance.
