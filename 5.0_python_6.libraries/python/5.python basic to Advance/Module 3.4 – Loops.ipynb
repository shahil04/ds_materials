## Loop

In programming, a loop is a control flow statement that executes a block of code repeatedly until a specified condition is met

* types of Loop
1. for
2. while

- 1. for Loop:

for loops are used for iterating(get one by one ) over a sequence (like a list, tuple, string, or range) or other iterable objects.  (fixed number of iterations)

2. while loops:
while loops execute a block of code as long as a specified condition remains True. It's crucial to ensure that the condition eventually becomes False to avoid an infinite loop.

#### Continue


The **continue** statement in Python is a control flow tool used inside loops (for and while) to skip the rest of the code in the current iteration and immediately move to the next iteration of the loop. 

```python
# Continue
# The continue statement in Python is a control flow statement used within loops (both for and while loops) 
# to skip the current iteration and proceed to the next iteration of the loop
for i in range(10):
  print(i)
  if i ==7:
    continue
  print(i+10)
  print("hii")
```

```python
# continue 
for  i in range(10):
    if i==3:
        continue
    print(i)
```
