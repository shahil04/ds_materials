## Loop

In programming, a loop is a control flow statement that executes a block of code repeatedly until a specified condition is met

* types of Loop
1. for
2. while

- 1. for Loop:

for loops are used for iterating(get one by one ) over a sequence (like a list, tuple, string, or range) or other iterable objects.  (fixed number of iterations)

2. while loops:
while loops execute a block of code as long as a specified condition remains True. It's crucial to ensure that the condition eventually becomes False to avoid an infinite loop.

```python
# Q .give 3attempts to user for login using while
# run loop in 3times --> while
db_user_name ="shahil"
db_password = 123

attampts =1
while attampts<=3:
  user = input("enter your user name: ")

  if user == db_user_name:
    password = int(input("Enter your password:(only numbers) "))
    if password == db_password:
      print("login success")
      break
    else:
      print("login failed incorrect password")
  else:
    print("login failed incorrect username")

  attampts +=1
```

```python
# Q .give 3attempts to user for login using while
# run loop in 3times --> while
db_user_name ="shahil"
db_password = 123

attampts =3
while attampts>0:
  print("you have only",attampts, "attampts")
  user = input("enter your user name: ")

  if user == db_user_name:
    password = int(input("Enter your password:(only numbers) "))
    if password == db_password:
      print("login success")
      break
    else:
      print("login failed incorrect password")
      attampts -=1

  else:
    print("login failed incorrect username")

  attampts -=1
```
