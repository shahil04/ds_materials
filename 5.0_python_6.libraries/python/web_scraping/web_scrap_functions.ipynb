```python
import requests
from bs4 import BeautifulSoup as bs

url ='https://www.yellowpages.com/search?search_terms=Italian+Restaurants&geo_location_terms=Ottawa%2C+ON'

def  get_links(url):

    page = requests.get(url)
    page.status_code

    soup = bs(page.content, "html.parser")

    return soup

var =get_links(url)
# extract number and details

def extract_number(soup):
    details=[]
    indivisual_ph=[]
    indivisual_name=[]
    ph = soup.findAll("div",class_="phones phone primary")
    print(ph)
    for p in ph:
        print(p.get_text())
        indivisual_ph.append(p.get_text())
        

    name = soup.findAll("a",class_="business-name")
    print(name)
    base_url = "https://www.yellowpages.com"
    for p in name:
        print(p.get_text())
        print(base_url+p["href"])
        indivisual_name.append(p.get_text())

    for i  in zip(indivisual_name,indivisual_ph):
        details.append(i)
        
    print(details)
# image extract and save 
def image_extract(soup1):
    name = soup1.find("div",class_="carousel")
    img = name.findAll("img")
    urls=[]
    for i in img:
        # print(i)
        urls.append(i["data-url"])
        print(i["data-url"])

    for i in range(0, len(img)):
        load = requests.get(urls[i])
        print(load)
        with open("img"+str(i)+".jpg", "wb") as file:
            file.write(load.content) 
```
