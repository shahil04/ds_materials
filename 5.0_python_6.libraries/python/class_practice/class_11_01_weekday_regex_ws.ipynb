#  Web scraping

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url ='https://www.yellowpages.com/search?search_terms=restaurant&geo_location_terms=Los+Angeles%2C+CA'

page = requests.get(url)
page.status_code
```

```python
print(page.content)
```

### Find the phone numbers from website

```python
soup = bs(page.content, "html.parser")

ph = soup.find("div",class_="phones phone primary")
print(ph.get_text())
ph
```

```python
soup = bs(page.content, "html.parser")

ph = soup.findAll("div",class_="phones phone primary")
for p in ph:
    print(p.get_text())
```

#### Find the restaurant name and link

```python
name = soup.findAll("a",class_="business-name")
print(name)
for t in name:
    print(t.get_text())
```

```python
name = soup.findAll("a",class_="business-name")
print(name)
link =[]
for t in name:
    print(t.get_text())
    l = "https://www.yellowpages.com"+t["href"]
    print(l)
    link.append(l)
```

```python
res_link = "https://www.yellowpages.com/los-angeles-ca/mip/oyabun-seafood-555210849?lid=1002166518717"

html = requests.get(res_link)
print(html.status_code)

soup1 = bs(html.content, "html.parser")

name = soup1.find("div",class_="carousel")
# print(name)


img = name.findAll("img")
urls=[]
for i in img:
    # print(i)
    urls.append(i["data-url"])
    print(i["data-url"])
```

```python
for i in range(0, len(img)):
    load = requests.get(urls[i])
    print(load)
    with open("img"+str(i)+".jpg", "wb") as file:
        file.write(load.content) 
```
