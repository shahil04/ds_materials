### **1. Line Plot**

**Definition:**  
A line plot is a type of plot that displays data points connected by straight lines. It's used to visualize the trend of data over a continuous interval or time series.

**Use Case:**  
Line plots are often used in time series data to understand trends over time.

**Syntax:**  
```python
plt.plot(x, y)
sns.lineplot(x, y)
```

**Example Code:**
```python

### **2. Histogram / Distplot**

**Definition:**  
A histogram is a plot that lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data.

**Use Case:**  
Histograms are useful for visualizing the distribution of data and identifying patterns such as skewness or modality.

**Syntax:**  
```python
plt.hist(data, bins)
sns.histplot(data, bins)
```

**Example Code:**

````python
# Generate random data
data = np.random.randn(1000)
data
````

````python

# Using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, alpha=0.7,color="red")
plt.title('Histogram')
plt.show()

# Using Seaborn
sns.histplot(data, bins=30,color="r")
plt.title('Histogram with Seaborn')
plt.show()
````

````python

# Using Seaborn
sns.histplot(data, bins=10,color="g")
plt.title('Histogram with Seaborn')
plt.show()
````

### **3. Bar Plot / Count Plot**

**Definition:**  
A bar plot represents categorical data with rectangular bars. The length of each bar is proportional to the value it represents.

**Use Case:**  
Bar plots are commonly used to compare the quantities of different categories.

**Syntax:** 
```python
plt.bar(x, height)
sns.barplot(x, y)
sns.countplot(x)
```

````python
# Generate random data
categories = ['A', 'B', 'C']
values = np.random.randint(10, 100, size=3)
print(categories)
print(values)
````

````python
# Using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.show()
````

````python
sns.barplot(x=categories,y=values,hue=categories)
````

````python
x=np.random.choice(categories, 100)
x
````

````python

# Using Seaborn for count plot
sns.countplot(x=np.random.choice(categories, 100))
plt.title('Count Plot with Seaborn')
plt.show()
````

### **4. Pie Chart**

**Definition:**  
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportions.

**Use Case:**  
Pie charts are used to show the proportions of categories in a dataset.

**Syntax:**  
```python
plt.pie(x, labels)
```

````python
# Generate random data
sizes = np.random.randint(10, 100, size=3)

# Using Matplotlib
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=categories, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'salmon'])
plt.title('Pie Chart')
plt.show()
````


### **5. Heat Map**

**Definition:**  
A heatmap is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions.

**Use Case:**  
Heatmaps are useful for showing the correlation between different variables.

**Syntax:**  
```python
sns.heatmap(data)
```

````python

# Generate random data
data = np.random.rand(10, 10)

# Using Seaborn
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
````

### **6. Scatter Plot**

**Definition:**  
A scatter plot displays values for two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.

**Use Case:**  
Scatter plots are used to identify the relationship between two variables.

**Syntax:**  
```python
plt.scatter(x, y)
sns.scatterplot(x, y)
```

````python
# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)

# Using Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.7)
plt.title('Scatter Plot')
plt.show()

# Using Seaborn
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot with Seaborn')
plt.show()
````

### **7. Box Plot**

**Definition:**  
A box plot is a standardized way of displaying the distribution of data based on a five-number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum").

**Use Case:**  
Box plots are used to understand the distribution of data, identify outliers, and compare distributions.

**Syntax:**  
```python
plt.boxplot(data)
sns.boxplot(data)
```


````python
# Generate random data
data = np.random.randn(100)

# Using Matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title('Box Plot')
plt.show()

# Using Seaborn
sns.boxplot(x=data)
plt.title('Box Plot with Seaborn')
plt.show()
````

## Titanic data visualizations

````python
### Step 1: Import Libraries and Load Titanic Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv('titanic.csv')

# Preview the first few rows
df.head()
````

````python

### Step 2: Basic Plotting with Matplotlib

#### 2.1 Line Plot

# Plot the Age distribution using a simple line plot
plt.figure(figsize=(8, 5))
plt.plot(df['Age'], label='Age')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.title('Age Distribution of Titanic Passengers')
plt.legend()
plt.show()
````

````python

# **Question**: How do you create a simple line plot with Matplotlib?

#### 2.2 Bar Plot


# Bar plot showing the count of passengers by class
plt.figure(figsize=(8, 5))
df['Pclass'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Class')
plt.show()
````

````python

# **Question**: How can you create a bar plot using Matplotlib?

#### 2.3 Histogram


# Histogram for Age distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=20, color='green', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')
plt.show()
````

````python

# **Question**: How do you visualize the distribution of a continuous variable using a histogram?

#### 2.4 Scatter Plot

# Scatter plot for Age vs. Fare
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Fare'], color='purple', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare Scatter Plot')
plt.show()
````

````python
# **Question**: How can you create a scatter plot to visualize relationships between two numerical variables?

#### 2.5 Box Plot


# Box plot of Fare
plt.figure(figsize=(8, 5))
plt.boxplot(df['Fare'].dropna(), vert=False, patch_artist=True)
plt.xlabel('Fare')
plt.title('Box Plot of Fare')
plt.show()
````

````python
# **Question**: How do you create a box plot to check for outliers?

### Step 3: Advanced Plotting with Seaborn

#### 3.1 Count Plot (Categorical Variable)

# Count plot for Passenger Class using Seaborn
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title('Count of Passengers by Class')
plt.show()
````

````python
# **Question**: How can you visualize the count of categorical variables?

#### 3.2 Distribution Plot

# Distribution plot for Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='blue')
plt.title('Age Distribution with KDE')
plt.show()
````

````python
# **Question**: How do you visualize the distribution of data with a KDE (Kernel Density Estimation)?
# 
#### 3.3 Violin Plot


# Violin plot for Age distribution by Passenger Class
plt.figure(figsize=(8, 5))
sns.violinplot(x='Pclass', y='Age', data=df, palette='coolwarm', inner='quartile')
plt.title('Age Distribution by Class (Violin Plot)')
plt.show()
````

````python
import pandas as pd
import numpy as np

# Load the Titanic dataset
titanic_df = pd.read_csv('titanic.csv')

# Select only the numeric columns
numeric_titanic_df = titanic_df.select_dtypes(include=[np.number])

# Print the numeric columns
print(numeric_titanic_df.columns)
````

````python
# **Question**: How do you create a violin plot to show data distribution with respect to a categorical variable?

#### 3.4 Heatmap (Correlation Matrix)

# Correlation heatmap for numerical columns
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_titanic_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
````

````python
# **Question**: How do you visualize correlations between numerical variables in a dataset?

#### 3.5 Pair Plot

# Pair plot for selected features
plt.figure(figsize=(8, 5))
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']].dropna(), hue='Survived', palette='Set1')
plt.title('Pair Plot of Titanic Dataset Features')
plt.show()
````

````python
# **Question**: How do you visualize relationships between multiple variables using pair plots?

#### 3.6 Bar Plot (Grouped Data)


# Bar plot for average fare by class and survival status
plt.figure(figsize=(8, 5))
sns.barplot(x='Pclass', y='Fare', hue='Survived', data=df, palette='Set1')
plt.title('Average Fare by Class and Survival Status')
plt.show()
````

````python
# **Question**: How can you visualize comparisons between two categories using grouped data?

#### 3.7 Box Plot (Categorical Comparison)

# Box plot for Fare by Passenger Class and Survival
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df, palette='Set3')
plt.title('Box Plot of Fare by Class and Survival')
plt.show()
````

````python

# **Question**: How do you compare numerical variables between categories using box plots?

#### 3.8 FacetGrid (Multiple Plots)

# FacetGrid for Age distribution by Survival
g = sns.FacetGrid(df, col='Survived', height=5, aspect=1.2)
g.map(sns.histplot, 'Age', kde=True)
plt.show()
````

````python
# **Question**: How do you create multiple plots based on subsets of data using Seaborn's `FacetGrid`?

### Step 4: Customizing and Styling Plots

#### 4.1 Adding Titles and Labels

# Customizing plot with title, labels, and grid
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='red')
plt.title('Age Distribution of Titanic Passengers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(True)
plt.show()
````

````python
# **Question**: How can you add titles, labels, and grid lines to customize a plot?

#### 4.2 Changing Style with Seaborn

# Changing Seaborn style
sns.set_style('whitegrid')

# Create a plot with the new style
plt.figure(figsize=(8, 5))
sns.histplot(df['Fare'], color='purple', bins=30, kde=True)
plt.title('Fare Distribution with Whitegrid Style')
plt.show()
````

````python
# **Question**: How do you change the visual style of plots using Seaborn?

#### 4.3 Saving the Plot

# Save a plot as an image file
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title('Passenger Count by Class')
plt.savefig('passenger_count_by_class.png')
plt.show()
````

````python

# **Question**: How can you save your plot as an image file?

### Step 5: Combining Matplotlib and Seaborn

# You can combine Seaborn with Matplotlib to further enhance customizations:

# Combine Matplotlib and Seaborn for custom styling
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=True, color='blue')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Customize ticks and grid with Matplotlib
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.7)
plt.show()
````
