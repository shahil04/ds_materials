```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url ='https://www.yellowpages.com/search?search_terms=restaurant&geo_location_terms=Los+Angeles%2C+CA'

page = requests.get(url)
page.status_code

```

```python
print(page.content)
```

### Find the phone numbers from website

```python
soup = bs(page.content, "html.parser")

ph = soup.find("div",class_="phones phone primary")
ph.get_text()
```

```python
soup = bs(page.content, "html.parser")

ph = soup.findAll("div",class_="phones phone primary")
for p in ph:
    print(p.get_text())
```

```python

name = soup.findAll("a",class_="business-name")
print(name)
base_url = "https://www.yellowpages.com"
for p in name:
    print(p.get_text())
    print(base_url+p["href"])
```
