```python
# generate the data using linspace
x2 = np.linspace(1,10, 100)

plt.title("sin Graph")
# line 1
plt.plot(np.sin(x2),c="r",ls ="-.",label="sin line")

# line 2 
plt.plot(np.cos(x2),c="r",ls ="-",marker="o",mfc="m" ,mec="g",label="sin line")
plt.xlabel("data points")
plt.ylabel("sin value")

plt.xticks(np.arange(0,101,10))
plt.yticks(np.arange(-2,1,.20))

plt.ylim(-2,2)
plt.xlim(0,150)

plt.grid()
plt.legend(loc="lower left")

plt.show()
# seprate plot for cos graph
plt.plot(np.cos(x2),c="r",ls ="-",marker="s",mfc="m" ,mec="g",label="sin line")
```

```python
sales= [300,200,340,560]
cat = ["A","b","c","d"]
plt.bar(cat,sales,)
```

```python
sns.barplot(x=cat,y=sales,color="g")
```

```python
# indivisual color each bar
sns.barplot(x=cat,y=sales,palette=["g","b","r","m"])
```

```python
# indivisual color each bar
sns.barplot(x=cat,y=sales,hue=cat)
```

```python
# indivisual color each bar as gradient
ax = sns.barplot(x=cat,y=sales,hue=cat, palette="Greens",legend=True)

```

```python
ax.bar_label(ax.containers[3])
```

```python
# add data label in  each bar
sns.lineplot(x=cat,y=sales)
ax = sns.barplot(x=cat,y=sales,hue=cat, palette="Greens",legend=True,width=.5)

for i in range(len(cat)):
    ax.bar_label(ax.containers[i])
plt.title("catagory wise sales")
plt.show()
```

```python
# Creating explode data
exp = (0.1, 0.0, 0.2, 0.3)

# Creating color parameters
col = ("orange", "cyan", "brown",
          "grey")

a,b,c = plt.pie(sales,labels=cat,autopct="%.2f%%",explode=exp,colors=col)

plt.title('Sales by Category')

plt.show()
```
