### Create a DashBoard Using Gradio 

```python
import pandas as pd
import plotly.express as px
import gradio as gr

# -----------------------------
# Load Data
# -----------------------------

df = pd.read_csv("Module 2 Bengaluru_House_Data.csv")

# -----------------------------
# Data Cleaning
# -----------------------------
df['society'] = df['society'].fillna("Unknown")
df['balcony'] = df['balcony'].fillna(0)

# Clean location
df['location'] = df['location'].str.strip()

# Extract BHK
df['bhk'] = df['size'].str.extract('(\d+)').astype(float)

# Convert total_sqft
def convert_sqft(x):
    try:
        if '-' in str(x):
            a, b = x.split('-')
            return (float(a) + float(b)) / 2
        return float(x)
    except:
        return None

df['total_sqft'] = df['total_sqft'].apply(convert_sqft)

# Remove invalid rows
df = df[df['total_sqft'].notnull()]

# Feature Engineering
df['price_per_sqft'] = (df['price'] * 100000) / df['total_sqft']

# -----------------------------
# Filter Function
# -----------------------------
def filter_data(location, society, availability, price_range):
    data = df.copy()

    if location != "All":
        data = data[data['location'] == location]

    if society != "All":
        data = data[data['society'] == society]

    if availability != "All":
        data = data[data['availability'] == availability]

    # FIX: unpack tuple
    min_price, max_price = price_range

    data = data[
        (data['price'] >= min_price) &
        (data['price'] <= max_price)
    ]

    return data

# -----------------------------
# Dashboard Function
# -----------------------------
def update_dashboard(location, society, availability, price_range):

    data = filter_data(location, society, availability, price_range)

    if data.empty:
        return None, None, None, None

    # Scatter Plot
    fig1 = px.scatter(
        data,
        x="total_sqft",
        y="price",
        color="bhk",
        title="Price vs Total Sqft"
    )

    # Histogram
    fig2 = px.histogram(
        data,
        x="bhk",
        title="BHK Distribution"
    )

    # Box Plot
    fig3 = px.box(
        data,
        y="price_per_sqft",
        title="Price per Sqft Distribution"
    )

    # Top Locations
    top_loc = data.groupby('location')['price'].mean().sort_values(ascending=False).head(10)

    fig4 = px.bar(
        x=top_loc.index,
        y=top_loc.values,
        title="Top Locations by Avg Price"
    )

    return fig1, fig2, fig3, fig4

# -----------------------------
# Dropdown Values
# -----------------------------
locations = ["All"] + sorted(df['location'].dropna().unique().tolist())
societies = ["All"] + sorted(df['society'].dropna().unique().tolist())
availability_list = ["All"] + sorted(df['availability'].dropna().unique().tolist())

# Price range
min_price = int(df['price'].min())
max_price = int(df['price'].max())

# -----------------------------
# Gradio UI
# -----------------------------
with gr.Blocks() as demo:

    gr.Markdown("# 🏠 Bangalore Housing EDA Dashboard")

    # Filters
    with gr.Row():
        location_dd = gr.Dropdown(locations, value="All", label="Location")
        society_dd = gr.Dropdown(societies, value="All", label="Society")
        availability_dd = gr.Dropdown(availability_list, value="All", label="Availability")

    price_slider = gr.RangeSlider(
        minimum=min_price,
        maximum=max_price,
        value=(min_price, max_price),
        step=1,
        label="Price Range (Lakhs)"
    )

    btn = gr.Button("Apply Filters")

    # Charts
    with gr.Row():
        plot1 = gr.Plot()
        plot2 = gr.Plot()

    with gr.Row():
        plot3 = gr.Plot()
        plot4 = gr.Plot()

    # Button Action
    btn.click(
        fn=update_dashboard,
        inputs=[location_dd, society_dd, availability_dd, price_slider],
        outputs=[plot1, plot2, plot3, plot4]
    )

demo.launch()
```

```python
import pandas as pd
import plotly.express as px
import gradio as gr

# -----------------------------
# Load Data
# -----------------------------
df = pd.read_csv("clean_data_EDA.csv")

# -----------------------------
# Basic Cleaning (Safety)
# -----------------------------
df['location'] = df['location'].astype(str).str.strip()
df['society'] = df['society'].astype(str).str.strip().fillna("unknown")
df['availability'] = df['availability'].astype(str).str.strip()

# Ensure numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['total_sqft'] = pd.to_numeric(df['total_sqft'], errors='coerce')
df['price_per_sqft'] = pd.to_numeric(df['price_per_sqft'], errors='coerce')

# Drop nulls
df = df.dropna(subset=['price', 'total_sqft', 'price_per_sqft'])

# -----------------------------
# Filter Function
# -----------------------------
def filter_data(location, society, availability, price_range):
    data = df.copy()

    if location != "All":
        data = data[data['location'] == location]

    if society != "All":
        data = data[data['society'] == society]

    if availability != "All":
        data = data[data['availability'] == availability]

    min_price, max_price = price_range

    data = data[
        (data['price'] >= min_price) &
        (data['price'] <= max_price)
    ]

    return data

# -----------------------------
# Dashboard Function
# -----------------------------
def update_dashboard(location, society, availability, price_range):

    data = filter_data(location, society, availability, price_range)

    if data.empty:
        return None, None, None, None

    # Scatter Plot
    fig1 = px.scatter(
        data,
        x="total_sqft",
        y="price",
        color="size",
        title="Price vs Total Sqft"
    )

    # BHK / Size Distribution
    fig2 = px.histogram(
        data,
        x="size",
        title="BHK Distribution"
    )

    # Price per Sqft
    fig3 = px.box(
        data,
        y="price_per_sqft",
        title="Price per Sqft Distribution"
    )

    # Top Locations
    top_loc = data.groupby('location')['price'].mean().sort_values(ascending=False).head(10)

    fig4 = px.bar(
        x=top_loc.index,
        y=top_loc.values,
        title="Top Locations by Avg Price"
    )

    return fig1, fig2, fig3, fig4

# -----------------------------
# Dropdown Values
# -----------------------------
locations = ["All"] + sorted(df['location'].unique().tolist())
societies = ["All"] + sorted(df['society'].unique().tolist())
availability_list = ["All"] + sorted(df['availability'].unique().tolist())

# Price Range
min_price = int(df['price'].min())
max_price = int(df['price'].max())

# -----------------------------
# Gradio UI
# -----------------------------
with gr.Blocks() as demo:

    gr.Markdown("# 🏠 Bangalore Housing EDA Dashboard")

    with gr.Row():
        location_dd = gr.Dropdown(locations, value="All", label="Location")
        society_dd = gr.Dropdown(societies, value="All", label="Society")
        availability_dd = gr.Dropdown(availability_list, value="All", label="Availability")

    price_slider = gr.RangeSlider(
        minimum=min_price,
        maximum=max_price,
        value=(min_price, max_price),
        step=1,
        label="Price Range (Lakhs)"
    )

    btn = gr.Button("Apply Filters")

    with gr.Row():
        plot1 = gr.Plot()
        plot2 = gr.Plot()

    with gr.Row():
        plot3 = gr.Plot()
        plot4 = gr.Plot()

    btn.click(
        fn=update_dashboard,
        inputs=[location_dd, society_dd, availability_dd, price_slider],
        outputs=[plot1, plot2, plot3, plot4]
    )

demo.launch()
```
