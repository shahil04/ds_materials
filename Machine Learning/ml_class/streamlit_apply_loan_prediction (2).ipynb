```python
# bar chart  --> coutinous ,categorical data
product = ["apple", "banana", "cherry"]
sales = [300,340,370]

# create bar chart using matplotlib
plt.figure(figsize=(4,2))

plt.bar(product,sales, zorder = 2)
plt.grid(zorder=0)
plt.title("categories_wise sales")
plt.xlabel("categories")
plt.ylabel("sales amount")

plt.show()
```

```python
plt.barh(product,sales)
```

```python
# bar seaborn
import seaborn as sns
plt.figure(figsize=(4,2))

# 
ax = sns.barplot( x= product, y= sales, palette=["red","blue","green"], label="b")

# data label
for i in range(len(product)):
    ax.bar_label(ax.containers[i])

plt.title("categories_wise sales")
plt.xlabel("categories")
plt.ylabel("sales amount")
plt.legend()
plt.show()
```

```python
# pie chart 
e = [.2, 0.1, 0]
plt.pie(sales, labels=product,autopct="%.2f%%", explode=e)

plt.show()
```
