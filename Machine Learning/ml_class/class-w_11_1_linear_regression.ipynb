```python
#  librarise for ml 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python

# #read the data
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()

diabetes
```

```python
print(diabetes.DESCR)
```

```python
diabetes.feature_names
```

```python
diabetes.target
```

```python
diabetes.data
```

```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
data.head()
```

```python
data['target'] = diabetes.target
data.head()
```

```python
# data prepocessing
#  data cleaning and transform
```

```python
#  define x and y

# simple linear regression --> one feature and 1 target
x =data[["bmi"]]
x
```

```python
y = data["target"]
```

```python
# split the data for training and  testing 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20,    
                                                        random_state = 111)
X_train
```

```python
# model impoert
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(X_train,y_train)
```

```python
# y =mx+c
lr.coef_ # m ,beta1
```

```python
# c ,intercept
lr.intercept_

# y = 967*x + 150
```

```python
#  predict 
pr = lr.predict(X_test)
```

```python
pr
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mean_squared_error(y_test, pr)
```

```python
mean_absolute_error(y_test,pr)
```

```python
r2_score(y_test,pr)
```

```python
#visualised the result
plt.scatter(X_test, y_test, color = 'black', label = "Actual data")
plt.plot(X_test, pr, color = 'red', linewidth = 3, label = "linear regression line")
# y = 967*x + 150
plt.xlabel("BMI")
plt.ylabel("One year preogression-target")
plt.title("My first linear regression model")
plt.legend()
plt.show()
```
