### 2. Face detection

````python
# Import necessary libraries
import cv2
import numpy as np
# Load the pre-trained Haar cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Start capturing video from the webcam
video_capture = cv2.VideoCapture(0)
if not video_capture.isOpened():
    print("Error accessing the camera")
    exit()
# Main loop to process each video frame
while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    if not ret:
        print("Error reading frame from webcam")
        break

    # Convert the captured frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the grayscale image
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    # Draw a rectangle around each detected face
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        # text add 
        cv2.putText(frame, 'Hello', (x, y), cv2.FONT_HERSHEY_SIMPLEX,
            2, (255, 255, 255), 3)

        
    # Display the resulting frame
    cv2.imshow('Face Detection', frame)
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
video_capture.release()
cv2.destroyAllWindows()
````

````python
import cv2
import time

# Load the Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Open the webcam (0 = default camera)
cap = cv2.VideoCapture(0)

# Check if camera opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Define the codec and create VideoWriter object
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out = cv2.VideoWriter('output_face.avi', cv2.VideoWriter_fourcc(*'XVID'), 30.0, (frame_width, frame_height))

start_time = time.time()
duration = 100  # seconds

print("Recording started...")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('Face Detection', frame)

    # Write the frame to the video file
    out.write(frame)

    # Stop after 10 seconds
    if time.time() - start_time > duration:
        print("Recording finished.")
        break

    # Press 'q' to exit early
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Recording stopped by user.")
        break

# Release everything
cap.release()
out.release()
cv2.destroyAllWindows()
````

````python
import cv2
import time

# Load the Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# Open the webcam (0 = default camera)
cap = cv2.VideoCapture(0)

# Check if camera opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Set capture properties (optional, for some webcams)
cap.set(cv2.CAP_PROP_FPS, 30)  # Try to force 30 FPS

# Get frame dimensions
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Get actual FPS from webcam (some webcams ignore CAP_PROP_FPS)
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0:  # fallback if camera returns 0
    fps = 30.0

print(f"Recording at {fps} FPS")

# Define the codec and create VideoWriter object
out = cv2.VideoWriter(
    'output_face.avi',
    cv2.VideoWriter_fourcc(*'XVID'),
    fps,  # Match actual capture FPS
    (frame_width, frame_height)
)

start_time = time.time()
duration = 10  # seconds

print("Recording started... Press 'q' to stop.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5
    )

    # Draw rectangles around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('Face Detection', frame)

    # Write the frame to the video file
    out.write(frame)

    # Stop after the set duration
    if time.time() - start_time > duration:
        print("Recording finished.")
        break

    # Press 'q' to exit early
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Recording stopped by user.")
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
````

````python
import cv2
import tkinter as tk
from PIL import Image, ImageTk
from datetime import datetime
import os

# ===============================
# Load Haar Cascade
# ===============================
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# ===============================
# Tkinter Camera App
# ===============================
class CameraApp:
    def __init__(self, window):
        self.window = window
        self.window.title("Camera App - Face Detection")
        self.window.geometry("900x650")

        self.cap = cv2.VideoCapture(0)
        self.current_frame = None
        self.running = False

        # Video display
        self.video_label = tk.Label(window)
        self.video_label.pack()

        # Buttons
        btn_frame = tk.Frame(window)
        btn_frame.pack(pady=10)

        tk.Button(btn_frame, text="â–¶ Start", command=self.start_camera).grid(row=0, column=0, padx=10)
        tk.Button(btn_frame, text="â¹ Stop", command=self.stop_camera).grid(row=0, column=1, padx=10)
        tk.Button(btn_frame, text="ğŸ“¸ Capture", command=self.capture_image).grid(row=0, column=2, padx=10)

        self.update_frame()
        self.window.protocol("WM_DELETE_WINDOW", self.on_close)

    # ===============================
    def start_camera(self):
        self.running = True

    def stop_camera(self):
        self.running = False

    # ===============================
    def capture_image(self):
        if self.current_frame is not None:
            os.makedirs("captures", exist_ok=True)
            filename = datetime.now().strftime("captures/capture_%Y%m%d_%H%M%S.jpg")
            cv2.imwrite(filename, self.current_frame)
            print(f"âœ… Image saved: {filename}")

    # ===============================
    def update_frame(self):
        if self.running:
            ret, frame = self.cap.read()
            if ret:
                self.current_frame = frame.copy()

                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                faces = face_cascade.detectMultiScale(gray, 1.1, 5, minSize=(30, 30))

                for (x, y, w, h) in faces:
                    cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

                frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                img = Image.fromarray(frame_rgb)
                imgtk = ImageTk.PhotoImage(image=img)

                self.video_label.imgtk = imgtk
                self.video_label.configure(image=imgtk)

        self.window.after(10, self.update_frame)

    # ===============================
    def on_close(self):
        self.running = False
        self.cap.release()
        self.window.destroy()

# ===============================
# Run App
# ===============================
root = tk.Tk()
CameraApp(root)
root.mainloop()
````
