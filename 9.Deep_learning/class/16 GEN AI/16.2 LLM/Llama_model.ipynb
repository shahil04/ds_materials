````python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
from google.colab import userdata
import torch # Import torch to specify dtype

# Load tokenizer and model
model_name = "meta-llama/Llama-2-7b-chat-hf"

# Get Hugging Face token from Colab secrets
hf_token = userdata.get('HF_TOKEN') # Assuming you named your secret 'HF_TOKEN'

# Configure 4-bit quantization
bnb_config = BitsAndBytesConfig(load_in_4bit=True)

tokenizer = AutoTokenizer.from_pretrained(model_name, token=hf_token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16, # Keep float16 for potential compatibility, though quantization is the main factor now
    token=hf_token,
    quantization_config=bnb_config
)

# Text generation pipeline
llama_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Generate response
prompt = "Explain what is quantum computing in simple terms."
output = llama_pipeline(prompt, max_length=200, do_sample=True, temperature=0.7)

print(output[0]['generated_text'])
````

````python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
import torch
from google.colab import userdata

model_name = "meta-llama/Llama-2-7b-chat-hf"
hf_token = userdata.get('HF_TOKEN')  # Or set manually: hf_token = "hf_..."

bnb_config = BitsAndBytesConfig(load_in_4bit=True)

tokenizer = AutoTokenizer.from_pretrained(model_name, token=hf_token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16,
    use_auth_token=hf_token,
    quantization_config=bnb_config
)

llama_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

prompt = "Explain what is quantum computing in simple terms."
output = llama_pipeline(prompt, max_length=200, do_sample=True, temperature=0.7)

print(output[0]['generated_text'])
````
