```python
from moviepy.editor import VideoFileClip

def convert_mp4_to_mp3(mp4_path, mp3_path):
    """
    Converts an MP4 video file to an MP3 audio file.

    Args:
        mp4_path (str): The path to the input MP4 file.
        mp3_path (str): The desired path for the output MP3 file.
    """
    try:
        video_clip = VideoFileClip(mp4_path)
        audio_clip = video_clip.audio
        audio_clip.write_audiofile(mp3_path)
        audio_clip.close()
        video_clip.close()
        print(f"Successfully converted '{mp4_path}' to '{mp3_path}'")
    except Exception as e:
        print(f"Error converting file: {e}")

# Example usage:
input_mp4_file = "/content/video1584088381.mp4"  # Replace with the actual path to your MP4 file
output_mp3_file = "audio1.mp3" # Desired name for the output MP3 file

convert_mp4_to_mp3(input_mp4_file, output_mp3_file)
```
