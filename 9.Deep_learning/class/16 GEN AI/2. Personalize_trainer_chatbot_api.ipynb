# ü§ñ **RASA Chatbot Development**

---

## üí¨ **1. What is RASA?**

### üîπ Definition:

**RASA** is an open-source framework for building **AI-powered conversational chatbots** that can understand natural language and respond intelligently.
It helps developers create **contextual**, **customizable**, and **on-premise** chatbots using **NLP + Machine Learning**.

---

### üîπ Components of RASA:

| Component                                     | Description                                                     |
| --------------------------------------------- | --------------------------------------------------------------- |
| **Rasa NLU (Natural Language Understanding)** | Extracts intent and entities from user input (what user wants). |
| **Rasa Core**                                 | Manages conversation flow (decides next response/action).       |
| **Rasa Action Server**                        | Executes custom Python code (like calling APIs or databases).   |

---

### üîπ Why Use RASA?

‚úÖ Open-source (no cloud dependency)
‚úÖ Integrates with any front-end (Telegram, WhatsApp, Slack, etc.)
‚úÖ Supports custom ML models
‚úÖ Secure, can be deployed on private servers

---

### üîπ Example Chatbot Flow

```
User: Hi
Bot: Hello! How can I help you today?
User: Book a flight to Delhi
Bot: Sure! What date would you like to travel?
```

## üí¨ **5. Intents, Entities, Responses, and Stories**

### üîπ **Intents**

The **goal** or **purpose** behind a user‚Äôs message.

Example from `data/nlu.yml`:

```yaml
nlu:
- intent: greet
  examples: |
    - hello
    - hi
    - hey there

- intent: ask_weather
  examples: |
    - what's the weather
    - tell me the weather in Delhi
    - weather of Mumbai
```

---

### üîπ **Entities**

Specific **data values** extracted from user messages (like names, cities, dates).

Example:

```
User: Weather in Delhi
Entity: city = "Delhi"
```

---

### üîπ **Responses**

Predefined bot replies defined in `domain.yml`.

```yaml
responses:
  utter_greet:
    - text: "Hi there! How can I help you?"
```

---

### üîπ **Stories**

Define **conversation paths** ‚Äî how bot should respond to certain intents.

Example from `data/stories.yml`:

```yaml
stories:
- story: weather path
  steps:
  - intent: greet
  - action: utter_greet
  - intent: ask_weather
  - action: utter_ask_city
  - action: action_weather_api
```

---

### üîπ **Rules**

Used for specific conditions like greetings or FAQ-type replies.

```yaml
rules:
- rule: Say goodbye anytime user says goodbye
  steps:
  - intent: goodbye
  - action: utter_goodbye
```

## üß∞ Step 7: Custom Action for Weather

**File:** `actions/actions.py`

```python
from rasa_sdk import Action, Tracker
from rasa_sdk.executor import CollectingDispatcher
import requests

class ActionGetWeather(Action):

    def name(self) -> str:
        return "action_get_weather"

    def run(self, dispatcher: CollectingDispatcher,
            tracker: Tracker,
            domain: dict):

        city = tracker.get_slot("city")

        if not city:
            dispatcher.utter_message(text="Please provide a city name.")
            return []

        api_key = "YOUR_OPENWEATHER_API_KEY"
        url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

        response = requests.get(url).json()

        if response.get("cod") != 200:
            dispatcher.utter_message(text=f"Sorry, I couldn't find weather for {city}.")
        else:
            temp = response["main"]["temp"]
            desc = response["weather"][0]["description"]
            dispatcher.utter_message(text=f"The weather in {city} is {temp}¬∞C with {desc}.")

        return []
```

## üí¨ Step 3: Create Streamlit App

Create a new file named **`app.py`** inside your RASA project folder.

```python
import streamlit as st
import requests

# Streamlit page config
st.set_page_config(page_title="Weather Chatbot", page_icon="‚õÖ", layout="centered")

st.title("üå¶Ô∏è Weather Chatbot (RASA + Streamlit)")
st.markdown("Ask me about the current weather in any city!")

# Initialize session state for chat history
if "messages" not in st.session_state:
    st.session_state.messages = []

# Function to send message to RASA REST API
def send_message_to_rasa(message):
    response = requests.post(
        "http://localhost:5005/webhooks/rest/webhook",
        json={"sender": "user", "message": message},
    )
    messages = response.json()
    return messages

# Chat input
user_input = st.chat_input("Type your message...")

if user_input:
    # Display user message
    st.session_state.messages.append({"role": "user", "content": user_input})
    rasa_responses = send_message_to_rasa(user_input)

    # Display RASA response(s)
    for res in rasa_responses:
        if "text" in res:
            st.session_state.messages.append({"role": "assistant", "content": res["text"]})

# Display conversation
for msg in st.session_state.messages:
    if msg["role"] == "user":
        with st.chat_message("user"):
            st.markdown(msg["content"])
    else:
        with st.chat_message("assistant"):
            st.markdown(msg["content"])
```

## **9Ô∏è‚É£ Custom Action: Weather API**

**File:** `actions/actions.py`

```python
from rasa_sdk import Action, Tracker
from rasa_sdk.executor import CollectingDispatcher
import requests

class ActionGetWeather(Action):

    def name(self) -> str:
        return "action_get_weather"

    def run(self, dispatcher: CollectingDispatcher,
            tracker: Tracker,
            domain: dict):

        city = tracker.get_slot("city")

        if not city:
            dispatcher.utter_message(text="Please provide a city name.")
            return []

        api_key = "YOUR_OPENWEATHER_API_KEY"
        url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

        response = requests.get(url).json()

        if response.get("cod") != 200:
            dispatcher.utter_message(text=f"Sorry, I couldn't find weather for {city}.")
        else:
            temp = response["main"]["temp"]
            desc = response["weather"][0]["description"]
            dispatcher.utter_message(text=f"The weather in {city} is {temp}¬∞C with {desc}.")

        return []
```

> Replace `YOUR_OPENWEATHER_API_KEY` with your OpenWeatherMap API key.
