### use generative models 

A generative AI model is a type of artificial intelligence that creates new, original content—such as text, images, audio, or video—based on patterns learned from vast amounts of data. Unlike traditional AI that follows pre-programmed rules, generative models can produce novel outputs by learning from existing data and predicting new possibilities.
