# remove urls

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```

### remove punctuations

the marks used for dividing writing into sentences and phrases

Here are 14 common punctuation marks with examples: period (.), comma (,), question mark (?), exclamation mark (!), colon (:), semicolon (;), apostrophe ('), quotation marks (" "), hyphen (-), dash (—), parentheses ( ), brackets (), braces ({ }), and ellipsis (...)

```python
import string
string.punctuation
```

```python
exclude = string.punctuation

def remove_punc(text):
  for c in exclude:
    text = text.replace(c, "")
  return text
# remove_punc(r1)
```

```python
# for all data
df["review"] = df["review"].apply(remove_punc)
```

### chat word treatment

```python
chat_words = {
    "AFAIK": "As Far As I Know",
    "AFK": "Away From Keyboard",
    "ASAP": "As Soon As Possible",
    "ATK": "At The Keyboard"}
# https://github.com/rishabhverma17/sms_slang_translator
```

```python
chat_words["AFK"]
```

```python
def chat_conversion(text):
  new_text = []
  for w in text.split():
      if w.upper() in chat_words:
          new_text.append(chat_words[w.upper()])
      else:
          new_text.append(w)

  return " ".join(new_text)
```

```python
chat_conversion("AFK is not working")
```

### remove stop words

```python
#  a the of are my
import nltk
from nltk.corpus import stopwords
# corpus --> a collection of written or spoken texts

# Download the stopwords data
nltk.download('stopwords')

# stopwords.words('english')
# add custom words in stopwords

# remove some  words from stopwords
```

```python
def remove_stopwords(text):
  new_text = []
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)

  x = new_text[:]
  new_text.clear()
  return " ".join(x)

remove_stopwords(r1)
```

### Handling Emojis


```python
import re

# Sample text with an emoji
text = "Hello world! 👋"

# Compile a regex to match emojis
# This pattern includes emoticons and symbols
emoji_pattern = re.compile("[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF]+", flags=re.UNICODE)

# Remove the emoji from the text
text_without_emoji = emoji_pattern.sub(r'', text)

print(text_without_emoji)
```

#### Stemming: Reduce words to root form (may be crude)
types

1.Porter Stemmer: running → run


2.Snowball Stemmer: More aggressive/stable than Porter



```python
from nltk.stem import PorterStemmer, SnowballStemmer
ps = PorterStemmer()
ss = SnowballStemmer("english")

print(ps.stem("running"))   # run
print(ss.stem("running"))   # run
```

```python
ps.stem("studies")
ss.stem("studies")
```
