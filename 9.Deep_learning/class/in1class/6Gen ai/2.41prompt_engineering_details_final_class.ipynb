#### 1.Tokens:

LLMs do not read words; they read "tokens" (chunks of characters, syllables, or entire words). Roughly 1 token equals 0.75 words. This means 1,000 tokens ≈ 750 words

#### 2. Context Window: 
This is the model's short-term memory—the total number of tokens (input + output) the AI can "see" and process at once. 

#### 3. Temperature: 
This setting (usually ranging from 0 to 2) controls the "risk" the model takes when predicting the next token. A low temperature (0 to 0.2) is predictable and factual, making it ideal for coding, data extraction, or math. A high temperature (0.7 to 1) introduces more randomness, making responses creative and diverse, but increasing errors

#### 4. Hallucinations: 
LLMs are essentially advanced autocomplete tools; they predict what word should logically come next based on probability, not hard facts. When they confidently output false, illogical, or completely fabricated information, it is called a hallucination. They do this because they are attempting to generate plausible-sounding text rather than verifying the actual truth

Prompt engineering is a relatively new discipline for developing and optimizing prompts to efficiently use language models (LMs) for a wide variety of applications and research topics. Prompt engineering skills help to better understand the capabilities and limitations of large language models (LLMs).


#### Elements of a Prompt

A prompt contains any of the following elements:

Instruction - a specific task or instruction you want the model to perform

Context - external information or additional context that can steer the model to better responses

Input Data - the input or question that we are interested to find a response for

Output Indicator - the type or format of the output.

````python
import os
from google import genai
from google.genai import types

# 1. Initialize the client
# Ensure you have set your environment variable: export GEMINI_API_KEY="your_api_key"
client = genai.Client(api_key="")

# We use gemini-2.5-flash as the standard model for general text tasks
model_id = 'gemini-3.5-flash'

# A prompt designed to test creativity and factual boundaries
prompt = "Write a highly detailed, realistic historical fact about the invention of the internet, but invent a fictional secret society that secretly funded it."
````

````python
def generate_demo(temperature_value: float, description: str):
    print(f"\n--- Running Demo with {description} (Temperature: {temperature_value}) ---")
    
    try:
        response = client.models.generate_content(
            model=model_id,
            contents=prompt,
            config=types.GenerateContentConfig(
                temperature=temperature_value,
                # Optional: You can also limit max_output_tokens here to control the response size
                max_output_tokens=150 
            )
        )
        
        # Display the results
        print(response.text)
        
        # Display token usage metadata provided by the context window
        if response.usage_metadata:
            print("\n[Token Metrics]")
            print(f"Input Tokens: {response.usage_metadata.prompt_token_count}")
            print(f"Output Tokens: {response.usage_metadata.candidates_token_count}")
            print(f"Total Context Used: {response.usage_metadata.total_token_count}")
            
    except Exception as e:
        print(f"An error occurred: {e}")
````

````python
#  contents="Suggest an AI startup name.",
````

````python
# Test 1: Low Temperature (Deterministic, highly factual, struggles to hallucinate creatively)
generate_demo( temperature_value=0.1, description="Low Temperature / Deterministic Mode" )
````

````python
# Test 2: High Temperature (Creative, unpredictable, prone to intentional hallucinations)
generate_demo(
    temperature_value=1.3, 
    description="High Temperature / Creative (Hallucination-prone) Mode"
)
````

- Top-P Sampling

config=types.GenerateContentConfig(
    top_p=0.8
)

- Lower values → More focused output.
- Higher values → More diverse output.


* Top-K Sampling

config=types.GenerateContentConfig(
    top_k=40
)

- Lower Top-K → More deterministic.
- Higher Top-K → More varied responses.

````python
from google import genai
from google.genai import types

client = genai.Client(api_key="")

response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents="Write a short AI poem.",
    config=types.GenerateContentConfig(
        temperature=0.8,
        top_p=0.9,
        top_k=40,
        max_output_tokens=150
    )
)
print(response.text)
````

````python
# Hallucination Example
response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents="Who invented Python in 1850?"
)
print(response.text)
````

````python
# Safety-Oriented Prompt
response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents="""
 Q. how is harsh singh in ai field 
"""
)

print(response.text)
````

````python
# Safety-Oriented Prompt
response = client.models.generate_content(
    model="gemini-3.5-flash",
    contents="""
Answer using verified facts only.

If you don't know the answer,
say 'I don't know.'

Do not guess. Q. how is harsh singh in ai field 
"""
)

print(response.text)
````

| Parameter            | Purpose                               | Typical Range    |
| -------------------- | ------------------------------------- | ---------------- |
| `temperature`        | Controls creativity/randomness        | 0.0–2.0          |
| `top_p`              | Nucleus sampling (probability mass)   | 0.0–1.0          |
| `top_k`              | Number of candidate tokens considered | 1–100+           |
| `max_output_tokens`  | Maximum response length               | Positive integer |
| `system_instruction` | Defines the model's behavior and role | Text             |

# 1. Role + Task + Context + Format Framework

This is one of the simplest and most effective prompt structures.

## Structure

```
Role
↓
Task
↓
Context
↓
Output Format
```

### 1. Role

Tell the AI who it should act as.

Examples:

* Data Scientist
* Python Teacher
* Financial Analyst
* Marketing Expert

**Example**

```
You are an experienced Data Science trainer.
```

---

### 2. Task

Tell the AI exactly what you want.

Examples

* Explain
* Summarize
* Generate
* Analyze

**Example**

```
Explain Machine Learning.
```

---

### 3. Context

Provide background information.

Without context:

```
Explain Python.
```

With context:

```
Explain Python to a beginner who has never written code.
```

Better context produces better results.

Examples

```
Audience: College students

Level: Beginner

Purpose: Interview preparation

Length: 300 words
```

---

### 4. Format

Tell AI how to present the answer.

Examples

* Bullet points
* Table
* JSON
* Markdown
* Step-by-step guide
* Presentation slides

---

## Complete Example

```
Role:
You are a Data Science Trainer.

Task:
Explain Linear Regression.

Context:
The audience is non-technical working professionals learning Data Science for the first time.

Format:
Use simple English with real-life examples and bullet points.
```

---

## Output

The response will usually be:

* Beginner friendly
* Practical
* Well structured
* Easy to understand

### React
ReAct (Reasoning + Acting) prompting is an AI framework where a large language model iteratively alternates between thinking step-by-step and taking physical or digital actions to solve a task. By looping through Thought, Action, and Observation, the model limits hallucinations and continuously grounds its answers in real-world data.

1. Thought: The model reasons through the problem, identifies what it needs to know, and plans its next move.
2. Action: The model determines the best external tool to use (e.g., a calculator, search engine, or API) and formats the exact request.
3. Observation: The tool executes the action and feeds the raw results back to the model

```
# react prompt
You are a helpful AI assistant. You have access to the following tools:

{tools}

Use the following format:

Question: the input question you must answer
Thought: you should always think about what to do next
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
... (this Thought/Action/Action Input/Observation can repeat N times)
Thought: I now know the final answer
Final Answer: the final answer to the original input question

Begin!

Question: {input}
Thought: {agent_scratchpad}

```
