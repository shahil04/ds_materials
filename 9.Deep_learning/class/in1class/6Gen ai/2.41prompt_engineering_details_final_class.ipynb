# 🔹 Module 3: Prompt Engineering Fundamentals
* What is a Prompt?
* Components of a Prompt
* Context Design
* Task Instructions
* Constraints


Prompt engineering is a relatively new discipline for developing and optimizing prompts to efficiently use language models (LMs) for a wide variety of applications and research topics. Prompt engineering skills help to better understand the capabilities and limitations of large language models (LLMs).



# Why Prompt Engineering Matters

LLMs are powerful but cannot read your mind.

A vague prompt produces vague results.

A clear prompt produces better results.

---

## Example

### Bad Prompt

```text
Tell me about Python.
```

Output:

Generic explanation.

---

### Good Prompt

```text
Explain Python for beginners.

Include:
1. Definition
2. Features
3. Applications
4. Example Code
5. Interview Questions
```

Output:

Structured and useful.
