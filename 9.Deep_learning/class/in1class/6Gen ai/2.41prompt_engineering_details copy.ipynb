# What is Prompt Engineering?

## Definition

Prompt Engineering is the process of designing and optimizing prompts to obtain accurate, relevant, and useful responses from Large Language Models (LLMs).

---

## Interview Definition

Prompt Engineering is the practice of crafting effective instructions and context for an AI model to improve output quality without modifying the model’s parameters.

---

# Why Prompt Engineering Matters

LLMs are powerful but cannot read your mind.

A vague prompt produces vague results.

A clear prompt produces better results.

---

## Example

### Bad Prompt

```text
Tell me about Python.
```

Output:

Generic explanation.

---

### Good Prompt

```text
Explain Python for beginners.

Include:
1. Definition
2. Features
3. Applications
4. Example Code
5. Interview Questions
```

Output:

Structured and useful.

---

# Prompt Engineering as a Solution

Prompt Engineering helps:

* Improve output quality
* Reduce ambiguity
* Control response style
* Improve reasoning
* Generate structured outputs
* Reduce hallucinations

---

# Components of a Prompt

A good prompt contains:

## 1. Task

What should the model do?

Example:

```text
Summarize the article.
```

---

## 2. Context

Background information.

Example:

```text
This article is about climate change.
```

---

## 3. Input Data

The content to process.

Example:

```text
Article Text: ...
```

---

## 4. Output Format

Expected structure.

Example:

```text
Return output as bullet points.
```

---

# Prompt Formula

```text
Role
+
Task
+
Context
+
Instructions
+
Output Format
```

Example:

```text
You are a Data Analyst.

Analyze the sales dataset.

Provide:
1. Insights
2. Trends
3. Recommendations

Output in table format.
```


### Types of prompts
1. number of example based (zero ,one ,few shorts)
2. Based on Desired outcome (instructions based prompt, role-playing ,educational prompt, creative writing prompt)

# 1. Zero-Shot Prompting

## What?

Model receives only a task.

No examples.

---

## Example

```text
Translate:

I love AI.

To Hindi.
```

Output:

```text
मुझे AI पसंद है।
```

---

## Advantages

* Simple
* Fast
* No examples needed

---

## Disadvantages

* May produce inconsistent outputs

---

## When to Use

* Simple tasks
* General knowledge

---

# 2. One-Shot Prompting

#### What?

 - Provide one example.
---
#### Example
```text
Example:

Question:
What is Python?

Answer:
Python is a programming language.

Now answer:
What is Java?
```
---
#### Why?

Shows expected format.

# 3. Few-Shot Prompting

## What?

Provide multiple examples.

---

## Example

```text
Input: Apple
Output: Fruit

Input: Mango
Output: Fruit

Input: Carrot
Output: Vegetable

Input: Potato
Output: ?
```

Output:

```text
Vegetable
```

---

## Advantages

* Higher accuracy
* Better consistency

# 4. Role Prompting

## What?

Assign a role to the model.

---

## Example

```text
You are an experienced Python Trainer.

Explain Lists to beginners.
```

---

## Real-World Examples

```text
You are a Doctor.
```

```text
You are a Lawyer.
```

```text
You are a Financial Advisor.
```

---

## Why It Works

The model adapts its tone and expertise.

---

# 5. Chain of Thought (CoT)

## What?

Ask the model to think step-by-step.

---

## Example

```text
Solve step-by-step:

A shop sells 5 books at ₹200 each.

What is total revenue?
```

---

## Benefits

Improves reasoning.

Useful for:

* Mathematics
* Coding
* Logical tasks

---

# Example

```text
Let's think step by step.
```

Often improves accuracy.

---

# 6. Structured Output Prompting

## Problem

Normal output:

```text
John is 25 years old.
```

---

## Desired Output

```json
{
 "name":"John",
 "age":25
}
```

---

## Prompt

```text
Return output only in JSON format.
```

---

# Why Important?

Used in:

* APIs
* Automation
* Workflows
* Data Pipelines

# Advanced Prompting Techniques

## Delimiter Prompting

Separate sections clearly.

Example:

```text
###
Article:
...
###
Summarize.
```

---

## Constraint Prompting

Specify limits.

Example:

```text
Explain in exactly 50 words.
```

---

## Output Formatting Prompting

Example:

```text
Return output as:

| Product | Sales |
|----------|-------|
```

# Prompt Optimization Techniques

## Technique 1

Be Specific

Bad:

```text
Explain SQL.
```

Good:

```text
Explain SQL joins with examples.
```

---

## Technique 2

Provide Context

Bad:

```text
Write email.
```

Good:

```text
Write professional leave email to manager.
```

---

## Technique 3

Specify Audience

Example:

```text
Explain Machine Learning to a 10-year-old child.
```

---

## Technique 4

Specify Format

Example:

```text
Output in markdown table.
```

# Streamlit Project

## AI Prompt Playground

```python
import streamlit as st
from openai import OpenAI

client = OpenAI()

st.title("Prompt Engineering Playground")

user_prompt = st.text_area(
    "Enter Prompt"
)

if st.button("Generate"):

    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[
            {
                "role":"user",
                "content":user_prompt
            }
        ]
    )

    st.write(
        response.choices[0].message.content
    )
```
