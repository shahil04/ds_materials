# Types of Fine-Tuning

## 1. Full Fine-Tuning

Train all model parameters.

Example:

```text
Llama 3
70 Billion Parameters
```

Update:

```text
All 70 Billion
```

Pros:

* Best accuracy

Cons:

* Very expensive

---

## 2. PEFT

Parameter Efficient Fine-Tuning

Update only small parameters.

Pros:

* Cheap
* Fast

---

## 3. LoRA

Low Rank Adaptation

Most popular method.

Train:

```text
0.1% - 1%
```

parameters only.

Pros:

* Extremely efficient
* Industry standard

````python
# Fine-Tuning Pipeline

## Install Libraries
pip install transformers
pip install datasets
pip install peft
pip install trl
pip install accelerate
````

````python
# Open Source Example
# Fine-Tuning Llama 3 Using LoRA
## Step 1 Load Dataset

from datasets import load_dataset

dataset = load_dataset("json",
                      data_files="customer_support.json")

## Step 2 Load Model
from transformers import AutoTokenizer
from transformers import AutoModelForCausalLM

model_name = "meta-llama/Meta-Llama-3-8B"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name
)


## Step 3 Configure LoRA
from peft import LoraConfig

lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    task_type="CAUSAL_LM"
)

## Step 4 Trainer
from trl import SFTTrainer

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset["train"],
    peft_config=lora_config
)

## Step 5 Train
trainer.train()

## Step 6 Save Model
trainer.save_model("./llama_customer_support")

# Streamlit UI for Open Source Model

import streamlit as st
from transformers import pipeline

st.title("Customer Support Assistant")

model = pipeline(
    "text-generation",
    model="./llama_customer_support"
)

question = st.text_input("Ask Question")

if st.button("Generate"):
    response = model(question)
    st.write(response[0]["generated_text"])
````

# Closed Source Example

# Fine-Tuning GPT Model (OpenAI)

## Why Closed Source?

Cannot access model weights.

But OpenAI allows fine-tuning supported models.

---

# Dataset Format

JSONL

````python

# ```json
# {"messages":[
# {"role":"system","content":"You are HR assistant"},
# {"role":"user","content":"Leave policy"},
# {"role":"assistant","content":"Employees get 20 annual leaves"}
# ]}

## Upload Dataset
from openai import OpenAI

client = OpenAI()

client.files.create(
 file=open("hr_data.jsonl","rb"),
 purpose="fine-tune"
)

## Create Fine-Tuning Job
job = client.fine_tuning.jobs.create(
 training_file="file-abc123",
 model="gpt-4.1-mini"
)

## Monitor

client.fine_tuning.jobs.retrieve(job.id)

## Use Fine-Tuned Model
response = client.chat.completions.create(
 model="ft:gpt-4.1-mini:company:model"
)

# Streamlit UI for Fine-Tuned GPT
import streamlit as st
from openai import OpenAI

client = OpenAI()

st.title("HR Assistant")

question = st.text_input("Ask Question")

if st.button("Submit"):

    response = client.chat.completions.create(
        model="ft:gpt-4.1-mini:company:model",
        messages=[
            {
              "role":"user",
              "content":question
            }
        ]
    )

    st.write(
      response.choices[0].message.content
    )
````

# What is Prompt Engineering?

### Definition

Prompt Engineering is the process of designing effective prompts to guide a model toward desired outputs without changing the model itself.

### Example

Prompt:

```text
You are an expert Data Science trainer.

Explain Linear Regression using:
1. Definition
2. Formula
3. Real-world example
4. Interview questions
```

Result:

The model follows your instructions.

👉 The model weights remain unchanged.

---

# What is Fine-Tuning?

### Definition

Fine-Tuning is the process of updating a pretrained model's parameters using custom training data.

### Example

Dataset:

```json
{
  "question":"What is our company's refund policy?",
  "answer":"Refund allowed within 15 days."
}
```

Train thousands of examples.

Now the model naturally learns:

* Company terminology
* Company policies
* Response style

👉 Model weights are changed.

# Interview Question

### Is Prompt Engineering a Fine-Tuning Technique?

**Answer:**

No. Prompt Engineering is not a Fine-Tuning technique. Prompt Engineering improves model outputs by crafting better instructions, while Fine-Tuning modifies the model's weights using additional training data. Prompt Engineering does not require retraining, whereas Fine-Tuning requires a training process and updates the model parameters.

### One-Line Interview Answer

> Prompt Engineering guides the model externally through instructions, whereas Fine-Tuning changes the model internally by updating its learned parameters.
