````python
MODEL_REGISTRY = {
    "Mistral (Customer Support)": "mistralai/Mistral-7B-Instruct-v0.2",
    "LLaMA (HR Resume Analysis)": "meta-llama/Llama-2-7b-chat-hf",
    "Falcon (Document Summarization)": "tiiuae/falcon-7b-instruct",
    "Qwen (Multilingual Assistant)": "Qwen/Qwen-7B-Chat"
}
# Model Registry (All Models in One Place)

# Universal Model Loader (4-bit for Colab)
loaded_models = {}

def load_model(model_name):
    if model_name in loaded_models:
        return loaded_models[model_name]

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map="auto",
        load_in_4bit=True,
        torch_dtype=torch.float16
    )

    loaded_models[model_name] = (tokenizer, model)
    return tokenizer, model
# 💡 Teaching tip:
# Explain “We load models only once to save GPU memory.”
````

````python
# Text Generation Function
def generate_text(model_key, user_prompt, max_tokens=200):
    model_name = MODEL_REGISTRY[model_key]
    tokenizer, model = load_model(model_name)

    inputs = tokenizer(user_prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        do_sample=True,
        temperature=0.7
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
````

## 2. Ollama Open-Source Model – Python Demo

**Ollama** allows you to run open-source LLMs such as Llama, Mistral, Qwen, Gemma, and Phi locally on your machine and access them through Python. ([Ollama Docs][1])

### Step 1: Install Ollama

Download and install:

[Ollama Official Website](https://ollama.com?utm_source=chatgpt.com)

Pull a model:

```bash
ollama pull llama3
```

or

```bash
ollama pull mistral
```

Start Ollama:

```bash
ollama serve
```

---

### Step 2: Install Python Library

```bash
pip install ollama
```

The official Python library provides `chat()` and `generate()` APIs for interacting with local models. ([GitHub][2])

---

### Step 3: Simple Chat Example

```python
from ollama import chat

response = chat(
    model='llama3',
    messages=[
        {
            'role': 'user',
            'content': 'Explain AI in simple words'
        }
    ]
)

print(response['message']['content'])
```

This sends a chat request to the local Ollama server and returns the model's response. ([GitHub][2])

---

### Step 4: Text Generation Example

```python
from ollama import generate

response = generate(
    model='mistral',
    prompt='Write a short business email for project approval.'
)

print(response['response'])
```

The `generate()` API is useful for single-prompt text generation tasks. ([GitHub][2])

---

### Step 5: Business Chatbot Demo

```python
from ollama import chat

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    response = chat(
        model='llama3',
        messages=[
            {
                'role': 'system',
                'content': 'You are a professional HR assistant.'
            },
            {
                'role': 'user',
                'content': user_input
            }
        ]
    )

    print("Bot:", response['message']['content'])
```

⚖️ Open-Source vs Closed-Source LLMs    

| **Criteria**     | **Open-Source LLMs**    | **Closed-Source LLMs** |
| ---------------- | ----------------------- | ---------------------- |
| Model Weights    | Available               | Not Available          |

✅ Advantages
- No infrastructure management
- State-of-the-art performance
- Faster deployment
- Regular model updates
- Enterprise support options
⚠️ Limitations
- Ongoing API costs
- Limited customization
- Vendor dependency
- Less transparency into model internals
- Potential data governance restrictions depending on provider and plan

🎯 When to Choose Closed-Source Models

Choose a closed-source LLM when:

- You need a solution quickly.
- You do not want to manage GPUs or model hosting.
- High accuracy and reliability are priorities.
- Customization requirements are limited.
- API-based pricing fits your budget.

## 3. Closed-Source Model (ChatGPT) – Python Demo

**ChatGPT** is a closed-source LLM provided by [OpenAI](https://openai.com?utm_source=chatgpt.com). Unlike Ollama, you access the model through an API rather than running model weights locally.


````python

### Step 1: Install the OpenAI SDK

pip install openai

### Step 2: Get an API Key

# 1. Create an account at [OpenAI Platform](https://platform.openai.com?utm_source=chatgpt.com)
# 2. Generate an API key from the dashboard.
# 3. Store it securely as an environment variable.

# Windows:

# set OPENAI_API_KEY=your_api_key

### Step 3: Simple Chat Example
from openai import OpenAI

client = OpenAI()

response = client.responses.create(
    model="gpt-5",
    input="Explain Artificial Intelligence in simple words."
)

print(response.output_text)
````

````python

### Step 4: Business Assistant Example
from openai import OpenAI

client = OpenAI()

response = client.responses.create(
    model="gpt-5",
    instructions="You are a professional HR assistant.",
    input="What are the key skills required for a Data Scientist?"
)

print(response.output_text)
````

````python

### Step 5: Interactive Chatbot
from openai import OpenAI

client = OpenAI()

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        break

    response = client.responses.create(
        model="gpt-5",
        input=user_input
    )

    print("Bot:", response.output_text)
````

````python

### Step 6: Document Q&A (Simple RAG-Style Prompting)

from openai import OpenAI

client = OpenAI()

document = """
Company Leave Policy:
Employees receive 20 paid leave days annually.
"""

question = "How many paid leave days do employees receive?"

prompt = f"""
Context:
{document}

Question:
{question}

Answer only from the provided context.
"""

response = client.responses.create(
    model="gpt-5",
    input=prompt
)
print(response.output_text)
````
