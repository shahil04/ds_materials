## 2. Ollam use for open source

Ollama is an open-source tool that allows you to run Large Language Models (LLMs) locally on your own computer. When paired with Python, it provides a powerful, completely free, and private environment to build AI applications without relying on third-party cloud APIs like OpenAI or Anthropic


1. Prerequisites & Installation

* Download Ollama: Download and install the app for macOS, Linux, or Windows from the  Official Ollama Website .

- ollama pull llama3.2

* Install the Python Library: Install the official Python package via pip:

2. Core Python Implementations
The official Ollama Python GitHub Library provides two primary ways to interact with your local models:

One-Shot Text Generation Use  for quick, single-prompt tasks like summarization, drafting, or code generation.

````python
import ollama

response = ollama.generate(
    model='llama3.2',
    prompt='Write a python function to generate the Fibonacci sequence.'
)
print(response['response'])
````

````python

# Multi-Turn Chat (Streaming) Use  to manage conversational context. Setting  prints
#  the output piece-by-piece as it generates, preventing lag in your application.
import ollama

messages = [
    {'role': 'user', 'content': 'Why is the sky blue?'}
]

stream = ollama.chat(
    model='llama3.2',
    messages=messages,
    stream=True
)

for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
````

````python
!pip install -q transformers accelerate bitsandbytes sentencepiece gradio
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import gradio as gr
````

````python
MODEL_REGISTRY = {
    "Mistral (Customer Support)": "mistralai/Mistral-7B-Instruct-v0.2",
    "LLaMA (HR Resume Analysis)": "meta-llama/Llama-2-7b-chat-hf",
    "Falcon (Document Summarization)": "tiiuae/falcon-7b-instruct",
    "Qwen (Multilingual Assistant)": "Qwen/Qwen-7B-Chat"
}
# Model Registry (All Models in One Place)

# Universal Model Loader (4-bit for Colab)
loaded_models = {}

def load_model(model_name):
    if model_name in loaded_models:
        return loaded_models[model_name]

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map="auto",
        load_in_4bit=True,
        torch_dtype=torch.float16
    )

    loaded_models[model_name] = (tokenizer, model)
    return tokenizer, model
# 💡 Teaching tip:
# Explain “We load models only once to save GPU memory.”
````

````python
# Text Generation Function
def generate_text(model_key, user_prompt, max_tokens=200):
    model_name = MODEL_REGISTRY[model_key]
    tokenizer, model = load_model(model_name)

    inputs = tokenizer(user_prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        do_sample=True,
        temperature=0.7
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
````
