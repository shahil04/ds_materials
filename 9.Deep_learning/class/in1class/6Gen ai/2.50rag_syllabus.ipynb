# 14. Bonus

## Industry Tips

* Retrieval quality is often more important than model size.
* Invest time in chunking strategy design.
* Always evaluate retrieval before generation.
* Learn Agentic RAG and GraphRAG for advanced roles.
* Build production deployments, not just notebooks.

## Common Mistakes to Avoid

* Using poor chunk sizes
* Ignoring metadata
* Skipping reranking
* No evaluation framework
* Lack of monitoring
