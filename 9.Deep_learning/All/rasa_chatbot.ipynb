## **ğŸ”° 3ï¸âƒ£ Natural Language Understanding (NLU)**
NLU is responsible for **understanding user inputs** (intent & entities).  

### **Define Intents in `data/nlu.yml`**
```yaml
version: "3.0"
nlu:
- intent: greet
  examples: |
    - Hello
    - Hi
    - Hey there!

- intent: goodbye
  examples: |
    - Bye
    - See you later!
```
### **Train the Model**
```bash
rasa train
```
