```python
# drop the 
# Drop the Agency_name column
df = df.drop("Agency_name", axis=1)

# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d", errors="coerce")

# Check updated dataframe
print(df.info())
print(df.head())
```

```python
# load the model 
import joblib

# Path to the model
model_path = r"C:\Users\hp\Documents\ds_materials\9.Deep_learning\project\Rainfall Pattern Prediction\rainfall_prediction_project\notebooks\models\rf_model.joblib"

# Load the model
model = joblib.load(model_path)

# Test the model (optional)
print(model)
```

```python
def create_time_features(df, group_col='District', target='Avg_rainfall', lags=[1,2,3], windows=[7]):
    """
    Adds lag features and rolling mean features for a target column grouped by a category.
    """
    df = df.sort_values([group_col,'Date']).copy()
    
    # Lag features
    for lag in lags:
        df[f'lag_{lag}'] = df.groupby(group_col)[target].shift(lag)
    
    # Rolling mean features
    for w in windows:
        df[f'roll_mean_{w}'] = df.groupby(group_col)[target].shift(1).rolling(window=w, min_periods=1).mean().reset_index(level=0, drop=True)
    
    return df


# --- User input for prediction ---
state_input = input("Enter State: ")
district_input = input("Enter District: ")
date_input = input("Enter Date (YYYY-MM-DD): ")

# Convert to datetime
date_input = pd.to_datetime(date_input)

# Prepare input dictionary
input_dict = {
    'State': state_input,
    'District': district_input,
    'Date': date_input,
    'Day': date_input.day,
    'Month': date_input.month,
    'Year': date_input.year
}

# Create a DataFrame for input
df_input = pd.DataFrame([input_dict])

# Get historical data for this district to compute lag & rolling features
df_hist = df[df['District']==district_input].sort_values('Date')
df_combined = pd.concat([df_hist, df_input], ignore_index=True)

# Create lag and rolling features
df_combined = create_time_features(df_combined)

# Take last row as feature input
features = ['lag_1','lag_2','lag_3','roll_mean_7','Day','Month','District']
X_pred = pd.DataFrame([df_combined.iloc[-1]])[features]

# Predict rainfall
predicted_rainfall = model.predict(X_pred)[0]
print(f"\nPredicted Avg_rainfall for {district_input}, {state_input} on {date_input.date()}: {predicted_rainfall:.3f} mm")
```
