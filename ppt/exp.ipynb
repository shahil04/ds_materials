```python
import os
import json
import requests

# MAIN PPT FOLDER
main_path = r"c:\Users\hp\Documents\ppt"
print("MAIN PATH:", main_path)

# List only subfolders inside main_path
folders = [
    name for name in os.listdir(main_path)
    if os.path.isdir(os.path.join(main_path, name))
]

print("SUBFOLDERS FOUND:", folders)

# ------------------------------------------------------------
# PROCESS EACH SUBFOLDER
# ------------------------------------------------------------
for folder in folders:

    subfolder_path = os.path.join(main_path, folder)
    print("\nüìÅ Entering:", subfolder_path)

    # List items inside subfolder
    items = os.listdir(subfolder_path)

    for file in items:

        if not file.endswith(".txt"):
            continue  # skip non-txt files

        print(f"\n=== Processing TXT: {file}")

        # Create folder based on txt file name
        json_folder_name = file[:-4]  # remove .txt
        json_folder_path = os.path.join(subfolder_path, json_folder_name)

        # Create this folder
        os.makedirs(json_folder_path, exist_ok=True)
        print("Created/Exists ‚Üí", json_folder_path)

        # Read the JSON txt file
        input_file_path = os.path.join(subfolder_path, file)

        with open(input_file_path, "r", encoding="utf-8") as f:
            content = f.read().strip()
            data = json.loads(content)

        # Create the "Presentations" folder inside json_folder
        presentations_path = os.path.join(json_folder_path, "Presentations")
        os.makedirs(presentations_path, exist_ok=True)

        # Function to sanitize file/folder names
        def sanitize_filename(name):
            return "".join(
                c if c.isalnum() or c in " -_&" else "_"
                for c in name
            )

        # --------------------------------------------------
        # PROCESS JSON DATA
        # --------------------------------------------------
        for module_idx, module in enumerate(data, start=1):
            heading = sanitize_filename(module["heading"])
            module_folder = f"{module_idx:02d}_{heading}"
            module_dir = os.path.join(presentations_path, module_folder)
            os.makedirs(module_dir, exist_ok=True)

            print(f"\nüìÇ Module Folder Created: {module_folder}")

            for item_idx, item in enumerate(module["items"], start=1):
                title = sanitize_filename(item["title"])
                link = item["link"]

                if not link:
                    print(f"   ‚ö†Ô∏è SKIPPED (No Link): {title}")
                    continue

                ext = os.path.splitext(link)[1] or ".html"
                file_name = f"{item_idx:02d}_{title}{ext}"
                file_path = os.path.join(module_dir, file_name)

                print(f"   ‚¨áÔ∏è Downloading ‚Üí {title}")

                try:
                    resp = requests.get(link, timeout=20)
                    resp.raise_for_status()
                    with open(file_path, "wb") as f:
                        f.write(resp.content)
                    print(f"   ‚úÖ Saved: {file_name}")

                except Exception as e:
                    print(f"   ‚ùå Failed: {title} ({e})")

print("\nüéâ ALL DONE SUCCESSFULLY!")
```

```python
import os
import json
import requests

# Path to your .txt file (update this if needed)
# Path to your .txt file (update this)
INPUT_FILE = r"C:\Users\hp\Documents\class work or project\ds_materials\ppt\aws_itv_fsd\aws_pp.txt"
BASE_DIR = "Presentations"
# Step 1: Read the .txt file and load JSON
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    content = f.read().strip()
    data = json.loads(content)

# Step 2: Create base folder
os.makedirs(BASE_DIR, exist_ok=True)

# Helper function: sanitize names
def sanitize_filename(name):
    return "".join(c if c.isalnum() or c in " -_&" else "_" for c in name)

# Step 3: Loop through the data with sequence numbers
for module_idx, module in enumerate(data, start=1):
    heading = sanitize_filename(module["heading"]).strip()
    folder_name = f"{module_idx:02d}_{heading}"
    folder_path = os.path.join(BASE_DIR, folder_name)
    os.makedirs(folder_path, exist_ok=True)

    print(f"\nüìÅ Folder: {folder_name}")

    for item_idx, item in enumerate(module["items"], start=1):
        title = sanitize_filename(item["title"]).strip()
        link = item["link"]

        if not link:
            print(f"   ‚ö†Ô∏è Skipping '{title}' (no link)")
            continue

        # Detect or default extension
        ext = os.path.splitext(link)[1] or ".html"
        file_name = f"{item_idx:02d}_{title}{ext}"
        file_path = os.path.join(folder_path, file_name)

        print(f"   ‚¨áÔ∏è Downloading {title}...")

        try:
            resp = requests.get(link, timeout=20)
            resp.raise_for_status()
            with open(file_path, "wb") as f:
                f.write(resp.content)
            print(f"   ‚úÖ Saved as: {file_name}")
        except Exception as e:
            print(f"   ‚ùå Failed to download {title}: {e}")

print("\nüéâ All downloads completed successfully!")
```
