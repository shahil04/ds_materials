```python
# sub plot 
plt.figure()
plt.subplot(2,1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2,1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
# ploting in 4 parts
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))


plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))


plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x))


plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x+1))
plt.show()
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), color="red")

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.ylim(-1, 2)
plt.xlim(0,20)
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'blue', label = 'Sinx')
plt.plot(x, np.cos(x), color = 'red', label = 'Cosx')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.grid()
plt.title("This is sin plot before real dataset execution to understand basic syntax")
plt.xlabel("x axis")
plt.ylabel("y axis")
# plt.axis('equal')
plt.legend()
plt.show()
```

```python
#pie chart

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
my_explode = (0.0, 0.1, 0.1)

langs = ['C', 'Python', 'Java']
students = [20, 100, 40]

plt.pie(students, labels = langs, autopct = '%1.1f%%', explode = my_explode, shadow =  True)
plt.title("Pie chart")
plt.show()
```

```python
# 3 d plot

x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x, y, z, c = "#FF5733", alpha = .9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
