```python
url = "https://www.yellowpages.com/lima-oh/mip/fazolis-3954441"
page = requests.get(url)
page.status_code
```

```python
soup = BeautifulSoup(page.content,"html.parser")
```

```python
soup2 = soup.find("div", class_="carousel")
soup2
```

```python
soup2.findAll('img')
```

```python
img_url=[]
for i in soup2.findAll('img'):
    print(i["data-url"])
    img_url.append(i["data-url"])
```

```python
for i in range (0, len(img_url)):
    with open("img"+str(i)+".jpg","wb") as f:
        r=  requests.get(img_url[i])
        f.write(r.content)
```
