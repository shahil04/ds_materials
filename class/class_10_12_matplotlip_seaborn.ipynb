```python
# addd title
fig = plt.figure(figsize=(6,3))
plt.plot(np.sin(x1),"--", label='sin')  # label is use for represent the line in legend
plt.plot(np.cos(x1), label='cos')

plt.legend(loc ="upper right")  # legend is use for represent the line in graph

# add title in graph
plt.title(" sin cos graph")

# add x and y label in graph
plt.xlabel("X axis value")
plt.ylabel("Y axis value")

# define the range of x and y axis
plt.xlim(0,210)
plt.ylim(-1 ,1.5)

# tickts --use to show custom value in axis 
plt.xticks(np.arange(0,210, 10), rotation=90)
plt.show()
```

## Bar Chart

```python
# data 
cat = ["a","b","c"]
sales = [343,384,300]

# create bar chart using matplotlib
plt.figure(figsize=(4,2))

plt.bar(cat,sales)
plt.title("categories_wise sales")
plt.xlabel("categories")
plt.ylabel("sales amount")
plt.show()
```

#### Bar Chart using Seaborn

```python
import seaborn as sns 

cat = ["a","b","c"]
sales = [343,384,300]
# create bar chart using seaborn 
sns.barplot(x=cat,y=sales ,hue=cat, legend=cat, palette="Greens")
# hue--> use for color , legend --> show the label
# palette="Greens --> use for custom color 

plt.title("categories_wise sales")
plt.xlabel("categories")
plt.ylabel("sales amount")
plt.show()

# https://seaborn.pydata.org/generated/seaborn.barplot.html
```

### pie plot

```python
#pie chart
import matplotlib.pyplot as plt 
langs = ['C', 'Python', 'Java']
students = [20, 100, 40]
exp = [.12,0,0]
color =["g","r","b"]

plt.pie(students, labels = langs, autopct = '%0.2f%%',explode=exp , colors=color)
plt.title("Pie chart")
plt.show()
```
