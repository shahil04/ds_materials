```python
for i in range(len(cards)):
    print(i)
    print(cards[i].find("h2",class_="title is-5").text)
    
    a= cards[i].findAll("a")[1]["href"]
    print(a)
```

```python
soup2.find("p", class_= "location").text.strip()
```

```python
title =[]
city =[]
apply_link=[]

for i in range(len(cards)):
    print(i)
    print(cards[i].find("h2",class_="title is-5").text)
    # title 
    t = cards[i].find("h2", class_="title is-5").text
    title.append(t)
    # city
    c = cards[i].find("p", class_= "location").text.strip()
    city.append(c)
    # link
    a= cards[i].findAll("a")[1]["href"]
    apply_link.append(a)
```

```python
import pandas as pd
df = pd.DataFrame({"title": title , "city":city, "link": apply_link})
df
```

```python
# save file in csv format
df.to_csv("jd.csv")
```
