```python
# important functions -->apply 
df["customer segment"] ="premium"
df

# create new columns customer segment where store 
# fare >30  premium  
# fare >10 and fare <20 --> medium 
# fare <10 --> low 


```

```python
# sigle fare value
fare =6
if fare >=30:
    print("premium")
elif  fare >=10 and fare <30:
    print("medium")
else:
    print("low")
```

```python
def fare_c(fare):
    if fare >=30:
        return("premium")
    elif  fare >=10 and fare <30:
        return("medium")
    else:
        return("low")
```

```python
fare_c(34)
```

```python
df["Fare"].apply(fare_c)
```

```python
df["customer segment"] =  df["Fare"].apply(fare_c)
df
```

```python
# group by
df.groupby(["customer segment"]).count()
```

```python
import pandas as pd
#make dataframe is dictionary
df1 = {"key1": [2, 3, 4, 5],
    "key2": [4, 5, 6, 7],
    "key3": [2, 3, 4, 5]}
df1 = pd.DataFrame(df1)
df1
```

```python
#make dataframe is dictionary
df2= {"key4": (2, 323, 4, 5),
    "key3": (4, 4, 6, 7),
    "key6": (2, 3,24, 5)}
df2 = pd.DataFrame(df2)
df2
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
pd.merge(df1, df2, how = 'left', left_on = 'key3', right_on = 'key3')  
```
